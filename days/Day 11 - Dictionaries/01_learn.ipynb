# Day 11 — Dictionaries

> Map keys to values. The most useful Python data structure after lists.


## Creating & accessing

```python
player = {"name": "Rocket", "score": 9500, "rank": 1}
player["name"]       # "Rocket"
player["score"]      # 9500
```


````python
player = {"name": "Rocket", "score": 9500, "rank": 1}

print(player["name"])    # direct access — KeyError if missing
print(player.get("score"))       # safe — None if missing
print(player.get("level", 1))    # safe with default
````

## Adding, updating, deleting


````python
player = {"name": "Rocket"}

player["score"] = 9500          # add
player["score"] = 10000         # update
player["active"] = True
del player["active"]            # delete
removed = player.pop('score', None)  # pop with default
print(player)
````

## Looping


````python
scores = {"Rocket": 9500, "Alice": 8700, "Bob": 7200}

for key in scores:                    # keys only
    print(key)

for val in scores.values():           # values only
    print(val)

for name, score in scores.items():    # both
    print(f"{name}: {score}")
````

## Frequency counter pattern

Build a dict of how many times each item appears:


````python
words = ["the","cat","sat","on","the","mat","the","cat"]

freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

print(freq)
print(max(freq, key=freq.get))  # most common word
````

## Dict comprehension


````python
names = ["Rocket", "Alice", "Bob"]
scores = [9500, 8700, 7200]

scoreboard = {name: score for name, score in zip(names, scores)}
print(scoreboard)

# Only high scorers
elite = {k: v for k, v in scoreboard.items() if v >= 8000}
print(elite)
````
