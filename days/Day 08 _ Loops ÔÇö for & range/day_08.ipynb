# Day 08 — Loops: for & range

> **Goal:** Loop over sequences and use range() to repeat code a set number of times.


## 📖 The for loop

Use `for` when you know what you're iterating over:

```python
for item in sequence:
    # do something with item
```


````python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
````

## 📖 range()

```python
range(5)        # 0, 1, 2, 3, 4
range(1, 6)     # 1, 2, 3, 4, 5
range(0, 10, 2) # 0, 2, 4, 6, 8  (step of 2)
range(5, 0, -1) # 5, 4, 3, 2, 1  (countdown)
```


````python
for i in range(5):
    print(i)

print('---')
for i in range(1, 6):
    print(i)

print('---')
for i in range(0, 10, 2):
    print(i)
````

## 📖 enumerate() and zip()

```python
# enumerate — get index AND value
for i, fruit in enumerate(['apple','banana']):
    print(i, fruit)   # 0 apple, 1 banana

# zip — loop two lists together
for name, score in zip(['Rocket','Alice'], [95, 87]):
    print(name, score)
```


````python
fruits = ['apple', 'banana', 'cherry']
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")

names  = ['Rocket', 'Alice', 'Bob']
scores = [95, 87, 72]
for name, score in zip(names, scores):
    print(f"{name}: {score}")
````

````python
# FizzBuzz test
def fizzbuzz(n):
    result = []
    for i in range(1, n+1):
        if i % 15 == 0: result.append('FizzBuzz')
        elif i % 3 == 0: result.append('Fizz')
        elif i % 5 == 0: result.append('Buzz')
        else: result.append(str(i))
    return result

fb = fizzbuzz(15)
assert fb[2] == 'Fizz'
assert fb[4] == 'Buzz'
assert fb[14] == 'FizzBuzz'

numbers = [10, 25, 3, 47, 8, 16]
total = 0
for n in numbers: total += n
assert total == 109
print('✅ Tests passed!')
````
