## Matrix basics


```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

rows, cols = len(matrix), len(matrix[0])
print(matrix[1][2])   # 6

# Transpose
transposed = [[matrix[r][c] for r in range(rows)] for c in range(cols)]
print(transposed)
```

## Common traversal patterns


```python
# Spiral
def spiral(matrix):
    result = []
    while matrix:
        result += matrix.pop(0)
        matrix = list(zip(*matrix))[::-1]
    return result

print(spiral([[1,2,3],[4,5,6],[7,8,9]]))
```
