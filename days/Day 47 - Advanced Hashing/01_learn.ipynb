# Day 47 — Advanced Hashing

> When you need O(1) lookup, reach for a hash map.


## collections module tools


```python
from collections import Counter, defaultdict, OrderedDict

# Counter
c = Counter('aabbbcccc')
print(c)                        # Counter({'c':4,'b':3,'a':2})
print(c.most_common(2))         # [('c',4),('b',3)]
print(c - Counter('abc'))       # Counter subtraction

# defaultdict
dd = defaultdict(list)
dd['key'].append('value')       # no KeyError

# OrderedDict (LRU cache pattern)
od = OrderedDict()
od['a'] = 1
od.move_to_end('a')             # move to end
od.popitem(last=False)          # pop from front
```
