## What is a linked list?

A chain of nodes. Each node holds a value and a pointer to the next.


```python
class Node:
    def __init__(self, val):
        self.val  = val
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, val):
        if not self.head:
            self.head = Node(val); return
        cur = self.head
        while cur.next: cur = cur.next
        cur.next = Node(val)

    def to_list(self):
        result, cur = [], self.head
        while cur: result.append(cur.val); cur=cur.next
        return result

ll = LinkedList()
ll.append(1); ll.append(2); ll.append(3)
print(ll.to_list())   # [1,2,3]
```
