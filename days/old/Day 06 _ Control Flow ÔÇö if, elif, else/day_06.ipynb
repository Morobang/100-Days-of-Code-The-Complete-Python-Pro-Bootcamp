# Day 06 — Control Flow: if / elif / else

> **Goal:** Make decisions in your code based on conditions.


## 📖 if / elif / else

```python
if condition:
    # runs if condition is True
elif other_condition:
    # runs if first was False, this is True
else:
    # runs if nothing above was True
```


````python
age = 20

if age >= 18:
    print("Adult")
elif age >= 13:
    print("Teenager")
else:
    print("Child")
````

## 📖 Comparison operators

```python
==   # equal to
!=   # not equal to
>    # greater than
<    # less than
>=   # greater than or equal
<=   # less than or equal
```


````python
print(5 == 5)    # True
print(5 != 4)    # True
print(5 > 3)     # True
print("a" == "a") # True
print("A" == "a") # False — case matters
````

## 📖 Logical operators

```python
and   # both must be True
or    # at least one must be True
not   # flips True ↔ False
```


````python
age = 20
has_id = True

if age >= 18 and has_id:
    print("Entry allowed")

raining = False
cold = True
if raining or cold:
    print("Wear a jacket")
````

## 📖 Truthy and falsy

These values are **falsy** (treated as False in an if statement):
`0`, `0.0`, `""`, `[]`, `{}`, `None`, `False`

Everything else is **truthy**.


````python
name = ""
if name:
    print(f"Hello {name}")
else:
    print("No name provided")
````

## ✅ Tests


````python
def grade(score):
    if score >= 90: return 'A'
    elif score >= 80: return 'B'
    elif score >= 70: return 'C'
    elif score >= 60: return 'D'
    else: return 'F'

assert grade(95) == 'A'
assert grade(85) == 'B'
assert grade(55) == 'F'

def is_leap(year):
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

assert is_leap(2000) == True
assert is_leap(1900) == False
assert is_leap(2024) == True
assert is_leap(2023) == False
print('✅ Tests passed!')
````
