# Day 01 — Printing & Output

> **Goal:** Understand how Python outputs text to the screen.


## 📖 What is `print()`?

`print()` is a built-in Python function that displays output on the screen.

```python
print("Hello, World!")
```

Run the cell below to see it in action:


````python
print("Hello, World!")
````

## 📖 Printing multiple things

You can pass multiple values to `print()` separated by commas.
Python puts a space between them by default.

```python
print("Hello", "Rocket")
```


````python
print("Hello", "Rocket")
print("I", "am", "learning", "Python")
````

## 📖 The `sep` and `end` parameters

- `sep` — what goes **between** values (default is a space)
- `end` — what goes at the **end** of the line (default is `\n` — a new line)

```python
print("a", "b", "c", sep="-")   # a-b-c
print("Hello", end=" ")         # no newline after
print("World")                  # prints on same line
```


````python
print("a", "b", "c", sep="-")
print("Hello", end=" ")
print("World")
````
