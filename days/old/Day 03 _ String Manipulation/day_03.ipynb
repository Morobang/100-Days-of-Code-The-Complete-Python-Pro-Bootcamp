## 📖 String concatenation

Join strings together with `+`:

```python
first = "Rocket"
last = "Tshigidimisa"
full = first + " " + last
```


````python
first = "Rocket"
last = "Tshigidimisa"
full = first + " " + last
print(full)
````

## 📖 f-strings (the modern way)

Put an `f` before the quote and embed variables directly with `{}`:

```python
name = "Rocket"
age = 23
print(f"My name is {name} and I am {age} years old")
```

You can even do math inside the `{}`:


````python
name = "Rocket"
age = 23
print(f"My name is {name} and I am {age} years old")
print(f"Next year I will be {age + 1}")
print(f"Name has {len(name)} characters")
````

## 📖 String methods

```python
s = "  Hello World  "
s.upper()          # '  HELLO WORLD  '
s.lower()          # '  hello world  '
s.strip()          # 'Hello World'  — removes whitespace
s.replace('o','0') # '  Hell0 W0rld  '
s.split()          # ['Hello', 'World']
s.count("l")       # 3
```


````python
s = "  Hello World  "
print(s.upper())
print(s.lower())
print(s.strip())
print(s.replace('o','0'))
print(s.split())
print(s.count("l"))
````

## 📖 String indexing & slicing

```python
s = "Python"
s[0]     # 'P'   — first character
s[-1]    # 'n'   — last character
s[0:3]   # 'Pyt' — index 0,1,2
s[::-1]  # 'nohtyP' — reversed
```


````python
s = "Python"
print(s[0])
print(s[-1])
print(s[0:3])
print(s[2:])
print(s[::-1])
````
