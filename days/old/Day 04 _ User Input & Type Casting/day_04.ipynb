## 📖 The `input()` function

`input()` pauses the program and waits for the user to type something.
It **always returns a string**.

```python
name = input("What is your name? ")
print(f"Hello {name}!")
```

> ⚠️ In a notebook, run the cell and type in the box that appears below it.


````python
name = input("What is your name? ")
print(f"Hello {name}!")
````

## 📖 input() always returns a string

This is the most common beginner mistake:

```python
age = input("How old are you? ")  # returns "23" — a string!
age + 1   # ❌ TypeError — can't add int to str
```

Fix it by converting:

```python
age = int(input("How old are you? "))
age + 1   # ✅ works
```


````python
# Safe version — always wrap input() with the right type converter
age = int(input("How old are you? "))
print(f"Next year you will be {age + 1}")
````

## 📖 Type casting recap

```python
int("42")      # → 42
float("3.14")  # → 3.14
str(99)        # → "99"
bool(0)        # → False
```


````python
print(int("42") + 8)       # 50
print(float("3.14") * 2)   # 6.28
print(str(99) + " bottles")  # 99 bottles
````

````python
# Test the conversion logic (not the input part)
def bmi(weight, height): return round(weight / height**2, 1)
assert bmi(70, 1.75) == 22.9
assert bmi(90, 1.80) == 27.8

def seconds_to_hms(s):
    h = s // 3600
    m = (s % 3600) // 60
    sec = s % 60
    return h, m, sec
assert seconds_to_hms(3661) == (1, 1, 1)
assert seconds_to_hms(7200) == (2, 0, 0)
print('✅ Tests passed!')
````
