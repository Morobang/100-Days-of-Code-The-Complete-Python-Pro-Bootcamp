## 📖 The operators

| Operator | Symbol | Example | Result |
|----------|--------|---------|--------|
| Addition | `+` | `3 + 2` | `5` |
| Subtraction | `-` | `3 - 2` | `1` |
| Multiplication | `*` | `3 * 2` | `6` |
| Division | `/` | `7 / 2` | `3.5` |
| Floor division | `//` | `7 // 2` | `3` |
| Modulo | `%` | `7 % 2` | `1` |
| Exponent | `**` | `3 ** 2` | `9` |


````python
print(3 + 2)
print(3 - 2)
print(3 * 2)
print(7 / 2)    # always float
print(7 // 2)   # floor — drops decimal
print(7 % 2)    # remainder
print(3 ** 2)   # power
````

## 📖 Operator precedence (BODMAS)

Python follows standard maths order: `**` → `* / // %` → `+ -`

Use brackets to control the order:

```python
2 + 3 * 4      # → 14  (not 20)
(2 + 3) * 4    # → 20
```


````python
print(2 + 3 * 4)
print((2 + 3) * 4)
print(10 - 2 ** 3)
print((10 - 2) ** 3)
````

## 📖 Augmented assignment

Shorthand for updating a variable:

```python
score = 100
score += 10   # same as: score = score + 10
score -= 5    # same as: score = score - 5
score *= 2    # same as: score = score * 2
```


````python
score = 100
print(score)
score += 10
print(score)
score -= 5
print(score)
score *= 2
print(score)
````

## 📖 The `round()` function

```python
round(3.14159, 2)   # → 3.14
round(3.5)          # → 4  (rounds to nearest even)
```


````python
print(round(3.14159, 2))
print(round(7/3, 3))
````
