## 📖 What is a class?

A class is a **blueprint**. An object is an **instance** of that blueprint.

```python
class Dog:                         # blueprint
    def __init__(self, name):
        self.name = name

rex = Dog('Rex')                   # instance
```


````python
class Dog:
    def __init__(self, name, breed):
        self.name  = name     # instance attributes
        self.breed = breed

    def bark(self):
        return f'{self.name} says: Woof!'

rex  = Dog('Rex', 'Husky')
fido = Dog('Fido', 'Labrador')

print(rex.name)
print(fido.bark())
print(rex.bark())
````

## 📖 Class attributes vs instance attributes

```python
class Dog:
    species = 'Canis lupus'  # class attribute — shared by all

    def __init__(self, name):
        self.name = name     # instance attribute — unique per dog
```


````python
class Dog:
    species = 'Canis lupus'

    def __init__(self, name):
        self.name = name

rex = Dog('Rex')
print(Dog.species)    # access via class
print(rex.species)    # access via instance — also works
print(rex.name)       # only on instance
````

## 📖 `__str__` — string representation

```python
class Dog:
    def __str__(self):
        return f'Dog({self.name})'

print(Dog('Rex'))   # → Dog(Rex)
```


````python
class Dog:
    def __init__(self, name, breed):
        self.name  = name
        self.breed = breed

    def __str__(self):
        return f'Dog(name={self.name}, breed={self.breed})'

rex = Dog('Rex', 'Husky')
print(rex)
print(str(rex))
````
