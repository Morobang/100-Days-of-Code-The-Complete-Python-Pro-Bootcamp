## 📖 What is inheritance?

A child class **inherits** all attributes and methods from its parent.

```python
class Animal:           # parent
    def __init__(self, name):
        self.name = name

class Dog(Animal):      # child — inherits from Animal
    def bark(self):
        return 'Woof!'
```


````python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return '...'

    def __str__(self):
        return f'{type(self).__name__}({self.name})'

class Dog(Animal):
    def speak(self):           # override parent method
        return 'Woof!'

class Cat(Animal):
    def speak(self):
        return 'Meow!'

animals = [Dog('Rex'), Cat('Whiskers'), Dog('Fido')]
for a in animals:
    print(f'{a} says {a.speak()}')
````

## 📖 super() — calling the parent

```python
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)   # call Animal.__init__
        self.breed = breed       # add Dog-specific attr
```


````python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)   # don't repeat yourself
        self.breed = breed

    def __str__(self):
        return f'Dog({self.name}, {self.breed})'

rex = Dog('Rex', 'Husky')
print(rex)
print(rex.name)    # inherited from Animal
print(rex.breed)   # Dog-specific
````

## 📖 isinstance() and issubclass()


````python
print(isinstance(rex, Dog))     # True
print(isinstance(rex, Animal))  # True — Dog IS an Animal
print(issubclass(Dog, Animal))  # True
````
