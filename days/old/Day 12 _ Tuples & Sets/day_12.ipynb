# Day 12 — Tuples & Sets

> **Goal:** Use tuples for fixed data and sets for unique collections.


## 📖 Tuples — immutable lists

A tuple is like a list but **you can't change it** after creation.

```python
coords = (10, 20)
rgb    = (255, 128, 0)
```


````python
coords = (10, 20)
print(coords[0])   # 10
print(coords[1])   # 20
print(len(coords)) # 2

# Unpacking
x, y = coords
print(x, y)
````

## 📖 Why use tuples?

- Function returning multiple values → tuple
- Data that shouldn't change (coordinates, RGB, DB rows)
- Faster than lists and can be used as dict keys

```python
def min_max(lst):
    return min(lst), max(lst)   # returns a tuple

lo, hi = min_max([3, 1, 4, 1, 5])
```


````python
def min_max(lst):
    return min(lst), max(lst)

lo, hi = min_max([3, 1, 4, 1, 5])
print(f"Min: {lo}, Max: {hi}")

# Swap variables using tuple unpacking
a, b = 10, 20
a, b = b, a
print(a, b)   # 20 10
````

## 📖 Sets — unique, unordered collections

A set stores only **unique values** and has no order.

```python
s = {1, 2, 3, 2, 1}   # → {1, 2, 3}
```


````python
s = {1, 2, 3, 2, 1}
print(s)           # {1, 2, 3} — duplicates removed

s.add(4)
s.discard(2)       # remove, no error if missing
print(s)
print(3 in s)      # fast membership check
````

## 📖 Set operations

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

a | b    # union        → {1,2,3,4,5,6}
a & b    # intersection → {3,4}
a - b    # difference   → {1,2}
a ^ b    # sym diff     → {1,2,5,6}
```


````python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)   # union
print(a & b)   # intersection
print(a - b)   # in a but not b
print(a ^ b)   # in one but not both
````
