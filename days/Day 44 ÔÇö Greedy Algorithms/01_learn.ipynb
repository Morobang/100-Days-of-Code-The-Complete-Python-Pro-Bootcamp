# Day 44 — Greedy Algorithms

> Make the locally optimal choice at each step and hope it leads to a global optimum.


## When greedy works

Greedy works when:
- The problem has **optimal substructure**
- The **greedy choice property** holds (local best = global best)

Classic greedy: coin change (standard coins), activity selection, Huffman coding.


## Activity selection


```python
def max_activities(activities):
    # Sort by end time — classic greedy
    activities.sort(key=lambda x: x[1])
    count, last_end = 0, float('-inf')
    for start, end in activities:
        if start >= last_end:
            count += 1
            last_end = end
    return count

print(max_activities([(1,4),(3,5),(0,6),(5,7),(3,8),(5,9),(6,10),(8,11),(8,12),(2,13),(12,14)]))
```
