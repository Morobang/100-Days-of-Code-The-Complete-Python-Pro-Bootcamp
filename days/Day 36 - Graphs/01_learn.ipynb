# Day 36 — Graphs

> Graphs model relationships. Most hard interview problems are graph problems.


## Graph representation


```python
# Adjacency list — most common
graph = {
    'A': ['B','C'],
    'B': ['A','D','E'],
    'C': ['A','F'],
    'D': ['B'],
    'E': ['B'],
    'F': ['C'],
}
```

## DFS — depth first search


```python
def dfs(graph, node, visited=None):
    if visited is None: visited = set()
    visited.add(node)
    for neighbour in graph[node]:
        if neighbour not in visited:
            dfs(graph, neighbour, visited)
    return visited

print(dfs(graph, 'A'))
```

## BFS — breadth first search


```python
from collections import deque

def bfs(graph, start):
    visited = {start}
    queue   = deque([start])
    order   = []
    while queue:
        node = queue.popleft()
        order.append(node)
        for n in graph[node]:
            if n not in visited:
                visited.add(n)
                queue.append(n)
    return order

print(bfs(graph, 'A'))
```
