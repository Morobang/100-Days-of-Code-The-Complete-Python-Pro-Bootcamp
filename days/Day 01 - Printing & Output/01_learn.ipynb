## What is `print()`?

`print()` sends output to the screen. It's the first thing every Python developer learns.

```python
print("Hello, World!")
```


````python
print("Hello, World!")
````

## Printing multiple values

Pass multiple values separated by commas — Python puts a space between them.


````python
print("Name:", "Rocket")
print("Score:", 9500)
print("Active:", True)
````

## The `sep` parameter

`sep` controls what goes **between** values. Default is a space.

```python
print("a", "b", "c", sep="-")   # a-b-c
print("a", "b", "c", sep="")    # abc
```


````python
print("Python", "is", "cool", sep="-")
print("2025", "03", "21", sep="/")
print("a", "b", "c", sep="")
````

## The `end` parameter

`end` controls what goes **after** the last value. Default is a newline `\n`.

```python
print("Hello", end=" ")   # no newline — next print continues on same line
print("World!")           # Hello World!
```


````python
print("Loading", end="")
print(".", end="")
print(".", end="")
print(". Done!")
````

## Escape sequences

| Sequence | Meaning |
|----------|---------|
| `\n` | New line |
| `\t` | Tab |
| `\\` | Backslash |
| `\'` / `\"` | Quote inside string |


````python
print("Line 1\nLine 2\nLine 3")
print("Name:\tRocket")
print("He said \"hello\"")
````

## f-strings — embedding values

Put `f` before the quote, then embed variables directly with `{}`.


````python
name = "Rocket"
score = 9500
print(f"Player: {name}")
print(f"Score: {score}")
print(f"{name} scored {score} points!")
````
