## Challenge 4 — Receipt ⭐⭐⭐

Print this receipt using f-strings. Store each value in a variable first.

```
===== RECEIPT =====
Item:   Tekken 8
Price:  R1299
Player: Rocket
===================
```


````python
# write your solution
item = "Tekken 8"
price = 1299
player = "Rocket"

print("===== RECEIPT =====")
print(f"Item:   {item}")
print(f"Price:  R{price}")
print(f"Player: {player}")
print("===================")
````
