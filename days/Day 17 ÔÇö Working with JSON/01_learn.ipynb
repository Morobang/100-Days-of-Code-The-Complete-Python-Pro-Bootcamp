> JSON is the language of APIs, configs, and data exchange.


## What is JSON?

JSON looks like a Python dict. Python ↔ JSON conversions:

| Python | JSON |
|--------|------|
| `dict` | `{}` |
| `list` | `[]` |
| `str` | string |
| `int`/`float` | number |
| `True`/`False` | `true`/`false` |
| `None` | `null` |


```python
import json

player = {'name':'Rocket','score':9500,'active':True,'tag':None}

json_str = json.dumps(player)           # Python → JSON string
print(json_str)

print(json.dumps(player, indent=2))     # pretty print
```

## json.loads() — JSON string → Python


```python
import json

raw = '{"name":"Rocket","score":9500,"active":true}'

data = json.loads(raw)
print(data['name'])     # Rocket
print(type(data))       # <class 'dict'>
```

## Reading and writing JSON files


```python
import json

players = [{'name':'Rocket','score':9500},{'name':'Alice','score':8700}]

with open('/tmp/players.json','w') as f:
    json.dump(players, f, indent=2)

with open('/tmp/players.json') as f:
    loaded = json.load(f)

print(loaded[0]['name'])
```

## Transform JSON data


```python
import json

raw = '[{"name":"Rocket","score":9500},{"name":"Alice","score":7800},{"name":"Bob","score":8700}]'
players = json.loads(raw)

# Filter and sort
top = [p for p in players if p['score'] > 8000]
top.sort(key=lambda p: p['score'], reverse=True)
print(json.dumps(top, indent=2))
```
