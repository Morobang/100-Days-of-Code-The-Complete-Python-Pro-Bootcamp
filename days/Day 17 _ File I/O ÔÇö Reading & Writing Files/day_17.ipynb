# Day 17 — File I/O: Reading & Writing Files

> **Goal:** Read from and write to files — essential for any real program.


## 📖 Writing a file

```python
with open('filename.txt', 'w') as f:
    f.write('Hello World\n')
```

Modes: `'w'` = write (overwrites), `'a'` = append, `'r'` = read


````python
with open('/tmp/test.txt', 'w') as f:
    f.write('Line 1\n')
    f.write('Line 2\n')
    f.write('Line 3\n')

print('File written!')
````

## 📖 Reading a file

```python
with open('file.txt', 'r') as f:
    content = f.read()       # entire file as string
    lines = f.readlines()    # list of lines

# Or loop line by line
with open('file.txt') as f:
    for line in f:
        print(line.strip())
```


````python
with open('/tmp/test.txt', 'r') as f:
    content = f.read()
print(content)

with open('/tmp/test.txt') as f:
    lines = f.readlines()
print(lines)

with open('/tmp/test.txt') as f:
    for line in f:
        print(line.strip())
````

## 📖 Appending to a file


````python
with open('/tmp/test.txt', 'a') as f:
    f.write('Line 4 — appended\n')

with open('/tmp/test.txt') as f:
    print(f.read())
````

## 📖 Handling missing files


````python
try:
    with open('doesnt_exist.txt') as f:
        print(f.read())
except FileNotFoundError:
    print('File not found — creating it')
    with open('doesnt_exist.txt', 'w') as f:
        f.write('Created!')
````
