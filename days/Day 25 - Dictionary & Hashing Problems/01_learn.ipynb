# Day 25 — Dictionary & Hashing Problems

> A dict lookup is O(1). Use it to turn slow solutions into fast ones.


## The frequency map pattern


```python
from collections import Counter, defaultdict

# Counter
c = Counter('aabbccdd')
print(c)                    # Counter({'a':2,'b':2,...})
print(c.most_common(2))     # [('a',2),('b',2)]

# defaultdict — no KeyError
groups = defaultdict(list)
for word in ['eat','tea','tan','ate']:
    groups[tuple(sorted(word))].append(word)
print(dict(groups))
```
