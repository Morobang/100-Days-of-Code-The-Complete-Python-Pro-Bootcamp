## Async generators
```python
async def arange(n):
    for i in range(n):
        await asyncio.sleep(0)   # yield control
        yield i

async def main():
    async for i in arange(5):
        print(i)
```

## asyncio.Queue — async producer-consumer
```python
import asyncio

async def producer(q, items):
    for item in items:
        await q.put(item)
        await asyncio.sleep(0.1)
    await q.put(None)   # sentinel

async def consumer(q, results):
    while True:
        item = await q.get()
        if item is None: break
        results.append(item * 2)
```
