## What you're building

A function that converts Markdown to HTML:

| Markdown | HTML |
|----------|------|
| `# Heading` | `<h1>Heading</h1>` |
| `## Heading` | `<h2>Heading</h2>` |
| `**bold**` | `<strong>bold</strong>` |
| `*italic*` | `<em>italic</em>` |
| `` `code` `` | `<code>code</code>` |
| `[text](url)` | `<a href=url>text</a>` |
| `- item` | `<li>item</li>` |
| blank line | `<p>` paragraph break |


## Strategy

Process line by line. Apply regex substitutions in order.


```python
import re

def convert_line(line):
    # Headings
    for level in range(6,0,-1):
        if line.startswith('#'*level+' '):
            return f'<h{level}>{line[level+1:].strip()}</h{level}>'
    # Bold, italic, code, links
    line = re.sub(r'\*\*(.*?)\*\*', r'<strong>\1</strong>', line)
    line = re.sub(r'\*(.*?)\*',       r'<em>\1</em>',         line)
    line = re.sub(r'`(.*?)`',          r'<code>\1</code>',     line)
    return line

print(convert_line('# Hello World'))
print(convert_line('**bold** and *italic*'))
```
