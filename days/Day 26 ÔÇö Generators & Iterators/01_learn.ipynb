# Day 26 — Generators & Iterators

> Process data lazily — one item at a time. No memory bloat.


## yield — pause and resume


```python
def count_up(n):
    i = 0
    while i < n:
        yield i       # pause here, return i
        i += 1        # resume here next time

gen = count_up(5)
print(next(gen))  # 0
print(next(gen))  # 1
for val in count_up(5): print(val, end=' ')
```

## Infinite generators


```python
def integers_from(n=0):
    while True:
        yield n
        n += 1

gen = integers_from(10)
print([next(gen) for _ in range(5)])  # [10,11,12,13,14]
```

## Generator expressions


```python
# List — all in memory
squares_list = [x**2 for x in range(1_000_000)]

# Generator — lazy, almost no memory
squares_gen = (x**2 for x in range(1_000_000))

print(next(squares_gen))   # 0
print(sum(x for x in range(100) if x%2==0))  # 2450
```
