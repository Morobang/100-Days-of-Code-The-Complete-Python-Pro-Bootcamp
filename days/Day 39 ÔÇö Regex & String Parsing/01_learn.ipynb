# Day 39 — Regex & String Parsing

> Pattern matching. Clean messy text. Extract structured data.


## re basics


```python
import re

text = 'Call 083-111-2222 or 072-333-4444 for info'
phones = re.findall(r'\d{3}-\d{3}-\d{4}', text)
print(phones)

# Match at start
print(re.match(r'\d+', '123abc'))  # matches
print(re.match(r'\d+', 'abc123'))  # None

# Search anywhere
print(re.search(r'\d+', 'abc123').group())  # '123'
```

## Groups, substitution, flags


```python
import re

date = '2025-03-21'
m = re.search(r'(\d{4})-(\d{2})-(\d{2})', date)
year, month, day = m.groups()
print(year, month, day)

# Substitute
cleaned = re.sub(r'\s+', ' ', 'too   many   spaces')
print(cleaned)

# Case insensitive
print(re.findall(r'hello', 'Hello HELLO hello', re.IGNORECASE))
```
