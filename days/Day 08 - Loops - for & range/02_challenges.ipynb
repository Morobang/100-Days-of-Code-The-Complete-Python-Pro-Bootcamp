````python
def fizzbuzz(n):
    result = []
    for i in range(1, n+1):
        if i%15==0: result.append('FizzBuzz')
        elif i%3==0: result.append('Fizz')
        elif i%5==0: result.append('Buzz')
        else: result.append(str(i))
    return result

fb = fizzbuzz(15)
assert fb[2] == 'Fizz'
assert fb[4] == 'Buzz'
assert fb[14] == 'FizzBuzz'
print('✅ FizzBuzz passed!')
````
