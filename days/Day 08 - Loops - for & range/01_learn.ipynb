# Day 08 — Loops: for & range

> Use `for` when you know what you're iterating over.


## Basic for loop


````python
games = ["Tekken 8", "FC26", "KOF XV"]
for game in games:
    print(f"Playing: {game}")
````

## range()

```python
range(5)          # 0,1,2,3,4
range(1, 6)       # 1,2,3,4,5
range(0, 10, 2)   # 0,2,4,6,8
range(5, 0, -1)   # 5,4,3,2,1
```


````python
for i in range(5):
    print(i, end=' ')
print()

for i in range(1, 11):
    print(f"7 x {i} = {7*i}")
````

## enumerate() — index + value


````python
players = ["Rocket", "Alice", "Bob"]
for i, name in enumerate(players, 1):
    print(f"{i}. {name}")
````

## zip() — two lists together


````python
names  = ["Rocket", "Alice", "Bob"]
scores = [9500, 8700, 7200]
for name, score in zip(names, scores):
    print(f"{name}: {score}")
````
