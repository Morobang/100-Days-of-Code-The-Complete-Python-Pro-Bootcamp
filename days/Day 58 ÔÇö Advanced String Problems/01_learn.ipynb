## KMP substring search


```python
def kmp_search(text, pattern):
    '''Find all occurrences of pattern in text. O(n+m)'''
    n, m = len(text), len(pattern)
    if m == 0: return []

    # Build failure function
    fail = [0] * m
    j = 0
    for i in range(1, m):
        while j and pattern[i] != pattern[j]: j = fail[j-1]
        if pattern[i] == pattern[j]: j += 1
        fail[i] = j

    # Search
    matches, j = [], 0
    for i, c in enumerate(text):
        while j and c != pattern[j]: j = fail[j-1]
        if c == pattern[j]: j += 1
        if j == m: matches.append(i-m+1); j = fail[j-1]
    return matches

print(kmp_search('ababcabcabababd','ababd'))
```
