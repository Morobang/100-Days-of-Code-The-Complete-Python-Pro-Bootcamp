## Challenge 1 — Grade calculator ⭐⭐

Write `grade(score)` → 'A' (90+), 'B' (80+), 'C' (70+), 'D' (60+), 'F' (below 60).


```python
def grade(score: int) -> str:
    # write your solution
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'
```

## Challenge 2 — Leap year ⭐⭐

Write `is_leap(year)` → True if leap year.
Rule: divisible by 4, EXCEPT centuries unless also divisible by 400.


```python
def is_leap(year: int) -> bool:
    # write your solution
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
        else:
            return True

    else:
        return False
```

## Challenge 3 — Rock Paper Scissors ⭐⭐⭐

Write `rps(player, computer)` → 'Win', 'Lose', or 'Draw'.


```python
def rps(player: str, computer: str) -> str:
    if player == computer:
        return "Draw"
    
    if (
        (player == "rock" and computer == "scissors") or
        (player == "paper" and computer == "rock") or
        (player == "scissors" and computer == "paper")
    ):
        return "Win"
    
    return "Lose"
```

## ✅ Tests


```python
assert grade(95) == 'A'
assert grade(83) == 'B'
assert grade(55) == 'F'

assert is_leap(2000) == True
assert is_leap(1900) == False
assert is_leap(2024) == True
assert is_leap(2023) == False

assert rps('rock', 'scissors') == 'Win'
assert rps('rock', 'rock') == 'Draw'
assert rps('rock', 'paper') == 'Lose'
print('✅ All tests passed!')
```
