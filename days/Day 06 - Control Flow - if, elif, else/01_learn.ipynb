## if / elif / else

```python
if condition:
    ...
elif other:
    ...
else:
    ...
```


````python
score = 85

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'F'

print(grade)
````

## Comparison operators

```
==  !=  >  <  >=  <=
```


````python
print(5 == 5)      # True
print("A" == "a") # False â€” case sensitive
print(10 >= 10)    # True
````

## Logical operators

```python
and   # both True
or    # at least one True
not   # flip
```


````python
age = 20
has_id = True

if age >= 18 and has_id:
    print('Entry allowed')

raining = False
cold = True
if raining or cold:
    print('Wear a jacket')
````

## Truthy and falsy

Falsy: `0`, `0.0`, `""`, `[]`, `{}`, `None`, `False`
Everything else is truthy.


````python
name = ""
if name:
    print(f'Hello {name}')
else:
    print('No name provided')
````

## Ternary expression

```python
label = "even" if x % 2 == 0 else "odd"
```


````python
x = 7
label = "even" if x % 2 == 0 else "odd"
print(label)
````
