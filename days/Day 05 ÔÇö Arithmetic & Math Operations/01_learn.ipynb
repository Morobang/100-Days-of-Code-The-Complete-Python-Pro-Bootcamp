# Day 05 — Arithmetic & Math Operations

> Python as a calculator — plus a few tricks you'll use every day.


## The operators

| Op | Symbol | Example | Result |
|----|--------|---------|--------|
| Add | `+` | `3+2` | `5` |
| Subtract | `-` | `3-2` | `1` |
| Multiply | `*` | `3*2` | `6` |
| Divide | `/` | `7/2` | `3.5` |
| Floor div | `//` | `7//2` | `3` |
| Modulo | `%` | `7%2` | `1` |
| Power | `**` | `3**2` | `9` |


```python
print(7 / 2)    # 3.5  — always float
print(7 // 2)   # 3    — floor, drops decimal
print(7 % 2)    # 1    — remainder
print(2 ** 10)  # 1024 — power
```

## Operator precedence (BODMAS)

`**` → `* / // %` → `+ -`


```python
print(2 + 3 * 4)     # 14 (not 20)
print((2 + 3) * 4)   # 20
print(10 - 2 ** 3)   # 2
```

## Augmented assignment


```python
score = 100
score += 50    # score = score + 50 = 150
score *= 2     # score = score * 2  = 300
score -= 100   # score = score - 100 = 200
print(score)
```

## Useful built-ins


```python
print(round(3.14159, 2))   # 3.14
print(abs(-42))            # 42
print(max(1, 9, 3, 7))     # 9
print(min(1, 9, 3, 7))     # 1
print(divmod(17, 5))       # (3, 2)
```
