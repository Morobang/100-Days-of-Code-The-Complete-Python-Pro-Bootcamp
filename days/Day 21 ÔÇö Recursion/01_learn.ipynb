# Day 21 — Recursion

> A function that calls itself. Every recursive solution needs a BASE CASE.


## The two rules

1. **Base case** — when to stop
2. **Recursive case** — call itself with a smaller problem


```python
def countdown(n):
    if n <= 0:              # base case
        print('Blast off!')
    else:                  # recursive case
        print(n)
        countdown(n - 1)

countdown(5)
```

## Factorial


```python
def factorial(n):
    if n == 0: return 1          # base case: 0! = 1
    return n * factorial(n - 1)  # recursive case

print(factorial(5))   # 120
print(factorial(0))   # 1
```

## Fibonacci — naive vs memoized


```python
from functools import lru_cache

# Naive — O(2^n) — very slow
def fib(n):
    if n <= 1: return n
    return fib(n-1) + fib(n-2)

# Memoized — O(n)
@lru_cache(maxsize=None)
def fib_fast(n):
    if n <= 1: return n
    return fib_fast(n-1) + fib_fast(n-2)

print(fib_fast(50))
```
