## 📖 What is a dictionary?

A dictionary maps **keys** to **values**. Keys are unique.

```python
player = {"name": "Rocket", "game": "Tekken 8", "score": 9500}
```


````python
player = {"name": "Rocket", "game": "Tekken 8", "score": 9500}

print(player["name"])    # "Rocket"
print(player["score"])   # 9500
print(len(player))       # 3
````

## 📖 Adding, updating, deleting

```python
d["new_key"] = value    # add or update
del d["key"]            # delete
d.pop("key", None)      # pop with default
```


````python
player = {"name": "Rocket", "score": 9500}

player["level"] = 42          # add
player["score"] = 10000       # update
del player["level"]           # delete
print(player)
````

## 📖 Safe access with .get()

```python
d["missing"]           # ❌ KeyError
d.get("missing")       # ✅ returns None
d.get("missing", 0)    # ✅ returns 0
```


````python
player = {"name": "Rocket", "score": 9500}

print(player.get("score"))          # 9500
print(player.get("level"))          # None
print(player.get("level", 1))       # 1
````

## 📖 Looping through a dictionary

```python
for key in d:              # keys only
for key in d.keys():       # same
for val in d.values():     # values only
for key, val in d.items(): # both
```


````python
scores = {"Rocket": 9500, "Alice": 8700, "Bob": 7200}

for name, score in scores.items():
    print(f"{name}: {score}")
````
