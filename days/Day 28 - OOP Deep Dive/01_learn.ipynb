## Properties — controlled attribute access


```python
class Circle:
    def __init__(self, radius):
        self._radius = radius   # _ = private by convention

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if value < 0: raise ValueError('Radius must be positive')
        self._radius = value

    @property
    def area(self):   # computed, read-only
        import math; return round(math.pi*self._radius**2, 2)

c = Circle(5)
print(c.area)     # 78.54
c.radius = 10
print(c.area)     # 314.16
```
