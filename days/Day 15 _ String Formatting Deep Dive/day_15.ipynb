## 📖 f-string format spec

Inside `{}` you can add a `:` followed by a format spec:

```python
f'{value:format_spec}'
```


````python
pi = 3.14159265

print(f'{pi:.2f}')      # 2 decimal places → 3.14
print(f'{pi:.4f}')      # 4 decimal places → 3.1416
print(f'{pi:10.2f}')    # width 10, 2dp    →       3.14
print(f'{1000000:,}')   # comma separator  → 1,000,000
print(f'{0.85:.1%}')    # percentage       → 85.0%
````

## 📖 Alignment

```python
f'{value:<10}'   # left align,  width 10
f'{value:>10}'   # right align, width 10
f'{value:^10}'   # centre,      width 10
f'{value:*^10}'  # centre with * fill
```


````python
items = ['Apple', 'Banana', 'Kiwi']
prices = [1.50, 2.99, 0.89]

print(f'{'Item':<12} {'Price':>8}')
print('-' * 20)
for item, price in zip(items, prices):
    print(f'{item:<12} R{price:>7.2f}')
````

## 📖 Numbers

```python
f'{255:b}'    # binary  → 11111111
f'{255:o}'    # octal   → 377
f'{255:x}'    # hex     → ff
f'{255:X}'    # HEX     → FF
f'{1e6:.2e}'  # sci notation → 1.00e+06
```


````python
print(f'{255:b}')      # binary
print(f'{255:x}')      # hex lowercase
print(f'{255:X}')      # hex uppercase
print(f'{1234567:.2e}') # scientific notation
````
