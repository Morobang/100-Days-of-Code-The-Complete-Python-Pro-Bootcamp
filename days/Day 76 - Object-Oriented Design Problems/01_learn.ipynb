# Day 76 — Object-Oriented Design Problems

> 'Design a parking lot' is a real interview question. Know how to approach it.


## SOLID principles (quick reference)

| | Principle | Meaning |
|-|-----------|--------|
| S | Single Responsibility | One class, one job |
| O | Open/Closed | Open for extension, closed for modification |
| L | Liskov Substitution | Subclass can replace parent |
| I | Interface Segregation | Small, specific interfaces |
| D | Dependency Inversion | Depend on abstractions, not concretions |


## OOP design approach

1. Identify the nouns → classes
2. Identify the verbs → methods
3. Identify relationships → inheritance / composition
4. Define interfaces → what can each object do?
