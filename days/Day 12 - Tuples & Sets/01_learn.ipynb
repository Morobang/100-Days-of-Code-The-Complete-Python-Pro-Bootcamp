# Day 12 — Tuples & Sets

> Tuples = immutable lists. Sets = unordered unique collections.


## Tuples — immutable sequences


```python
coords = (10, 20)
rgb    = (255, 128, 0)

print(coords[0])    # 10
print(len(coords))  # 2

# Unpacking
x, y = coords
print(x, y)

# Starred unpacking
first, *rest = (1, 2, 3, 4, 5)
print(first, rest)   # 1  [2,3,4,5]
```

## Why tuples?

- Return multiple values from a function
- Keys in a dictionary (lists can't be keys)
- Faster than lists for fixed data


```python
def min_max(lst):
    return min(lst), max(lst)   # returns a tuple

lo, hi = min_max([3,1,4,1,5])
print(f"Min: {lo}, Max: {hi}")

# Tuple as dict key — coordinates
grid = {(0,0): 'start', (5,5): 'end'}
print(grid[(0,0)])
```

## Swap with tuples


```python
a, b = 10, 20
print(f'Before: a={a}, b={b}')
a, b = b, a
print(f'After:  a={a}, b={b}')
```

## Sets — unique, unordered


```python
s = {1, 2, 3, 2, 1}   # duplicates removed
print(s)              # {1, 2, 3}

s.add(4)
s.discard(99)         # no error if missing
print(3 in s)         # True — O(1) lookup!
```

## Set operations


```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)   # union        {1,2,3,4,5,6}
print(a & b)   # intersection {3,4}
print(a - b)   # difference   {1,2}
print(a ^ b)   # sym diff     {1,2,5,6}
```
