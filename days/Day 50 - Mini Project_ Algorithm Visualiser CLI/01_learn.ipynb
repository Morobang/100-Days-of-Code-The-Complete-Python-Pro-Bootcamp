Build a CLI tool that shows sorting algorithms step-by-step in the terminal.


## What it looks like

```
=== Bubble Sort ===
Step 1: [64, 34, 25, 12, 22]
         ^^
Step 2: [34, 64, 25, 12, 22]
             ^^
...
Sorted: [12, 22, 25, 34, 64]
Steps: 10 | Comparisons: 10
```


## Architecture

- Each algorithm yields `(list_state, highlighted_indices)` at each step
- Renderer prints the current state with `^` markers
- Stats tracker counts comparisons and swaps


````python
def bubble_sort_steps(lst):
    """Yield (current_list, (i, j)) for each comparison step."""
    lst = lst[:]
    n = len(lst)
    for i in range(n):
        for j in range(n-i-1):
            yield lst[:], (j, j+1)  # show comparison
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j]
                yield lst[:], (j, j+1)  # show swap
    yield lst[:], ()  # final state
````
