> Beyond comparison sort. When you need O(n) or special order.


## Quickselect — O(n) average kth element


```python
import random

def quickselect(lst, k):
    pivot = random.choice(lst)
    low  = [x for x in lst if x < pivot]
    mid  = [x for x in lst if x == pivot]
    high = [x for x in lst if x > pivot]
    if k < len(low):              return quickselect(low, k)
    elif k < len(low)+len(mid):   return pivot
    else:                         return quickselect(high, k-len(low)-len(mid))

print(quickselect([3,1,4,1,5,9,2,6], 4))  # 5th smallest
```

## Counting sort — O(n+k)


```python
def counting_sort(lst):
    if not lst: return lst
    mn, mx = min(lst), max(lst)
    count  = [0] * (mx - mn + 1)
    for x in lst: count[x - mn] += 1
    result = []
    for i, c in enumerate(count):
        result.extend([i + mn] * c)
    return result

print(counting_sort([4, 2, 2, 8, 3, 3, 1]))
```
