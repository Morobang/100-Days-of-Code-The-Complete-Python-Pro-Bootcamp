# Day 55 — Type Hints & Dataclasses

> Modern Python. Type hints make code self-documenting.


## Type hints basics


```python
from typing import List, Dict, Optional, Union, Tuple, Any

def greet(name: str) -> str:
    return f'Hello {name}'

def process(items: List[int]) -> Dict[str, int]:
    return {'sum': sum(items), 'max': max(items)}

def find(lst: List[int], target: int) -> Optional[int]:
    try: return lst.index(target)
    except ValueError: return None
```

## Dataclasses


```python
from dataclasses import dataclass, field

@dataclass
class Player:
    name: str
    score: int = 0
    games: list = field(default_factory=list)

    def add_game(self, game: str) -> None:
        self.games.append(game)

p = Player('Rocket')
p.add_game('Tekken 8')
print(p)   # Player(name='Rocket', score=0, games=['Tekken 8'])
```

## Frozen dataclass (immutable)


```python
from dataclasses import dataclass

@dataclass(frozen=True)  # immutable
class Point:
    x: float
    y: float

    def distance_to(self, other: 'Point') -> float:
        return ((self.x-other.x)**2 + (self.y-other.y)**2)**0.5

p1, p2 = Point(0, 0), Point(3, 4)
print(p1.distance_to(p2))  # 5.0
```
