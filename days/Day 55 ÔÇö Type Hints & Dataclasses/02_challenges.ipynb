## 1. Type-annotated functions ⭐

Rewrite these functions with full type hints:
- `def clamp(n, lo, hi)` → int clamped to [lo, hi]
- `def flatten(lst)` → flattened list
- `def group_by(lst, key_fn)` → dict of lists


```python
from typing import List, Dict, Callable, TypeVar
T = TypeVar('T')
K = TypeVar('K')

def clamp(n: int, lo: int, hi: int) -> int:
    return max(lo, min(hi, n))

def flatten(lst: List[List[T]]) -> List[T]:
    return [x for sub in lst for x in sub]

def group_by(lst: List[T], key_fn: Callable[[T], K]) -> Dict[K, List[T]]:
    result: Dict[K, List[T]] = {}
    for item in lst: result.setdefault(key_fn(item), []).append(item)
    return result

# Add type hints to the function signatures above
```
