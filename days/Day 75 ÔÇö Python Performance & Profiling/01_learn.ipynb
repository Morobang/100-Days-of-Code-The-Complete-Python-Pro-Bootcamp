> Write fast Python. Know where time is actually spent.


## timeit — measure small code


```python
import timeit

# Compare list vs set membership
lst = list(range(10000))
s   = set(range(10000))

t_list = timeit.timeit('9999 in lst', globals=globals(), number=10000)
t_set  = timeit.timeit('9999 in s',   globals=globals(), number=10000)
print(f'List: {t_list:.4f}s')
print(f'Set:  {t_set:.4f}s')
print(f'Set is {t_list/t_set:.0f}x faster')
```

## cProfile — find the bottleneck


```python
import cProfile

def slow_fn():
    return sum(i**2 for i in range(100000))

cProfile.run('slow_fn()')
```

## Memory efficiency


```python
import sys

lst = list(range(1000))
gen = (i for i in range(1000))

print(f'List size: {sys.getsizeof(lst):,} bytes')
print(f'Gen size:  {sys.getsizeof(gen):,} bytes')

# Slots reduce per-instance overhead
class WithSlots:
    __slots__ = ['x', 'y']
    def __init__(self, x, y): self.x=x; self.y=y

class WithoutSlots:
    def __init__(self, x, y): self.x=x; self.y=y

a = WithSlots(1,2)
b = WithoutSlots(1,2)
print(f'With slots:    {sys.getsizeof(a)} bytes')
print(f'Without slots: {sys.getsizeof(b)} bytes')
```
