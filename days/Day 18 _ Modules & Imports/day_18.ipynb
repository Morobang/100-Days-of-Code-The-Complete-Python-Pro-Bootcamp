## 📖 Importing modules

```python
import math                  # import the whole module
from math import pi, sqrt    # import specific names
import math as m             # alias
```


````python
import math

print(math.pi)
print(math.sqrt(16))
print(math.ceil(4.2))    # round up
print(math.floor(4.8))   # round down
print(math.factorial(5)) # 5! = 120
````

## 📖 The random module


````python
import random

print(random.randint(1, 10))         # int between 1 and 10
print(random.random())               # float 0.0–1.0
print(random.choice(['a','b','c']))  # pick one
lst = [1,2,3,4,5]
random.shuffle(lst)
print(lst)
print(random.sample(lst, 3))         # 3 unique picks
````

## 📖 The datetime module


````python
from datetime import datetime, date, timedelta

now = datetime.now()
print(now)
print(now.year, now.month, now.day)
print(now.strftime('%Y-%m-%d %H:%M'))  # format as string

# Date arithmetic
birthday = date(2002, 3, 15)
today    = date.today()
diff     = today - birthday
print(f'{diff.days} days old')
````

## 📖 The os module


````python
import os

print(os.getcwd())              # current directory
print(os.listdir('.'))          # files in current dir
print(os.path.exists('/tmp'))   # does path exist?
print(os.path.join('folder', 'file.txt'))  # safe path join
````
