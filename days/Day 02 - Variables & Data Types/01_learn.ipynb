## What is a variable?

A variable is a **named box** that holds a value.

```python
name = "Rocket"   # str
age  = 23         # int
gpa  = 3.8        # float
pro  = True       # bool
```

- `=` is **assignment** — stores the right side into the left side name
- Python figures out the type automatically


````python
name = "Rocket"
age  = 23
gpa  = 3.8
pro  = True

print(name, age, gpa, pro)
````

## The 4 basic types

| Type | Example | Meaning |
|------|---------|--------|
| `str` | `"hello"` | text |
| `int` | `23` | whole number |
| `float` | `3.8` | decimal number |
| `bool` | `True` / `False` | yes or no |

Use `type()` to check:


````python
print(type("Rocket"))   # <class 'str'>
print(type(23))         # <class 'int'>
print(type(3.8))        # <class 'float'>
print(type(True))       # <class 'bool'>
````

## Type conversion

```python
int("42")      # → 42
float("3.14")  # → 3.14
str(99)        # → "99"
bool(0)        # → False
bool("hello")  # → True  (non-empty string!)
```


````python
print(int("42") + 8)        # 50
print(float("3.14") * 2)    # 6.28
print(str(99) + " bottles") # 99 bottles
print(bool(0), bool(1))     # False True
print(bool(""), bool("hi")) # False True
````

## Variable naming rules

```python
player_name = 'ok'    # ✅ snake_case — Python standard
playerName  = 'ok'    # ✅ camelCase — avoid in Python
2name       = 'bad'   # ❌ can't start with a number
my-var      = 'bad'   # ❌ hyphens not allowed
class       = 'bad'   # ❌ reserved keyword
```


````python
# Good naming
player_name   = 'Rocket'
high_score    = 9500
is_logged_in  = True
max_lives     = 3

print(player_name, high_score)
````

## Multiple assignment


````python
# Assign multiple at once
x, y, z = 1, 2, 3
print(x, y, z)

# Swap without a temp variable
a, b = 10, 20
a, b = b, a
print(a, b)   # 20 10
````
