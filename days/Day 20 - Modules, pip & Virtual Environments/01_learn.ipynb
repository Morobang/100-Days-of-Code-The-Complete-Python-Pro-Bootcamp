## Importing modules


````python
import math
from datetime import datetime, date, timedelta
import random
import os

print(math.pi)
print(math.sqrt(16))
print(datetime.now().strftime('%Y-%m-%d %H:%M'))
print(random.randint(1, 10))
print(os.getcwd())
````

## random module


````python
import random

print(random.randint(1, 10))        # int 1–10
print(random.random())              # float 0.0–1.0
print(random.choice(['a','b','c'])) # pick one
lst = [1,2,3,4,5]
random.shuffle(lst)
print(lst)
print(random.sample(lst, 3))        # 3 unique picks
````

## datetime module


````python
from datetime import date, timedelta

today = date.today()
print(today)
print(today.year, today.month, today.day)

birthday = date(2002, 3, 15)
age_days = (today - birthday).days
print(f'{age_days} days old')

future = today + timedelta(days=30)
print(f'30 days from now: {future}')
````

## Virtual environments & pip

```bash
# Create
python -m venv venv

# Activate
source venv/bin/activate        # Mac/Linux
venv\Scripts\activate           # Windows

# Install
pip install requests
pip freeze > requirements.txt

# Deactivate
deactivate
```

**Always add `venv/` to `.gitignore`. Never commit it.**
