# Day 80 — Interview Prep: Hard Problems IV



## The complete algorithm playbook

| Pattern | Trigger words |
|---------|--------------|
| Two pointers | Sorted array, pair sum, palindrome |
| Sliding window | Substring, subarray with condition |
| Binary search | Sorted, O(log n), find minimum satisfying |
| BFS | Shortest path, level-order |
| DFS + backtracking | All combinations, permutations |
| DP | Count ways, min/max cost, optimal choice |
| Heap | Kth largest, merge k sorted |
| Union-Find | Connected components, cycle detection |
| Topological sort | Dependencies, ordering |
| Monotonic stack | Next greater, largest rectangle |
