## Bitwise operators

| Op | Symbol | Example | Result |
|----|--------|---------|--------|
| AND | `&` | `5 & 3` | `1` |
| OR | `\|` | `5 \| 3` | `7` |
| XOR | `^` | `5 ^ 3` | `6` |
| NOT | `~` | `~5` | `-6` |
| Left shift | `<<` | `1 << 3` | `8` |
| Right shift | `>>` | `8 >> 2` | `2` |


```python
# Common bit tricks
n & 1              # check if odd (last bit is 1)
n & (n-1)          # clear lowest set bit (0 if power of 2)
n ^ n              # always 0
n ^ 0              # always n
a ^ b ^ a          # gives b  (XOR cancels)
1 << k             # 2^k
n >> k             # divide by 2^k

print(5 & 3)       # 1   (101 & 011 = 001)
print(5 | 3)       # 7   (101 | 011 = 111)
print(5 ^ 3)       # 6   (101 ^ 011 = 110)
```
