# Day 54 — Testing with pytest

> Code without tests is legacy code waiting to happen.


## Basic pytest


```python
# test_calculator.py

def add(a, b): return a + b

def test_add_positive():
    assert add(1, 2) == 3

def test_add_negative():
    assert add(-1, -2) == -3

def test_add_zero():
    assert add(0, 5) == 5

# Run with: pytest test_calculator.py -v
```

## Fixtures and parametrize


```python
import pytest

@pytest.fixture
def sample_list():
    return [1, 2, 3, 4, 5]

def test_sum(sample_list):
    assert sum(sample_list) == 15

@pytest.mark.parametrize('a,b,expected', [
    (1, 2, 3),
    (-1, 1, 0),
    (0, 0, 0),
])
def test_add_param(a, b, expected):
    assert add(a, b) == expected
```

## Testing exceptions and mocks


```python
import pytest
from unittest.mock import patch, MagicMock

def test_raises_error():
    with pytest.raises(ValueError):
        raise ValueError('oops')

@patch('time.sleep')
def test_with_mock(mock_sleep):
    import time
    time.sleep(10)   # mocked — doesn't actually sleep
    mock_sleep.assert_called_once_with(10)
```
