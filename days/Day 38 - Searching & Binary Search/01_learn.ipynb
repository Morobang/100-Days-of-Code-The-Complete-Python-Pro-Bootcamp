# Day 38 — Searching & Binary Search

> If the input is sorted, think binary search first.


## Binary search template


```python
def binary_search(lst, target):
    lo, hi = 0, len(lst)-1
    while lo <= hi:
        mid = (lo + hi) // 2
        if lst[mid] == target:   return mid
        elif lst[mid] < target:  lo = mid + 1
        else:                    hi = mid - 1
    return -1

print(binary_search([1,3,5,7,9,11], 7))  # 3
print(binary_search([1,3,5,7,9,11], 4))  # -1
```

## Find first/last position


```python
def find_first(lst, target):
    lo, hi, result = 0, len(lst)-1, -1
    while lo <= hi:
        mid = (lo+hi)//2
        if lst[mid] == target: result=mid; hi=mid-1  # keep going left
        elif lst[mid] < target: lo=mid+1
        else: hi=mid-1
    return result

print(find_first([1,2,2,2,3,4],2))  # 1
```
