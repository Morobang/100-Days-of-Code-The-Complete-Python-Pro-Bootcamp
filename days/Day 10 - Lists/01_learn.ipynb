# Day 10 — Lists

> Ordered, mutable collections — the most used data structure.


## Creating and accessing


````python
games = ["Tekken 8", "FC26", "KOF XV", "Elden Ring"]

print(games[0])    # first
print(games[-1])   # last
print(len(games))  # 4
print("FC26" in games)  # True
````

## Modifying lists


````python
games = ["Tekken 8", "FC26"]

games.append("KOF XV")         # add to end
games.insert(1, "Mortal Kombat") # insert at index
games.remove('FC26')           # remove by value
last = games.pop()             # remove & return last
print(games)
````

## Slicing

```python
lst[1:3]   # index 1,2
lst[:2]    # first 2
lst[2:]    # from index 2 to end
lst[::-1]  # reversed
```


````python
nums = [0, 1, 2, 3, 4, 5]
print(nums[1:4])    # [1, 2, 3]
print(nums[:3])     # [0, 1, 2]
print(nums[::2])    # [0, 2, 4]
print(nums[::-1])   # [5, 4, 3, 2, 1, 0]
````

## sort() vs sorted()


````python
lst = [3,1,4,1,5,9,2,6]

# sorted() — returns NEW list
new = sorted(lst)
print(new, lst)   # lst unchanged

# sort() — modifies IN PLACE
lst.sort()
print(lst)
````

## Copying — important gotcha!


````python
original = [1, 2, 3]

bad_copy  = original       # SAME object!
good_copy = original.copy() # new list

bad_copy.append(99)
print(original)   # [1, 2, 3, 99] — oops!
good_copy.append(99)
print(original)   # [1, 2, 3, 99] — untouched
````
