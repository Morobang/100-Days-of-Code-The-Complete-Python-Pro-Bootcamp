# Day 32 — Stacks, Queues & Deques

> Simple structures, surprisingly powerful in interviews.


## Stack — LIFO (last in, first out)


```python
stack = []
stack.append(1)   # push
stack.append(2)
stack.append(3)
print(stack.pop())   # 3 — LIFO
print(stack[-1])     # peek without popping
```

## Queue — FIFO (first in, first out)


```python
from collections import deque

queue = deque()
queue.append('a')    # enqueue
queue.append('b')
queue.append('c')
print(queue.popleft())  # 'a' — FIFO
```

## Deque — fast at both ends


```python
from collections import deque

d = deque([1,2,3])
d.appendleft(0)    # add to front
d.append(4)        # add to back
d.popleft()        # remove from front
d.pop()            # remove from back
print(d)           # deque([1,2,3])
```
