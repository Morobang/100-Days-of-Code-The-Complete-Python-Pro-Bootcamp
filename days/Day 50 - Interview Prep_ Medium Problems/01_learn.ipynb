# Day 50 — Interview Prep: Medium Problems

> Harder problems. Take your time, think before coding.
> For each problem: state the approach before writing a line of code.


## Approach checklist

Before coding, say/write:
1. **What is the input/output?**
2. **What is my approach?** (two pointers, DP, BFS...)
3. **What is the time/space complexity?**
4. **What are the edge cases?**
