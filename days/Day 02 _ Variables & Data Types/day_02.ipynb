## 📖 What is a variable?

A variable is a **named container** that holds a value.

```python
name = "Rocket"
age = 23
```

- `=` is the **assignment operator** — it stores the value on the right into the name on the left
- No need to declare a type — Python figures it out


````python
name = "Rocket"
age = 23
height = 1.75
is_student = True

print(name)
print(age)
print(height)
print(is_student)
````

## 📖 The 4 basic data types

| Type | Example | Description |
|------|---------|-------------|
| `str` | `"hello"` | Text |
| `int` | `23` | Whole number |
| `float` | `1.75` | Decimal number |
| `bool` | `True` / `False` | True or False |

Use `type()` to check what type a value is:


````python
print(type("Rocket"))   # <class 'str'>
print(type(23))         # <class 'int'>
print(type(1.75))       # <class 'float'>
print(type(True))       # <class 'bool'>
````

## 📖 Type conversion

Convert between types using `int()`, `float()`, `str()`, `bool()`:

```python
int("42")       # → 42
float("3.14")   # → 3.14
str(100)        # → "100"
bool(0)         # → False
bool(1)         # → True
```


````python
print(int("42"))
print(float("3.14"))
print(str(100))
print(bool(0), bool(1), bool(""), bool("hello"))
````
