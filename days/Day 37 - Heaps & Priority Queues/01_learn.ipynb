# Day 37 — Heaps & Priority Queues

> Get the min or max in O(log n). The tool for 'find the k largest' problems.


## heapq — Python's min-heap


```python
import heapq

lst = [3,1,4,1,5,9,2,6]
heapq.heapify(lst)        # convert list to heap in-place O(n)
print(heapq.heappop(lst)) # 1 — always pops minimum
heapq.heappush(lst, 0)   # add element
print(heapq.nsmallest(3,lst))  # 3 smallest
print(heapq.nlargest(3,lst))   # 3 largest
```

## Max-heap trick — negate values


```python
import heapq

nums = [3,1,4,1,5,9,2,6]
max_heap = [-n for n in nums]
heapq.heapify(max_heap)
print(-heapq.heappop(max_heap))  # 9 — largest
```
