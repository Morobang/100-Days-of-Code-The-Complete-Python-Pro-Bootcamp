# Day 13 — List Comprehensions

> **Goal:** Write concise, readable loops in a single line.


## 📖 The basic pattern

```python
# Long way
squares = []
for x in range(10):
    squares.append(x**2)

# List comprehension — one line
squares = [x**2 for x in range(10)]
```

**Pattern:** `[expression for item in iterable]`


````python
# Long way
squares_long = []
for x in range(10):
    squares_long.append(x**2)

# Comprehension
squares = [x**2 for x in range(10)]

print(squares_long)
print(squares)
print(squares_long == squares)  # True
````

## 📖 With a filter (if condition)

**Pattern:** `[expression for item in iterable if condition]`

```python
evens = [x for x in range(20) if x % 2 == 0]
```


````python
evens = [x for x in range(20) if x % 2 == 0]
print(evens)

# Strings — filter and transform
words = ['hello', 'world', 'python', 'is', 'great']
long_words = [w.upper() for w in words if len(w) > 4]
print(long_words)
````

## 📖 Nested comprehension

```python
# Flatten a 2D list
matrix = [[1,2,3],[4,5,6],[7,8,9]]
flat   = [num for row in matrix for num in row]
```


````python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
flat = [num for row in matrix for num in row]
print(flat)
````

## 📖 Dict comprehension

**Pattern:** `{key: value for item in iterable}`

```python
squared = {x: x**2 for x in range(5)}
# {0:0, 1:1, 2:4, 3:9, 4:16}
```


````python
squared = {x: x**2 for x in range(5)}
print(squared)

words = ['hello', 'world', 'python']
lengths = {w: len(w) for w in words}
print(lengths)
````
