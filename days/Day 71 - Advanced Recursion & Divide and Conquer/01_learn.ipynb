## Divide and Conquer template


```python
def dc(problem):
    if base_case(problem):
        return solve_directly(problem)
    left, right = split(problem)
    left_sol  = dc(left)
    right_sol = dc(right)
    return merge(left_sol, right_sol)
```

## Merge sort (revisited — focus on merge step)


```python
def merge_sort(lst):
    if len(lst) <= 1: return lst
    mid = len(lst) // 2
    L = merge_sort(lst[:mid])
    R = merge_sort(lst[mid:])
    return merge(L, R)

def merge(L, R):
    result, i, j = [], 0, 0
    while i < len(L) and j < len(R):
        if L[i] <= R[j]: result.append(L[i]); i += 1
        else:            result.append(R[j]); j += 1
    return result + L[i:] + R[j:]

print(merge_sort([5, 3, 8, 1, 9, 2, 7, 4, 6]))
```

## Quick sort


```python
def quicksort(lst):
    if len(lst) <= 1: return lst
    pivot = lst[len(lst) // 2]
    left   = [x for x in lst if x < pivot]
    middle = [x for x in lst if x == pivot]
    right  = [x for x in lst if x > pivot]
    return quicksort(left) + middle + quicksort(right)

print(quicksort([3,6,8,10,1,2,1]))
```

## Master theorem (for interviews)

T(n) = aT(n/b) + f(n)

| Algorithm | Recurrence | Complexity |
|-----------|------------|------------|

| Quicksort (avg) | T(n)=2T(n/2)+n | O(n log n) |
