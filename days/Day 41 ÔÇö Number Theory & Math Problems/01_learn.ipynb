## GCD & LCM


```python
from math import gcd

print(gcd(12, 8))           # 4
print(gcd(100, 75))         # 25

def lcm(a, b): return abs(a*b) // gcd(a, b)
print(lcm(4, 6))            # 12
```

## Prime checking & factorisation


```python
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0: return False
    return True

def prime_factors(n):
    factors = []
    d = 2
    while d*d <= n:
        while n % d == 0: factors.append(d); n //= d
        d += 1
    if n > 1: factors.append(n)
    return factors

print(prime_factors(360))   # [2,2,2,3,3,5]
```
