## Defining and calling


```python
def greet(name):
    return f'Hello {name}!'

print(greet('Rocket'))
print(greet('Alice'))
```

## Default parameters


```python
def power(base, exp=2):
    return base ** exp

print(power(5))     # 25  — uses default
print(power(5, 3))  # 125 — overrides
```

## Multiple return values


```python
def min_max(lst):
    return min(lst), max(lst)   # returns a tuple

lo, hi = min_max([3,1,4,1,5,9])
print(lo, hi)   # 1 9
```

## Scope


```python
x = 'global'

def foo():
    x = 'local'    # this is a NEW local x
    print(x)       # 'local'

foo()
print(x)           # 'global' — unchanged
```

## Docstrings


```python
def bmi(weight_kg, height_m):
    """Calculate Body Mass Index.
    Args: weight_kg (float), height_m (float)
    Returns: float rounded to 1dp
    """
    return round(weight_kg / height_m**2, 1)

help(bmi)
```
