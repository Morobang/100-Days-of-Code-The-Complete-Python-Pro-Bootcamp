# Day 52 — Functional Programming Deep Dive

> Functions as first-class citizens. No side effects. Composable.


## functools toolkit


```python
from functools import reduce, partial, lru_cache, wraps

# reduce — fold list to single value
print(reduce(lambda acc, x: acc+x, [1,2,3,4,5]))  # 15

# partial — pre-fill arguments
multiply = lambda x,y: x*y
double = partial(multiply, 2)
print(double(5))  # 10
```

## itertools toolkit


```python
from itertools import chain, groupby, islice, product, combinations, permutations

# chain — merge iterables
print(list(chain([1,2],[3,4],[5,6])))  # [1,2,3,4,5,6]

# islice — lazy slice
print(list(islice(range(100), 5)))    # [0,1,2,3,4]

# groupby — group consecutive elements
data = [('A',1),('A',2),('B',3),('B',4)]
for key, group in groupby(data, key=lambda x: x[0]):
    print(key, list(group))
```
