# Day 56 — Context Managers & Protocols

> Context managers = automatic setup and teardown.
> Protocols = duck typing made explicit.


## Context managers


```python
# Class-based
class Timer:
    import time
    def __enter__(self):
        self.start = __import__('time').time()
        return self
    def __exit__(self, *args):
        self.elapsed = __import__('time').time() - self.start

with Timer() as t:
    sum(range(1_000_000))
print(f'Elapsed: {t.elapsed:.3f}s')
```

## contextlib


```python
from contextlib import contextmanager

@contextmanager
def managed_resource(name):
    print(f'Opening {name}')
    try:
        yield name.upper()
    finally:
        print(f'Closing {name}')

with managed_resource('db') as r:
    print(f'Using {r}')
```

## Python protocols


```python
class NumberRange:
    def __init__(self, start, end):
        self.start, self.end = start, end

    def __len__(self):      return self.end - self.start + 1
    def __contains__(self, x): return self.start <= x <= self.end
    def __iter__(self):
        return iter(range(self.start, self.end+1))

r = NumberRange(1, 5)
print(len(r))          # 5
print(3 in r)          # True
print(list(r))         # [1,2,3,4,5]
```
