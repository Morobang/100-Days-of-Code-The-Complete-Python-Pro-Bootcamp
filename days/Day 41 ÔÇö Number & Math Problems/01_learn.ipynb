## GCD and LCM


```python
from math import gcd

print(gcd(12, 8))           # 4
print(gcd(48, 18))          # 6

def lcm(a, b):
    return a * b // gcd(a, b)

print(lcm(4, 6))            # 12
```

## Prime checking


```python
def is_prime(n):
    if n < 2: return False
    if n == 2: return True
    if n % 2 == 0: return False
    for i in range(3, int(n**0.5)+1, 2):
        if n % i == 0: return False
    return True

print([n for n in range(2,20) if is_prime(n)])
```
