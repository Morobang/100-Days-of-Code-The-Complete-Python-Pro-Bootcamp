# Day 53 — Concurrency Basics

> Python has three concurrency models. Know which to use when.


## GIL — Python's Global Interpreter Lock

| Model | Best for | GIL? |
|-------|----------|------|
| `threading` | I/O-bound (network, files) | Yes — shares GIL |
| `multiprocessing` | CPU-bound (computation) | No — separate processes |
| `asyncio` | I/O-bound, high concurrency | Single thread |


## Threading


```python
import threading, time

def task(name, duration):
    print(f'{name} starting')
    time.sleep(duration)
    print(f'{name} done')

t1 = threading.Thread(target=task, args=('Task1', 1))
t2 = threading.Thread(target=task, args=('Task2', 1))
t1.start(); t2.start()
t1.join();  t2.join()
print('All done')
```

## asyncio


```python
import asyncio

async def fetch(url):
    print(f'Fetching {url}')
    await asyncio.sleep(1)   # simulate network
    return f'Result from {url}'

async def main():
    results = await asyncio.gather(
        fetch('url1'),
        fetch('url2'),
        fetch('url3'),
    )
    print(results)

# asyncio.run(main())
```
