## Singleton — one instance only


```python
class Singleton:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance

a = Singleton()
b = Singleton()
print(a is b)  # True
```

## Factory — create objects without specifying class


```python
class Dog:
    def speak(self): return 'Woof!'

class Cat:
    def speak(self): return 'Meow!'

def animal_factory(animal_type):
    animals = {'dog': Dog, 'cat': Cat}
    cls = animals.get(animal_type.lower())
    if cls is None: raise ValueError(f'Unknown: {animal_type}')
    return cls()

pet = animal_factory('dog')
print(pet.speak())
```

## Observer — notify many when one changes


```python
class EventEmitter:
    def __init__(self):
        self._listeners = {}

    def on(self, event, fn):
        self._listeners.setdefault(event, []).append(fn)

    def emit(self, event, *args):
        for fn in self._listeners.get(event, []):
            fn(*args)

emitter = EventEmitter()
emitter.on('login', lambda u: print(f'{u} logged in'))
emitter.emit('login', 'Rocket')
```
