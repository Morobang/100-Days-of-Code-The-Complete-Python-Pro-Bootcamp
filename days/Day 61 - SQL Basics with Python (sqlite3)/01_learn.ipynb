# Day 61 — SQL Basics with Python (sqlite3)

> Query databases from Python. Essential for any developer.


## Connect and create


```python
import sqlite3

conn = sqlite3.connect(':memory:')  # in-memory DB
cur  = conn.cursor()

cur.execute('''
    CREATE TABLE players (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        score INTEGER DEFAULT 0
    )
''')
conn.commit()
```

## Insert and query safely


```python
# Parameterised — NEVER use f-strings for SQL values!
cur.execute('INSERT INTO players (name,score) VALUES (?,?)', ('Rocket',9500))
cur.execute('INSERT INTO players (name,score) VALUES (?,?)', ('Alice',8700))
conn.commit()

cur.execute('SELECT * FROM players ORDER BY score DESC')
for row in cur.fetchall():
    print(row)
```

## Rows as dicts


```python
conn.row_factory = sqlite3.Row
cur = conn.cursor()
cur.execute('SELECT * FROM players')
for row in cur.fetchall():
    print(dict(row))
```
