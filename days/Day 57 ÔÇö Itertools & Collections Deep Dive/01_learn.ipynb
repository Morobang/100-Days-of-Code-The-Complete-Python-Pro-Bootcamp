## itertools essentials


```python
from itertools import (
    count, cycle, repeat,
    chain, chain_from_iterable,
    islice, takewhile, dropwhile,
    groupby, combinations, permutations,
    product, starmap, accumulate
)

# accumulate — running total
from itertools import accumulate
print(list(accumulate([1,2,3,4,5])))  # [1,3,6,10,15]

# groupby — group consecutive
data = [1,1,2,2,2,3,1,1]
for key, group in groupby(data):
    print(key, list(group))
```

## collections essentials


```python
from collections import Counter, defaultdict, deque, namedtuple, OrderedDict

# namedtuple
Point = namedtuple('Point', ['x','y'])
p = Point(1, 2)
print(p.x, p.y)    # 1 2
print(p._asdict()) # {'x':1,'y':2}

# deque with maxlen
recent = deque(maxlen=3)
for x in range(5): recent.append(x)
print(recent)      # deque([2,3,4], maxlen=3)
```
