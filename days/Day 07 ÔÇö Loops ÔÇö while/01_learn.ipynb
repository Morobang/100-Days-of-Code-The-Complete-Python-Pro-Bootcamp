# Day 07 — Loops: while

> Repeat code as long as a condition is True.


## Basic while loop


```python
count = 1
while count <= 5:
    print(count)
    count += 1       # MUST update — or infinite loop!
```

## break — exit immediately


```python
n = 0
while True:
    n += 1
    if n == 5: break
print(f'Stopped at {n}')
```

## continue — skip this iteration


```python
for i in range(10):
    if i % 2 == 0: continue   # skip evens
    print(i, end=' ')         # prints 1 3 5 7 9
```

## while True — input validation pattern


```python
while True:
    answer = input("Type quit to exit: ")
    if answer == "quit": break
    print(f"You said: {answer}")
print("Goodbye!")
```
