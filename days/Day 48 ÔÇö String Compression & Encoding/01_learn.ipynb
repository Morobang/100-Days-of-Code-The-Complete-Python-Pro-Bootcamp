## Run-length encoding


```python
def encode_rle(s):
    result = []
    i = 0
    while i < len(s):
        char = s[i]; count = 1
        while i+count < len(s) and s[i+count] == char:
            count += 1
        result.append(f'{count}{char}' if count > 1 else char)
        i += count
    return ''.join(result)

print(encode_rle('aaabbbcc'))  # 3a3b2c
```
