# Day 16 — File I/O

> Read and write files — every real program needs this.


## Writing a file


````python
with open('/tmp/games.txt', 'w') as f:
    f.write('Tekken 8\n')
    f.write('FC26\n')
    f.write('KOF XV\n')

print('Written!')
````

## Reading a file

```python
f.read()        # entire file as one string
f.readlines()   # list of lines (keeps \n)
for line in f:  # line by line (memory efficient)
```


````python
with open('/tmp/games.txt') as f:
    content = f.read()
print(content)

with open('/tmp/games.txt') as f:
    for line in f:
        print(line.strip())   # strip removes \n
````

## File modes

| Mode | Meaning |
|------|---------|
| `'r'` | Read (default) |
| `'w'` | Write — creates/overwrites |
| `'a'` | Append — adds to end |
| `'r+'` | Read + write |


````python
# Append to existing file
with open('/tmp/games.txt', 'a') as f:
    f.write('Elden Ring\n')

with open('/tmp/games.txt') as f:
    print(f.read())
````

## Handling missing files


````python
try:
    with open('doesnt_exist.txt') as f:
        print(f.read())
except FileNotFoundError:
    print('File not found!')
````

## writelines() — write a list


````python
games = ['Tekken 8', 'FC26', 'KOF XV']

with open('/tmp/games2.txt', 'w') as f:
    f.writelines(game + '\n' for game in games)

with open('/tmp/games2.txt') as f:
    print(f.read())
````
