## State machine DP


```python
# Stock trading with cooldown
def max_profit_cooldown(prices):
    hold = -prices[0]  # currently holding stock
    sold = 0           # just sold (now in cooldown)
    rest = 0           # resting (can buy)
    for p in prices[1:]:
        hold, sold, rest = max(hold, rest-p), hold+p, max(rest, sold)
    return max(sold, rest)

print(max_profit_cooldown([1,2,3,0,2]))  # 3
```

## Interval DP


```python
# Minimum cost to merge stones
def min_cost_merge(stones):
    '''Minimum cost to merge stones (each merge = sum of merged group)'''
    n = len(stones)
    prefix = [0] * (n+1)
    for i in range(n): prefix[i+1] = prefix[i] + stones[i]

    dp = [[0]*n for _ in range(n)]
    for length in range(2, n+1):
        for i in range(n-length+1):
            j = i + length - 1
            dp[i][j] = min(dp[i][k]+dp[k+1][j] for k in range(i,j))
            dp[i][j] += prefix[j+1] - prefix[i]
    return dp[0][n-1]

print(min_cost_merge([6,3,8,2]))  # 47
```
