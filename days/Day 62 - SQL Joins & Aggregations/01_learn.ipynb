## Setup


```python
import sqlite3

conn=sqlite3.connect(':memory:')
conn.executescript('''
CREATE TABLE customers (id INT PRIMARY KEY, name TEXT, city TEXT);
CREATE TABLE orders (id INT PRIMARY KEY, customer_id INT, total REAL, status TEXT);
INSERT INTO customers VALUES (1,'Rocket','Pretoria'),(2,'Alice','Cape Town'),(3,'Bob','Durban');
INSERT INTO orders VALUES (1,1,1299,'completed'),(2,1,999,'completed'),
  (3,2,599,'pending'),(4,3,1599,'completed'),(5,2,799,'refunded');
''')
conn.commit()
```

## JOIN types


```python
cur=conn.cursor()
cur.execute('''
    SELECT c.name, o.total, o.status
    FROM customers c
    INNER JOIN orders o ON c.id = o.customer_id
    ORDER BY o.total DESC
''')
for row in cur.fetchall(): print(row)
```

## GROUP BY + HAVING


```python
cur.execute('''
    SELECT c.name, SUM(o.total) as total_spent, COUNT(o.id) as order_count
    FROM customers c
    JOIN orders o ON c.id=o.customer_id
    WHERE o.status='completed'
    GROUP BY c.name
    HAVING total_spent > 1000
    ORDER BY total_spent DESC
''')
for row in cur.fetchall(): print(row)
```
