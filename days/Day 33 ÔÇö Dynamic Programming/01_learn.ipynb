# Day 33 — Dynamic Programming

> Break a problem into overlapping subproblems. Solve each once. Reuse.


## Top-down (memoization)


```python
from functools import lru_cache

@lru_cache(maxsize=None)
def fib(n):
    if n<=1: return n
    return fib(n-1)+fib(n-2)

print([fib(i) for i in range(10)])
```

## Bottom-up (tabulation)


```python
def fib_dp(n):
    if n<=1: return n
    dp=[0]*(n+1)
    dp[1]=1
    for i in range(2,n+1):
        dp[i]=dp[i-1]+dp[i-2]
    return dp[n]

print(fib_dp(10))  # 55
```

## The DP mindset

1. Define the subproblem
2. Find the recurrence: `dp[i] = f(dp[i-1], dp[i-2], ...)`
3. Handle base cases
4. Determine order of computation
