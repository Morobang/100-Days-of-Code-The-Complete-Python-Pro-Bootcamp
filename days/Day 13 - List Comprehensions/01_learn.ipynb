# Day 13 — List Comprehensions

> One-line loops. Faster, cleaner, more Pythonic.


## The pattern

```python
# Long way
squares = []
for x in range(10):
    squares.append(x**2)

# Comprehension
squares = [x**2 for x in range(10)]
```

**Pattern:** `[expression for item in iterable]`


````python
squares = [x**2 for x in range(10)]
print(squares)

words = ['hello','world','python']
upper = [w.upper() for w in words]
print(upper)
````

## With a filter

**Pattern:** `[expression for item in iterable if condition]`


````python
evens  = [x for x in range(20) if x % 2 == 0]
print(evens)

words = ['hi','hello','hey','ok','python','is','great']
long  = [w.upper() for w in words if len(w) > 3]
print(long)
````

## Dict comprehension


````python
names  = ["Rocket","Alice","Bob"]
scores = [9500, 8700, 7200]

board  = {name: score for name, score in zip(names, scores)}
print(board)

# Invert a dict
inverted = {v: k for k, v in board.items()}
print(inverted)
````

## Set comprehension


````python
nums = [-3,-2,-1,0,1,2,3]
unique_squares = {x**2 for x in nums}
print(unique_squares)   # {0,1,4,9} — no duplicates
````

## Generator expression — lazy

Use `()` instead of `[]` — produces values one at a time, never loads everything.


````python
# Huge range — generator uses almost no memory
gen = (x**2 for x in range(1_000_000))
print(next(gen))   # 0
print(next(gen))   # 1
print(sum(x for x in range(100) if x % 2 == 0))  # sum of even numbers
````
