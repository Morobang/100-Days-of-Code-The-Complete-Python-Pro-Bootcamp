## Two pointers


```python
# Two Sum — sorted array
def two_sum_sorted(lst, target):
    left, right = 0, len(lst)-1
    while left < right:
        total = lst[left] + lst[right]
        if total == target:   return (left, right)
        elif total < target:  left += 1
        else:                right -= 1
    return None

print(two_sum_sorted([1,2,3,4,6], 6))  # (1,3)
```

## Sliding window


```python
# Max sum subarray of size k
def max_sum_k(lst, k):
    window = sum(lst[:k])
    best   = window
    for i in range(k, len(lst)):
        window += lst[i] - lst[i-k]
        best = max(best, window)
    return best

print(max_sum_k([2,1,5,1,3,2], 3))  # 9
```
