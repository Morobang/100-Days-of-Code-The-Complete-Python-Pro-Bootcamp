## String methods

Methods are functions that belong to a string object. Call them with `.method()`.


````python
s = "  Hello, World!  "

print(s.upper())        # ALL CAPS
print(s.lower())        # all lowercase
print(s.strip())        # remove edge whitespace
print(s.replace(",","!"))  # replace
print(s.split())        # split into list
print(s.count("l"))     # count occurrences
print(s.startswith("  Hello"))  # True
print(s.find("World"))  # index of first match
````

## Slicing

```python
s = "Python"
s[0]     # 'P'   — index from start
s[-1]    # 'n'   — index from end
s[0:3]   # 'Pyt' — start:stop (stop not included)
s[::2]   # 'Pto' — every 2nd character
s[::-1]  # 'nohtyP' — reversed
```


````python
s = "Python"

print(s[0])      # first
print(s[-1])     # last
print(s[0:3])    # slice
print(s[::-1])   # reversed
print(s[::2])    # every 2nd
````

## Joining & splitting


````python
# Split string into list
sentence = "Tekken 8 is the best fighting game"
words = sentence.split()
print(words)

# Join list back into string
print(" | ".join(words[:3]))
print("-".join(["2025","03","21"]))
````

## f-strings in depth


````python
name = "Rocket"
price = 1299.5
rating = 0.876

print(f"{name:>15}")        # right-align in 15 chars
print(f"{name:<15}|")       # left-align
print(f"R{price:,.2f}")     # R1,299.50
print(f"{rating:.1%}")      # 87.6%
````

## Strings are immutable

You can't change a character in place. Methods return a **new** string.


````python
s = "hello"
# s[0] = "H"  ← TypeError!
# Instead:
new_s = "H" + s[1:]
print(new_s)   # Hello
````
