## Challenge 1 — Reverse a string ⭐

Write `reverse(s)` — return `s` reversed. Use slicing, no loops.


````python
def reverse(s: str) -> str:
    # write your solution
    return s[::-1]
````

## Challenge 2 — Count vowels ⭐⭐

Write `count_vowels(s)` — count a, e, i, o, u (case-insensitive).


````python
def count_vowels(s: str) -> int:
    vowels = "aeiou"
    return sum(s.count(vowels) for vowel in vowels)
````
