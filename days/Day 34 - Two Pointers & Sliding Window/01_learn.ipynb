# Day 34 — Two Pointers & Sliding Window

> Two of the most powerful patterns for array/string interviews.


## Two pointers — opposite ends


```python
# Check if array can form pairs summing to target
def has_pair_sum(lst, target):
    lst.sort()
    left, right = 0, len(lst)-1
    while left < right:
        s = lst[left] + lst[right]
        if s == target:   return True
        elif s < target:  left += 1
        else:             right -= 1
    return False

print(has_pair_sum([1,2,4,6,8,9], 14))  # True
```

## Sliding window — fixed size


```python
def max_avg_subarray(lst, k):
    window = sum(lst[:k])
    best   = window
    for i in range(k, len(lst)):
        window += lst[i] - lst[i-k]
        best = max(best, window)
    return best / k

print(max_avg_subarray([1,12,-5,-6,50,3], 4))  # 12.75
```

## Sliding window — variable size


```python
def min_subarray_sum(target, lst):
    left=total=0; result=float('inf')
    for right in range(len(lst)):
        total += lst[right]
        while total >= target:
            result = min(result, right-left+1)
            total -= lst[left]; left += 1
    return result if result != float('inf') else 0

print(min_subarray_sum(7, [2,3,1,2,4,3]))  # 2
```
