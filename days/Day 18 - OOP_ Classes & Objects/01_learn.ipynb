## Class basics


```python
class Dog:
    species = 'Canis lupus'   # class attribute — shared by all

    def __init__(self, name, breed):
        self.name  = name     # instance attributes — unique per dog
        self.breed = breed

    def bark(self):
        return f'{self.name} says: Woof!'

rex  = Dog('Rex', 'Husky')
fido = Dog('Fido', 'Labrador')

print(rex.bark())
print(Dog.species)
```

## `__str__` and `__repr__`


```python
class Dog:
    def __init__(self, name, breed):
        self.name  = name
        self.breed = breed

    def __str__(self):   # print(dog)
        return f'Dog({self.name}, {self.breed})'

    def __repr__(self):  # developer repr
        return f"Dog('{self.name}', '{self.breed}')"

rex = Dog('Rex','Husky')
print(rex)       # calls __str__
print(repr(rex)) # calls __repr__
```

## `@classmethod` — factory methods


```python
class Dog:
    def __init__(self, name, breed):
        self.name  = name
        self.breed = breed

    @classmethod
    def from_dict(cls, data):
        return cls(data['name'], data['breed'])

rex = Dog.from_dict({'name':'Rex','breed':'Husky'})
print(rex.name)
```

## `@staticmethod` — utility function


```python
class MathHelper:
    @staticmethod
    def is_even(n):
        return n % 2 == 0

print(MathHelper.is_even(4))   # True
print(MathHelper.is_even(7))   # False
```
