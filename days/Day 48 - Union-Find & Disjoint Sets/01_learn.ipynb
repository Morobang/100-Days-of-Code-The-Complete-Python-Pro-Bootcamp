# Day 48 — Union-Find & Disjoint Sets

> Track which elements are in the same group. Nearly O(1) per operation.


## Union-Find with path compression


```python
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank   = [0] * n

    def find(self, x):    # path compression
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):  # union by rank
        px, py = self.find(x), self.find(y)
        if px == py: return False
        if self.rank[px] < self.rank[py]: px, py = py, px
        self.parent[py] = px
        if self.rank[px] == self.rank[py]: self.rank[px] += 1
        return True

    def connected(self, x, y): return self.find(x) == self.find(y)

uf = UnionFind(5)
uf.union(0,1); uf.union(1,2)
print(uf.connected(0,2))  # True
print(uf.connected(0,3))  # False
```
