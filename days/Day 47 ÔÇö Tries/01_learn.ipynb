# Day 47 — Tries

> A tree for string prefix operations. O(L) search where L = word length.


## Trie structure


```python
class TrieNode:
    def __init__(self):
        self.children = {}  # char → TrieNode
        self.is_end   = False

class Trie:
    def __init__(self): self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            node = node.children.setdefault(char, TrieNode())
        node.is_end = True

    def search(self, word):
        node = self.root
        for char in word:
            if char not in node.children: return False
            node = node.children[char]
        return node.is_end

trie = Trie()
trie.insert('apple')
print(trie.search('apple'))   # True
print(trie.search('app'))     # False
```
