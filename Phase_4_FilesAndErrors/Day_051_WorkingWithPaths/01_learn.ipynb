# Day 51: Working with Paths

Hard-coding paths like `'C:\\Users\\Alice\\data.txt'` breaks on other machines.
Python's `os.path` module (and the modern `pathlib`) provide cross-platform tools
for building, inspecting, and navigating file paths.

## 1. Current Working Directory

**Relative paths** are resolved from the current working directory (cwd).
`os.getcwd()` tells you where Python is running from.

```python
import os

print(os.getcwd())   # absolute path to the cwd
```

## 2. Building Paths with `os.path.join()`

Never concatenate paths with `+` or hard-code `\` or `/`.
`os.path.join()` uses the correct separator for the current OS.

```python
import os

path = os.path.join('data', 'reports', 'summary.txt')
print(path)
# Windows: data\reports\summary.txt
# Mac/Linux: data/reports/summary.txt

# Building from an absolute base
home = os.path.expanduser('~')   # user's home directory
config = os.path.join(home, '.config', 'myapp.cfg')
print(config)
```

## 3. Checking Existence

```python
import os

# Create a test file so we have something to check
with open('test_file.txt', 'w', encoding='utf-8') as f:
    f.write('hello')

print(os.path.exists('test_file.txt'))    # True
print(os.path.exists('missing.txt'))      # False

print(os.path.isfile('test_file.txt'))    # True  — it's a file
print(os.path.isdir('test_file.txt'))     # False — not a directory
print(os.path.isdir('.'))                 # True  — '.' is the cwd
```

## 4. Dissecting a Path

```python
import os

path = os.path.join('docs', 'reports', 'annual_report.txt')
print(path)
print(os.path.basename(path))     # annual_report.txt
print(os.path.dirname(path))      # docs/reports
print(os.path.splitext(path))     # ('docs/reports/annual_report', '.txt')

name, ext = os.path.splitext(os.path.basename(path))
print(f'Name: {name}, Extension: {ext}')
```

## 5. Listing Directory Contents

```python
import os

entries = os.listdir('.')   # list names in the current directory
for name in sorted(entries):
    kind = 'DIR ' if os.path.isdir(name) else 'FILE'
    print(f'{kind}  {name}')
```

## 6. Modern Alternative: `pathlib.Path`

`pathlib` (added in Python 3.4) provides an object-oriented approach.
The `/` operator joins path segments — much more readable than `os.path.join()`.

```python
from pathlib import Path

p = Path('docs') / 'reports' / 'annual_report.txt'
print(p)             # docs/reports/annual_report.txt
print(p.name)        # annual_report.txt
print(p.stem)        # annual_report
print(p.suffix)      # .txt
print(p.parent)      # docs/reports

cwd = Path('.')
print(cwd.exists())  # True

# List files and dirs
for item in sorted(Path('.').iterdir()):
    tag = 'DIR ' if item.is_dir() else 'FILE'
    print(f'{tag}  {item.name}')
```
