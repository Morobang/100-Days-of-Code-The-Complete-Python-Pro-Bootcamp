# Day 50: Writing Files

Yesterday you read from files. Today you write to them.
The same `open()` function is used — the difference is the **mode**.

## 1. Write Mode `'w'`

Opens the file for writing. If the file exists it is **overwritten** completely.
If it does not exist, it is created.

```python
with open('output.txt', 'w', encoding='utf-8') as f:
    f.write('Hello, file world!\n')
    f.write('Second line here.\n')

# Read it back to verify
with open('output.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
# Opening again with 'w' OVERWRITES everything
with open('output.txt', 'w', encoding='utf-8') as f:
    f.write('This replaced the old content.\n')

with open('output.txt', 'r', encoding='utf-8') as f:
    print(f.read())   # only the new content
```

## 2. Append Mode `'a'`

Opens the file for writing at the **end**. Existing content is preserved.
If the file does not exist, it is created.

```python
# Create a fresh log file
with open('log.txt', 'w', encoding='utf-8') as f:
    f.write('Entry 1\n')

# Append more entries
with open('log.txt', 'a', encoding='utf-8') as f:
    f.write('Entry 2\n')

with open('log.txt', 'a', encoding='utf-8') as f:
    f.write('Entry 3\n')

with open('log.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

## 3. `.write()` vs `.writelines()`

`.write(s)` writes one string. `.writelines(iterable)` writes each string in the iterable.
Neither adds automatic newlines — you must include `\n` yourself.

```python
# .writelines() — each string must include its own newline
fruits = ['apple\n', 'banana\n', 'cherry\n']
with open('fruits.txt', 'w', encoding='utf-8') as f:
    f.writelines(fruits)

with open('fruits.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
# Common pattern: write a list of items, one per line
items = ['apple', 'banana', 'cherry']
with open('items.txt', 'w', encoding='utf-8') as f:
    for item in items:
        f.write(item + '\n')

with open('items.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

## 4. Writing Numbers

`f.write()` only accepts strings. Convert numbers with `str()` first.

```python
with open('numbers.txt', 'w', encoding='utf-8') as f:
    for i in range(1, 6):
        f.write(str(i) + '\n')  # must convert int to str

# Read back and sum
with open('numbers.txt', 'r', encoding='utf-8') as f:
    total = sum(int(line.strip()) for line in f)
print(f'Sum: {total}')  # 15
```

## 5. Exclusive Create Mode `'x'`

Use `'x'` when you want to create a new file and be certain not to overwrite anything.

```python
import os

# Clean up from a previous run if needed
if os.path.exists('new_file.txt'):
    os.remove('new_file.txt')

# 'x' creates the file — succeeds only if it doesn't exist
with open('new_file.txt', 'x', encoding='utf-8') as f:
    f.write('Created fresh!\n')
print('Created successfully')

# Trying again raises FileExistsError:
# with open('new_file.txt', 'x', encoding='utf-8') as f:
#     f.write('oops')  # FileExistsError
```
