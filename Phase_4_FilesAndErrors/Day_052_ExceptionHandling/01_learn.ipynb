# Day 52: Exception Handling

When something goes wrong at runtime, Python raises an **exception**.
Without handling, exceptions crash the program. The `try / except` block
lets you catch them and respond gracefully.

## 1. What Exceptions Look Like

Python raises different exception types depending on what went wrong.

```python
# Each line below raises a different exception — uncomment one at a time
# int('abc')          # ValueError
# 10 / 0              # ZeroDivisionError
# open('nope.txt')    # FileNotFoundError
# [1, 2, 3][99]       # IndexError
# {'a': 1}['b']       # KeyError
# 'hello' + 1         # TypeError
print('Uncomment a line above to see exceptions in action')
```

## 2. Basic try / except

Code in the `try` block runs normally.
If an exception is raised, Python jumps to the matching `except` block.

```python
try:
    result = 10 / 0
    print('This line never runs')
except ZeroDivisionError:
    print('Caught it: cannot divide by zero')

print('Program continues after the try/except block')
```

## 3. Catching Specific Exceptions

Always catch the most specific exception you can.
A bare `except:` catches everything, including errors you didn't expect — hiding real bugs.

```python
def safe_int(s):
    try:
        return int(s)
    except ValueError:
        print(f"Cannot convert '{s}' to int")
        return None

print(safe_int('42'))    # 42
print(safe_int('abc'))   # error message, then None
print(safe_int('3.5'))   # error message (int() doesn't accept floats as strings)
```

## 4. The `as e` Clause — Accessing the Error

`except SomeError as e:` binds the exception object to `e`.
You can convert it to a string to get the error message.

```python
try:
    int('not a number')
except ValueError as e:
    print(f'ValueError: {e}')

try:
    open('ghost.txt')
except FileNotFoundError as e:
    print(f'FileNotFoundError: {e}')
```

## 5. Multiple except Blocks

You can handle different exception types differently.

```python
def safe_divide(a, b):
    try:
        return a / b
    except TypeError:
        print('Both arguments must be numbers')
        return None
    except ZeroDivisionError:
        print('Cannot divide by zero')
        return None

print(safe_divide(10, 2))     # 5.0
print(safe_divide(10, 0))     # ZeroDivisionError message
print(safe_divide(10, 'x'))   # TypeError message
```

## 6. else — Runs Only if No Exception

The `else` block runs when the `try` block **succeeds** (no exception was raised).
It separates the "success" logic from the error handling.

```python
def parse(s):
    try:
        value = int(s)
    except ValueError:
        print(f"'{s}' is not a valid integer")
    else:
        print(f'Parsed successfully: {value}')   # only if no ValueError

parse('42')
parse('abc')
```

## 7. finally — Always Runs

The `finally` block runs whether or not an exception occurred.
Use it for clean-up code that must always execute.

```python
def attempt_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print('Error: division by zero')
        result = None
    else:
        print(f'Result: {result}')
    finally:
        print('attempt_divide done')   # always runs
    return result

attempt_divide(10, 2)
print()
attempt_divide(10, 0)
```
