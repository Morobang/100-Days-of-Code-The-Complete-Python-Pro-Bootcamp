# Day 53: Raising Exceptions

Yesterday you *caught* exceptions. Today you learn to *raise* them.
When your function receives invalid input, raising an exception is the right response —
it tells the caller exactly what went wrong.

## 1. The `raise` Statement

`raise ExceptionType('message')` raises an exception at that exact point.
Execution stops immediately and Python unwinds the call stack looking for a matching `except`.

```python
# You can raise any built-in exception with a custom message
raise ValueError('something is wrong with the value')  # uncomment to see
```

```python
# The caller catches it with try/except
def get_positive(n):
    if n <= 0:
        raise ValueError(f'{n} must be positive')
    return n

try:
    print(get_positive(5))    # 5
    print(get_positive(-3))   # raises ValueError
except ValueError as e:
    print(f'Error: {e}')
```

## 2. Using raise for Input Validation

Validating inputs at the top of a function and raising immediately keeps the
rest of the function clean — no need for deeply nested `if` checks.

```python
def divide(a, b):
    if b == 0:
        raise ZeroDivisionError('divisor cannot be zero')
    return a / b

def set_username(name):
    if not isinstance(name, str):
        raise TypeError(f'username must be a str, got {type(name).__name__}')
    if len(name) < 3:
        raise ValueError('username must be at least 3 characters')
    return name.strip()

for val in ['Alice', 'ab', 42]:
    try:
        print(set_username(val))
    except (ValueError, TypeError) as e:
        print(f'Error: {e}')
```

## 3. Raising Different Exception Types

Choose the exception type that best describes the problem.

```python
def validate_age(age):
    if not isinstance(age, int):
        raise TypeError(f'age must be int, got {type(age).__name__}')
    if age < 0 or age > 150:
        raise ValueError(f'age must be 0–150, got {age}')
    return age

for val in [25, -1, 200, 'thirty']:
    try:
        print(validate_age(val))
    except (ValueError, TypeError) as e:
        print(f'{type(e).__name__}: {e}')
```

## 4. raise vs return None

Both are ways to signal "something went wrong," but they communicate very different things.

- **`raise`**: the caller made a mistake — invalid input, impossible operation
- **`return None`**: the result is legitimately absent — optional data, not an error

```python
# raise — for programming errors (bad input)
def get_item(lst, index):
    if not isinstance(index, int):
        raise TypeError('index must be an integer')
    if index < 0 or index >= len(lst):
        raise IndexError(f'index {index} is out of range for list of length {len(lst)}')
    return lst[index]

# return None — for expected absence
def find_first(lst, target):
    for i, val in enumerate(lst):
        if val == target:
            return i
    return None   # not found is a valid, expected outcome

data = [10, 20, 30]
print(get_item(data, 1))         # 20
print(find_first(data, 99))      # None
```

## 5. Re-raising with bare `raise`

Inside an `except` block, `raise` with no arguments re-raises the current exception.
This lets you do something (log, clean up) and then let the exception propagate.

```python
def risky():
    raise RuntimeError('something went wrong inside risky()')

def wrapper():
    try:
        risky()
    except RuntimeError:
        print('wrapper: caught, logging error, then re-raising')
        raise   # same exception continues up

try:
    wrapper()
except RuntimeError as e:
    print(f'outer caught: {e}')
```
