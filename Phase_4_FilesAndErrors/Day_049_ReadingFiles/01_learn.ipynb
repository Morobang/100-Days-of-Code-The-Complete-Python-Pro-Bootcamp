# Day 49: Reading Files

Programs often need to read data from disk. Python makes this straightforward
with the built-in `open()` function and the `with` statement.

```python
# Create a sample file so every cell below works stand-alone
with open('demo.txt', 'w', encoding='utf-8') as f:
    f.write('The quick brown fox jumps over the lazy dog.\n')
    f.write('Python is a great programming language.\n')
    f.write('Reading files is an important skill.\n')
    f.write('The more you practice, the better you get.\n')
    f.write('Every expert was once a beginner.\n')
print('demo.txt ready')
```

## 1. Opening and Closing Files

`open(filename, mode, encoding)` returns a **file object**.
You must close it when done — otherwise the OS keeps the file locked.

```python
# Manual open/close — works but error-prone
f = open('demo.txt', 'r', encoding='utf-8')
content = f.read()
f.close()   # easy to forget, especially if an error occurs first

print(type(content))     # <class 'str'>
print(len(content), 'characters')
```

## 2. The `with` Statement

The `with` statement is the correct way to open files.
It guarantees the file is closed when the block ends — even if an exception occurs.

```python
with open('demo.txt', 'r', encoding='utf-8') as f:
    content = f.read()
# file is closed here automatically

print(content)
```

## 3. Reading Methods

There are several ways to read from a file object.

```python
# .read() — entire file as one string
with open('demo.txt', 'r', encoding='utf-8') as f:
    text = f.read()
print(repr(text[:60]))  # shows \n characters
```

```python
# .readlines() — list of lines, each ending with '\n'
with open('demo.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
print(f'{len(lines)} lines')
print(lines[0])   # includes the trailing \n
print(repr(lines[0]))
```

```python
# .readline() — one line at a time
with open('demo.txt', 'r', encoding='utf-8') as f:
    first  = f.readline()
    second = f.readline()
print(repr(first))
print(repr(second))
```

## 4. Iterating Line by Line

The most common and memory-efficient approach:
iterate directly over the file object — Python reads one line at a time.

```python
with open('demo.txt', 'r', encoding='utf-8') as f:
    for line in f:
        print(line, end='')   # line already has \n, so end='' avoids double spacing
```

## 5. Stripping Trailing Newlines

Lines read from a file include a trailing `\n`. Use `.strip()` to remove it.

```python
# Clean list of lines — no trailing whitespace
with open('demo.txt', 'r', encoding='utf-8') as f:
    lines = [line.strip() for line in f]

for line in lines:
    print(repr(line))  # no \n at the end
```

```python
# Practical: count words across all lines
with open('demo.txt', 'r', encoding='utf-8') as f:
    total_words = sum(len(line.split()) for line in f)
print(f'Total words: {total_words}')
```
