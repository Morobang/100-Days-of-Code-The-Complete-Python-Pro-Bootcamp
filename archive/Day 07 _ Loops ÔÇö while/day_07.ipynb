# Day 07 — Loops: while

> **Goal:** Repeat code as long as a condition is True.


## 📖 The while loop

```python
while condition:
    # repeats as long as condition is True
```


````python
count = 1
while count <= 5:
    print(f"Count: {count}")
    count += 1
````

## 📖 break and continue

- `break` — exit the loop immediately
- `continue` — skip the rest of this iteration, go to next


````python
# break example — stop when we find 3
n = 0
while True:
    n += 1
    if n == 3:
        break
print(f"Stopped at {n}")

# continue example — skip even numbers
n = 0
while n < 10:
    n += 1
    if n % 2 == 0:
        continue
    print(n)   # prints 1,3,5,7,9
````

## 📖 while True (infinite loop with break)

A common pattern for menus and input validation:

```python
while True:
    answer = input("Type quit to exit: ")
    if answer == "quit":
        break
    print(f"You said: {answer}")
```


````python
while True:
    answer = input("Type quit to exit: ")
    if answer == "quit":
        break
    print(f"You said: {answer}")
print("Goodbye!")
````
