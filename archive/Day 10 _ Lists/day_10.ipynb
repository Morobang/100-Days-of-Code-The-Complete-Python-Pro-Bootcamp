## 📖 What is a list?

A list is an **ordered, mutable** collection. It can hold any types.

```python
games = ["Tekken 8", "FC26", "KOF XV"]
scores = [95, 87, 72, 100]
mixed  = [1, 'hello', True, 3.14]
```


````python
games = ["Tekken 8", "FC26", "KOF XV"]
print(games)
print(games[0])   # first item
print(games[-1])  # last item
print(len(games)) # length
````

## 📖 Modifying lists

```python
lst.append(x)     # add to end
lst.insert(i, x)  # insert at index i
lst.remove(x)     # remove first occurrence
lst.pop()         # remove & return last
lst.pop(0)        # remove & return first
lst.sort()        # sort in place
lst.reverse()     # reverse in place
```


````python
games = ["Tekken 8", "FC26", "KOF XV"]
games.append("Naruto Storm")
print(games)

games.remove('FC26')
print(games)

games.sort()
print(games)
````

## 📖 Slicing

```python
lst = [0, 1, 2, 3, 4]
lst[1:3]   # [1, 2]
lst[:2]    # [0, 1]
lst[2:]    # [2, 3, 4]
lst[::-1]  # [4, 3, 2, 1, 0]  reversed
```


````python
nums = [0, 1, 2, 3, 4]
print(nums[1:3])
print(nums[:2])
print(nums[2:])
print(nums[::-1])
````

## 📖 Useful list functions

```python
sum([1,2,3])       # 6
max([1,2,3])       # 3
min([1,2,3])       # 1
sorted([3,1,2])    # [1,2,3] — returns new list
3 in [1,2,3]       # True
```


````python
scores = [95, 87, 72, 100, 65]
print(sum(scores))
print(max(scores))
print(min(scores))
print(sorted(scores))
print(100 in scores)
````
