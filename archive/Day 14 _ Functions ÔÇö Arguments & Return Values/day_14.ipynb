## 📖 Positional vs keyword arguments

```python
def greet(name, greeting):
    return f'{greeting} {name}'

greet('Rocket', 'Hello')        # positional
greet(greeting='Hello', name='Rocket')  # keyword
```


````python
def greet(name, greeting):
    return f'{greeting} {name}'

print(greet('Rocket', 'Hello'))                   # positional
print(greet(greeting='Yo', name='Rocket'))        # keyword
````

## 📖 Default arguments

```python
def greet(name, greeting='Hello'):
    return f'{greeting} {name}'

greet('Rocket')          # uses default
greet('Rocket', 'Yo')    # overrides default
```


````python
def power(base, exponent=2):
    return base ** exponent

print(power(5))      # 25  — uses default exponent
print(power(5, 3))   # 125 — overrides default
````

## 📖 *args — variable positional arguments

Use `*args` when you don't know how many arguments will be passed.
`args` is a **tuple** inside the function.

```python
def total(*args):
    return sum(args)

total(1, 2, 3)      # 6
total(1, 2, 3, 4)   # 10
```


````python
def total(*args):
    return sum(args)

print(total(1, 2, 3))
print(total(10, 20, 30, 40, 50))
print(total())   # 0
````

## 📖 **kwargs — variable keyword arguments

`kwargs` is a **dict** inside the function.

```python
def display(**kwargs):
    for key, val in kwargs.items():
        print(f'{key}: {val}')

display(name='Rocket', game='Tekken 8', score=9500)
```


````python
def display(**kwargs):
    for key, val in kwargs.items():
        print(f'{key}: {val}')

display(name='Rocket', game='Tekken 8', score=9500)
````
