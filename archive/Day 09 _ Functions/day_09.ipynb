## 📖 Parameters and return values

```python
def add(a, b):
    return a + b

result = add(3, 4)   # result = 7
```


````python
def add(a, b):
    return a + b

print(add(3, 4))
print(add(10, -2))
total = add(100, 50)
print(total)
````

## 📖 Default parameters

```python
def greet(name, greeting='Hello'):
    return f"{greeting} {name}"

greet("Rocket")           # "Hello Rocket"
greet("Rocket", "Yo")     # "Yo Rocket"
```


````python
def greet(name, greeting='Hello'):
    return f"{greeting} {name}"

print(greet("Rocket"))
print(greet("Rocket", "Yo"))
````

## 📖 Functions can call other functions


````python
def square(n):
    return n * n

def sum_of_squares(a, b):
    return square(a) + square(b)

print(sum_of_squares(3, 4))   # 9 + 16 = 25
````
