# Day 39: Default Parameters

A **default parameter** has a pre-set value that is used when the caller doesn't supply an argument. This makes functions more flexible — they work simply with minimal input, but accept more detail when needed.

## Basic syntax

```python
def greet(name, greeting='Hello'):
    print(f'{greeting}, {name}!')

greet('Alice')                  # uses default → Hello, Alice!
greet('Bob', 'Good morning')    # overrides default → Good morning, Bob!
greet('Charlie', 'Hey')         # Hey, Charlie!
```

## Required params must come before defaults

Python must be able to match positional arguments unambiguously.  
Required (no default) parameters always go first.

```python
# CORRECT — required first, defaults after
def connect(host, port=5432, timeout=30):
    print(f'Connecting to {host}:{port} (timeout={timeout}s)')

connect('localhost')             # port=5432, timeout=30
connect('myserver', 3306)        # port=3306, timeout=30
connect('myserver', 3306, 10)    # all supplied

# WRONG — SyntaxError: non-default argument follows default
# def broken(greeting='Hi', name):
#     pass
```

## Keyword arguments — naming them at the call site

You can use the parameter name when calling to skip over middle defaults or improve readability.

```python
def connect(host, port=5432, timeout=30):
    print(f'{host}:{port} timeout={timeout}')

# Skip port, set only timeout by name
connect('localhost', timeout=5)

# All by name — order doesn't matter
connect(timeout=60, host='prod-server', port=5000)

# Mix: first positional, rest by name
connect('staging', timeout=15)
```

## Practical examples

```python
def make_header(title, width=40, char='='):
    """Print a centred header line."""
    print(title.center(width, char))

make_header('Welcome')                    # default width and char
make_header('Settings', char='-')         # custom char, default width
make_header('Report', 60, '*')            # all custom

def repeat_print(message, times=1, end='\n'):
    for _ in range(times):
        print(message, end=end)

repeat_print('Hello')
repeat_print('Ha', times=5, end=' ')
print()  # newline after the Ha's
```

## The mutable default trap

**Never use a mutable object (list, dict) as a default.** It is created once when the function is defined, not each time it is called — so it is shared across all calls.

```python
# DANGEROUS — the list is shared across calls!
def add_item_bad(item, collection=[]):
    collection.append(item)
    return collection

print(add_item_bad('apple'))    # ['apple']
print(add_item_bad('banana'))   # ['apple', 'banana'] — unexpected!
print(add_item_bad('cherry'))   # ['apple', 'banana', 'cherry'] — keeps growing!

# CORRECT — use None as default, create inside the function
def add_item_good(item, collection=None):
    if collection is None:
        collection = []
    collection.append(item)
    return collection

print(add_item_good('apple'))    # ['apple']
print(add_item_good('banana'))   # ['banana'] — fresh list each time
```

## Summary

| Rule | Example |
|------|---------|
| Default syntax | `def f(x, y=10):` |
| Required before default | `def f(required, optional=val):` |
| Override at call | `f(1, y=99)` |
| Skip by name | `f('host', timeout=5)` |
| Never mutable default | Use `None` and create inside |
