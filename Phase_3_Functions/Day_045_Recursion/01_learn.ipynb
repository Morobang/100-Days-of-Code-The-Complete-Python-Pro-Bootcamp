# Day 45: Recursion

**Recursion** is when a function calls *itself* to solve a smaller version of the same problem.
It is a powerful pattern for problems that are naturally self-similar.

## 1. The Two Rules of Recursion

Every correct recursive function has exactly two parts:

| Part | Purpose |
|------|---------|
| **Base case** | A condition where the function returns directly — no more calls |
| **Recursive case** | The function calls itself with a *simpler* input, getting closer to the base case |

Without a base case the function calls itself forever — a **stack overflow**.

```python
def countdown(n):
    if n < 0:           # base case — stop
        return
    print(n)
    countdown(n - 1)    # recursive case — simpler input

countdown(5)
```

## 2. How the Call Stack Works

Each call to a recursive function adds a new **frame** to the call stack.
Frames are removed (and results bubble back up) once a call returns.

Adding depth markers lets you see this happening in real time.

```python
def countdown_traced(n, depth=0):
    indent = '  ' * depth
    print(f'{indent}countdown({n}) called')
    if n < 0:
        print(f'{indent}base case — returning')
        return
    countdown_traced(n - 1, depth + 1)
    print(f'{indent}countdown({n}) returning')

countdown_traced(3)
```

## 3. Classic Example: Factorial

`factorial(n)` = n × (n-1) × (n-2) × … × 1

Notice: `factorial(n) = n * factorial(n - 1)` — the definition is recursive.

```python
def factorial(n):
    if n == 0:                       # base case: 0! = 1
        return 1
    return n * factorial(n - 1)     # recursive case

print(factorial(0))   # 1
print(factorial(5))   # 120
print(factorial(10))  # 3628800
```

```python
# Trace how factorial(4) unwinds:
# factorial(4)
#   4 * factorial(3)
#         3 * factorial(2)
#               2 * factorial(1)
#                     1 * factorial(0)
#                           1           ← base case
#                     1 * 1 = 1
#               2 * 1 = 2
#         3 * 2 = 6
#   4 * 6 = 24
print(factorial(4))  # 24
```

## 4. Recursion vs Iteration

Any recursive function can be rewritten as a loop, and vice versa.
Recursion is often more readable when the problem is naturally self-similar;
iteration is usually faster and handles larger inputs safely.

```python
# Sum of a list — two versions

def sum_recursive(lst):
    if not lst:               # base case: empty list
        return 0
    return lst[0] + sum_recursive(lst[1:])  # first + sum of rest

def sum_iterative(lst):
    total = 0
    for n in lst:
        total += n
    return total

numbers = [1, 2, 3, 4, 5]
print(sum_recursive(numbers))  # 15
print(sum_iterative(numbers))  # 15
```

## 5. Recursion Depth Limit

Python tracks how many recursive calls are active at once.
The default limit is ~1 000. Exceeding it raises `RecursionError`.

```python
import sys
print(sys.getrecursionlimit())  # 1000 by default
```

```python
# This will hit the limit — missing base case or input too large
def infinite(n):
    return infinite(n + 1)  # no base case!

# infinite(0)  # uncomment to see: RecursionError: maximum recursion depth exceeded
print('Uncomment the line above to see RecursionError')
```
