## Challenge 5 ⭐⭐⭐

The code below uses `global` inside `add_entry`. Refactor it to remove the `global` keyword:
- Add `history` as a parameter to `add_entry`
- Return the updated history from `add_entry`
- Update the call site so the output stays the same

Expected output: `['first', 'second']`

```python
history = []

def add_entry(entry):
    global history
    history.append(entry)

add_entry('first')
add_entry('second')
print(history)
```
