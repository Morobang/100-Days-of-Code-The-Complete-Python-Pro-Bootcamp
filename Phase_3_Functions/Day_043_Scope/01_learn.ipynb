# Day 43: Scope

Every name in Python — a variable, a function, a parameter — lives in a **scope**: the region of code where it is visible.
Understanding scope explains why some variables are accessible and others are not.

## 1. Local Scope

Any variable **assigned inside a function** is **local** to that function.
It is created when the function is called and destroyed when it returns.

```python
def greet(name):
    message = f'Hello, {name}!'  # local to greet
    print(message)

greet('Alice')
# message no longer exists once greet() returns
```

```python
# Accessing a local variable outside its function raises NameError
def make_secret():
    secret = 42

make_secret()
# print(secret)  # uncomment to see: NameError: name 'secret' is not defined
```

## 2. Global Scope

Variables defined at the top level of a script (outside any function) are **global**.
They are readable inside functions — no keyword needed just to *read* them.

```python
score = 0  # global

def show_score():
    print(f'Current score: {score}')  # reading global — allowed

show_score()
```

## 3. The LEGB Rule

When Python looks up a name it searches scopes in this order and stops at the first match:

| Letter | Scope | Where |
|--------|-------|-------|
| **L** | Local | Inside the current function |
| **E** | Enclosing | An outer function (relevant for nested functions — Day 44) |
| **G** | Global | Module level |
| **B** | Built-in | Python's built-ins: `len`, `print`, `range`, … |

```python
x = 'global'

def show_x():
    print(x)   # no local x → LEGB finds global x

def shadow_x():
    x = 'local'  # local x shadows the global
    print(x)

show_x()    # global
shadow_x()  # local
print(x)    # global — shadow_x never changed the global
```

## 4. The `global` Keyword

Reading a global is free. But if you want to **reassign** it inside a function,
you must declare `global name` first.

Without it, Python sees any assignment as creating a *new local variable* —
and then trying to read the local before it is set raises `UnboundLocalError`.

```python
count = 0

def increment():
    global count   # I mean the global count
    count += 1

increment()
increment()
print(count)  # 2
```

```python
# Without 'global', count += 1 is treated as a local assignment
# Python tries to read the local count before it has a value → UnboundLocalError
counter = 0

def broken():
    counter += 1  # UnboundLocalError!

# broken()  # uncomment to see the error
```

## 5. Better Design — Pass and Return Instead of `global`

Using `global` to modify state is usually a **code smell**:
- Hidden dependency between the function and module-level state
- Hard to test — the function's behaviour depends on external state

The clean alternative: **pass the value in as a parameter, return the updated value**.

```python
# No global needed — pure function
def increment(count):
    return count + 1

count = 0
count = increment(count)
count = increment(count)
print(count)  # 2
# Same input always gives same output — easy to test and reason about
```

## 6. Mutable Globals — No Keyword Needed

The `global` keyword is only needed for **reassignment** (making a name point at a new object).

If you **mutate** a mutable global — call `.append()`, `.update()`, `.clear()` — you are
modifying the object in place, not reassigning the name. `global` is not required.

```python
items = []

def add_item(item):
    items.append(item)  # mutation — no 'global' needed

add_item('apple')
add_item('banana')
print(items)  # ['apple', 'banana']

# This WOULD need 'global' because it reassigns the name:
# def reset():
#     global items
#     items = []   # new list object
```
