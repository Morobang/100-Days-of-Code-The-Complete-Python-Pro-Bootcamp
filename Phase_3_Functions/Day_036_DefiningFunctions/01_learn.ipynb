# Day 36: Defining Functions

Until now every line of code ran top-to-bottom, once. A **function** lets you name a block of code and run it whenever you like — and as many times as you like. This is the single most important building block in any program.

## Why functions?

Without functions you repeat yourself. With functions you write the logic once and call it by name.

````python
# Without a function — repeated code
print('--- Welcome ---')
print('--- Welcome ---')
print('--- Welcome ---')

# With a function — define once, call many times
def print_welcome():
    print('--- Welcome ---')

print_welcome()
print_welcome()
print_welcome()
````

## The `def` syntax

```python
def function_name():
    # body — must be indented
    ...
```

- `def` is the keyword
- The name follows Python variable naming rules (lowercase, underscores)
- The `()` holds parameters — empty for now
- The `:` ends the header
- The **indented block** is the body — it runs when the function is called

````python
def greet():
    print('Hello!')
    print('Good to see you.')

def separator():
    print('-' * 30)

# Nothing runs until you CALL the function
separator()
greet()
separator()
greet()
separator()
````

## Docstrings — one-line descriptions

The first string inside a function body (before any code) is its **docstring**. It describes what the function does.

````python
def print_menu():
    """Print the main application menu."""
    print('1. Start')
    print('2. Settings')
    print('3. Quit')

print_menu()

# Python stores the docstring — you can read it with help()
help(print_menu)
````

## Functions can call other functions

This is how programs are built — small focused functions that call each other.

````python
def print_line():
    print('=' * 40)

def print_header():
    print_line()
    print('       DAILY REPORT')
    print_line()

def print_footer():
    print_line()
    print('       End of report')
    print_line()

def show_report():
    print_header()
    print('  Everything looks good today.')
    print_footer()

show_report()
````

## What a function returns when there is no `return`

Every function returns something. If you don't write `return`, it returns `None` implicitly.

````python
def say_hi():
    print('Hi!')

result = say_hi()   # call the function
print('Return value:', result)   # None
print(type(result))              # <class 'NoneType'>

# This is why printing the call of a print-only function looks odd
print(say_hi())   # prints 'Hi!', then prints 'None'
````

## Summary

```python
def function_name():
    """Optional one-line docstring."""
    # body
    ...

function_name()   # call it
```

| Concept | Key point |
|---------|----------|
| `def` | Defines (but does not run) the function |
| Call `name()` | Runs the body |
| Docstring | First string in body — describes purpose |
| No `return` | Function returns `None` implicitly |
