# Day 41: **kwargs — Variable Keyword Arguments

`**kwargs` lets a function accept **any number of keyword arguments**. Inside the function they arrive as a **dict** — keys are the argument names, values are what was passed. This is the partner to `*args`.

## Basic syntax

```python
def show_all(**kwargs):
    print(type(kwargs))   # <class 'dict'>
    print(kwargs)

show_all(name='Alice', age=30)            # {'name': 'Alice', 'age': 30}
show_all(x=1, y=2, z=3)                  # {'x': 1, 'y': 2, 'z': 3}
show_all()                               # {}   — zero kwargs is fine
```

## Iterating over kwargs

```python
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f'  {key}: {value}')

print_details(name='Alice', age=30, city='Cape Town')
print()
print_details(title='Python Crash Course', author='Matthes', year=2019)
```

## Combining params, `*args`, and `**kwargs`

The required order is: **regular params → `*args` → `**kwargs`**

```python
def full_example(required, *args, **kwargs):
    print('Required:', required)
    print('Args:', args)
    print('Kwargs:', kwargs)

full_example('must', 1, 2, 3, name='Alice', city='CT')
# Required: must
# Args: (1, 2, 3)
# Kwargs: {'name': 'Alice', 'city': 'CT'}

# Practical: required plus optional extras
def create_user(username, **profile):
    print(f'User: {username}')
    for key, val in profile.items():
        print(f'  {key}: {val}')

create_user('alice99')
create_user('bob42', email='bob@mail.com', age=25, city='Durban')
```

## Unpacking a dict with `**`

Prefix an existing dict with `**` at the call site to unpack it into keyword arguments.

```python
def connect(host, port, timeout):
    print(f'Connecting to {host}:{port} (timeout={timeout}s)')

config = {'host': 'localhost', 'port': 5432, 'timeout': 30}
connect(**config)   # same as connect(host='localhost', port=5432, timeout=30)

# Merge two dicts then unpack — useful for overriding defaults
defaults = {'host': 'localhost', 'port': 5432, 'timeout': 30}
overrides = {'port': 3306}
connect(**{**defaults, **overrides})   # timeout from defaults, port overridden
```

## Practical example: flexible tag builder

```python
def make_tag(tag, content, **attributes):
    """Build a simple HTML-like tag string."""
    attrs = ''
    for key, val in attributes.items():
        attrs += f' {key}="{val}"'
    return f'<{tag}{attrs}>{content}</{tag}>'

print(make_tag('p', 'Hello world'))
print(make_tag('a', 'Click here', href='https://python.org', target='_blank'))
print(make_tag('input', '', type='text', placeholder='Enter name', required='true'))
```

## Summary

| Concept | Syntax |
|---------|--------|
| Define | `def f(**kwargs):` |
| Type inside | `kwargs` is a `dict` |
| Iterate | `for k, v in kwargs.items():` |
| With all param types | `def f(req, *args, **kwargs):` |
| Unpack dict at call | `f(**my_dict)` |
| Order rule | regular → `*args` → `**kwargs` |
