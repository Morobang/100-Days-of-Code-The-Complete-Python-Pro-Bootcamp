# Day 37: Parameters and Arguments

Yesterday's functions always did the same thing. **Parameters** let you pass information into a function so it can work with different data each time it's called.

## Parameter vs argument

These two words describe the same value from different perspectives:
- **Parameter** — the variable name in the `def` line (what the function expects)
- **Argument** — the actual value you pass when calling the function

```python
def greet(name):       # 'name' is the parameter
    print(f'Hello, {name}!')

greet('Alice')         # 'Alice' is the argument
```

````python
def greet(name):
    print(f'Hello, {name}!')

greet('Alice')
greet('Bob')
greet('everyone')   # same function, different data each time
````

## Multiple parameters

````python
def describe_pet(name, animal):
    print(f'{name} is a {animal}.')

describe_pet('Fido', 'dog')
describe_pet('Whiskers', 'cat')
describe_pet('Nemo', 'fish')

# Positional — arguments match parameters left to right
def introduce(first, last, age):
    print(f'I am {first} {last}, aged {age}.')

introduce('Alice', 'Smith', 30)
introduce('Bob', 'Jones', 25)
````

## Any type can be an argument

````python
def double(value):
    print(value * 2)

double(5)          # 10
double(3.14)       # 6.28
double('hello ')   # 'hello hello '
double([1, 2, 3])  # [1, 2, 3, 1, 2, 3]

# Passing a list
def print_items(items):
    for i, item in enumerate(items, 1):
        print(f'  {i}. {item}')

print_items(['apples', 'bread', 'milk'])
print_items(['Python', 'Java', 'Rust'])
````

## Passing a dictionary

````python
def print_profile(person):
    print(f"Name:  {person['name']}")
    print(f"Age:   {person['age']}")
    print(f"City:  {person['city']}")

alice = {'name': 'Alice', 'age': 30, 'city': 'Cape Town'}
bob   = {'name': 'Bob',   'age': 25, 'city': 'Durban'}

print_profile(alice)
print()
print_profile(bob)
````

## Parameters are local variables

A parameter only exists inside the function. Changing it inside the function does not affect a variable of the same name outside.

````python
name = 'Alice'   # outer variable

def shout(name):           # 'name' here is a separate local parameter
    name = name.upper()    # changes the local copy only
    print(name)

shout('bob')   # BOB
shout(name)    # ALICE — passes the value, not the variable
print(name)    # 'Alice' — outer name unchanged
````

## Summary

| Concept | Example |
|---------|--------|
| One parameter | `def greet(name):` |
| Multiple parameters | `def add(a, b):` |
| Calling with arguments | `greet('Alice')` |
| Any type works | string, int, list, dict, ... |
| Parameters are local | changes inside don't affect outside |
