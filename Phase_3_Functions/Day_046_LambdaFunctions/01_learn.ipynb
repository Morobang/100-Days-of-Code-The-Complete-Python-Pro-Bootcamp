# Day 46: Lambda Functions

A **lambda** is a small, anonymous function defined in a single expression.
It is not a replacement for `def` — it shines in specific situations,
most often as an inline `key=` argument to sorting functions.

## 1. Lambda Syntax

```
lambda parameters: expression
```

The expression is evaluated and returned automatically — no `return` keyword.

````python
# A lambda that doubles its input
double = lambda x: x * 2
print(double(5))   # 10
print(double(3))   # 6

# Equivalent def:
def double_def(x):
    return x * 2

print(double_def(5))  # 10  — same result
````

````python
# Lambdas can take multiple parameters
add = lambda x, y: x + y
print(add(3, 7))   # 10

# Or no parameters at all
greeting = lambda: 'Hello!'
print(greeting())  # Hello!
````

## 2. Lambda as a Sort Key

The most common real-world use: passing a lambda as the `key=` argument to
`sorted()`, `.sort()`, `min()`, or `max()`.
The key function tells Python *what* to compare.

````python
words = ['banana', 'apple', 'cherry', 'date']

# Sort alphabetically (default)
print(sorted(words))

# Sort by length
print(sorted(words, key=lambda w: len(w)))

# Sort by length, longest first
print(sorted(words, key=lambda w: len(w), reverse=True))
````

````python
# Find the longest and shortest word
words = ['banana', 'apple', 'cherry', 'date']
print(max(words, key=lambda w: len(w)))  # banana (or cherry — both length 6)
print(min(words, key=lambda w: len(w)))  # date
````

## 3. Sorting Complex Structures

Lambda is especially useful when sorting lists of tuples or lists of dicts,
where you need to pick a specific field as the sort criterion.

````python
# Sort list of tuples by the second element
pairs = [(3, 'cherry'), (1, 'apple'), (2, 'banana')]
print(sorted(pairs, key=lambda t: t[1]))
# [(1, 'apple'), (2, 'banana'), (3, 'cherry')]
````

````python
# Sort list of dicts by a key
people = [
    {'name': 'Charlie', 'age': 30},
    {'name': 'Alice',   'age': 25},
    {'name': 'Bob',     'age': 35},
]

by_age  = sorted(people, key=lambda p: p['age'])
by_name = sorted(people, key=lambda p: p['name'])

for p in by_age:
    print(f"{p['name']}: {p['age']}")
````

## 4. Lambda vs def

Lambda and `def` produce the same kind of object — a function.
The difference is style and capability, not power.

| Prefer `lambda` | Prefer `def` |
|----------------|--------------|
| Short, one-expression key function | More than one line |
| Passed inline — no name needed | Function has a meaningful name |
| The logic is obvious at a glance | Needs a comment or docstring |

````python
# Good use of lambda — short, inline, self-explanatory
nums = [4, 1, 3, 2, 5]
nums.sort(key=lambda x: -x)  # descending
print(nums)

# Better as def — logic is non-trivial
def score_key(student):
    return (-student['score'], student['name'])  # desc score, asc name

students = [{'name': 'Bob', 'score': 90}, {'name': 'Alice', 'score': 90}]
print(sorted(students, key=score_key))
````

## 5. Lambda Limitations

A lambda body is a **single expression** — no statements, no blocks, no assignments.
If you need any of those, use `def`.

````python
# This is valid — a conditional expression (not an if statement)
abs_val = lambda x: x if x >= 0 else -x
print(abs_val(-5))  # 5
print(abs_val(3))   # 3

# But an if-statement, loop, or assignment in a lambda is a SyntaxError:
# bad = lambda x: if x > 0: x else -x   # SyntaxError
# bad = lambda x: y = x + 1             # SyntaxError
````
