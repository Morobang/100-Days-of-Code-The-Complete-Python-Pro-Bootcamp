# Day 38: Return Values

So far functions have only *printed* things. The `return` statement lets a function *produce a value* that the caller can use — store it, print it, pass it somewhere else. This is what makes functions genuinely reusable.

## The `return` statement

```python
def square(n):
    return n ** 2

# Use the return value directly
print(square(5))          # 25
print(square(3) + 1)      # 10

# Store it in a variable
result = square(7)
print(result)             # 49

# Pass it as an argument to another function
print(f'4 squared is {square(4)}')
```

## Return vs print — an important distinction

A function that **prints** shows output but produces no usable value.  
A function that **returns** produces a value the caller can work with.

```python
def square_print(n):
    print(n ** 2)   # shows output, returns None

def square_return(n):
    return n ** 2   # returns a value, shows nothing

# square_print — can see it but can't use it
a = square_print(4)   # prints 16
print('a =', a)       # a = None  — nothing to work with!

# square_return — can use it anywhere
b = square_return(4)
print('b =', b)            # b = 16
print('b + 1 =', b + 1)    # b + 1 = 17
```

## Returning different types

```python
def is_even(n):
    return n % 2 == 0      # returns a boolean

def grade(score):
    if score >= 80: return 'A'
    if score >= 70: return 'B'
    if score >= 60: return 'C'
    if score >= 50: return 'D'
    return 'F'             # returns a string

def doubled_list(items):
    return [x * 2 for x in items]  # returns a list

print(is_even(4))          # True
print(is_even(7))          # False
print(grade(88))           # A
print(grade(55))           # D
print(doubled_list([1, 2, 3]))  # [2, 4, 6]
```

## Returning multiple values

Python lets you return several values separated by commas. They come back as a **tuple** that you can unpack.

```python
def min_max(numbers):
    return min(numbers), max(numbers)   # returns a tuple

# Receive as a tuple
result = min_max([3, 1, 8, 2, 7])
print(result)          # (1, 8)
print(type(result))    # <class 'tuple'>

# Unpack immediately
lo, hi = min_max([3, 1, 8, 2, 7])
print(f'Min: {lo}, Max: {hi}')

def stats(numbers):
    total = sum(numbers)
    count = len(numbers)
    average = total / count
    return total, count, average

t, c, avg = stats([10, 20, 30, 40])
print(f'Sum={t}, Count={c}, Avg={avg}')
```

## Early return

`return` exits the function immediately — any code after it in that branch is skipped.

```python
def safe_divide(a, b):
    if b == 0:
        return None     # early return — avoids division by zero
    return a / b

print(safe_divide(10, 2))   # 5.0
print(safe_divide(10, 0))   # None

# Check the result before using it
result = safe_divide(10, 0)
if result is None:
    print('Cannot divide by zero')
else:
    print(result)
```

## Chaining function calls

Because functions return values, you can use them as arguments to other functions.

```python
def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

# Use return value directly as argument
result = multiply(add(2, 3), add(1, 4))   # multiply(5, 5)
print(result)   # 25

# Or step by step — same result
sum1 = add(2, 3)
sum2 = add(1, 4)
result2 = multiply(sum1, sum2)
print(result2)  # 25
```

## Summary

| Concept | Syntax |
|---------|--------|
| Return one value | `return value` |
| Return multiple | `return a, b, c` |
| Unpack multiple | `x, y, z = func()` |
| Early return | `return` before the end |
| No return | function returns `None` |
