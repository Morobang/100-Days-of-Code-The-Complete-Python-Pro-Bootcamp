# Day 40: *args — Variable Positional Arguments

`*args` lets you write a function that accepts **any number of positional arguments** — one, ten, or none. Inside the function, all those arguments arrive as a single **tuple**.

## Basic syntax

```python
def show_all(*args):
    print(type(args))   # <class 'tuple'>
    print(args)

show_all(1, 2, 3)           # (1, 2, 3)
show_all('a', 'b')          # ('a', 'b')
show_all(42)                # (42,)
show_all()                  # ()   — zero args is fine
```

## Iterating over args

```python
def total(*args):
    """Return the sum of any number of numbers."""
    result = 0
    for n in args:
        result += n
    return result

print(total(1, 2, 3))          # 6
print(total(10, 20))           # 30
print(total(5))                # 5
print(total())                 # 0

def biggest(*args):
    """Return the largest of any number of values."""
    if not args:
        return None
    result = args[0]
    for val in args[1:]:
        if val > result:
            result = val
    return result

print(biggest(3, 9, 1, 7, 5))  # 9
```

## Combining `*args` with regular parameters

Regular parameters must come **before** `*args`. Anything after `*args` must be passed by keyword.

```python
def greet_all(greeting, *names):
    """Greet any number of people with a given greeting."""
    for name in names:
        print(f'{greeting}, {name}!')

greet_all('Hello', 'Alice', 'Bob', 'Charlie')
greet_all('Hey', 'Diana')   # just one name
greet_all('Hi')             # no names — loop doesn't run

def make_sentence(verb, *nouns):
    items = ', '.join(nouns)
    print(f'{verb}: {items}')

make_sentence('Bought', 'apples', 'bread', 'milk')
```

## Unpacking a list with `*`

If you already have a list or tuple, prefix it with `*` to unpack it into positional arguments.

```python
def add(a, b, c):
    return a + b + c

numbers = [1, 2, 3]
print(add(*numbers))   # same as add(1, 2, 3)

# Works with *args functions too
def total(*args):
    return sum(args)

scores = [88, 75, 92, 61]
print(total(*scores))     # same as total(88, 75, 92, 61)

# Useful for built-ins too
print(max(*scores))       # 92
print(min(*scores))       # 61

# Unpack range
print(total(*range(1, 6)))   # 1+2+3+4+5 = 15
```

## Practical example: a flexible logger

```python
def log(level, *messages):
    """Print one or more messages with a log level prefix."""
    for msg in messages:
        print(f'[{level.upper()}] {msg}')

log('info', 'Server started')
log('warning', 'Disk space low', 'CPU at 95%')
log('error', 'Connection failed', 'Retrying in 5s', 'Attempt 1/3')
```

## Summary

| Concept | Syntax |
|---------|--------|
| Define | `def f(*args):` |
| Type inside | `args` is a `tuple` |
| Iterate | `for x in args:` |
| With regular params | `def f(required, *args):` |
| Unpack list at call | `f(*my_list)` |
