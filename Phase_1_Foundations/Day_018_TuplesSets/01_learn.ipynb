# Day 18: Tuples and Sets

You already know lists. Today you learn two related types, each with a different purpose:

| Type | Ordered | Mutable | Duplicates |
|------|---------|---------|------------|
| List | Yes | Yes | Yes |
| **Tuple** | Yes | **No** | Yes |
| **Set** | No | Yes | **No** |

## Tuples — ordered and immutable

```python
# Create with parentheses
point = (10, 25)
person = ('Alice', 30, 'Cape Town')
single = (42,)          # single item needs trailing comma

print(point)
print(type(point))      # <class 'tuple'>
```

```python
# Indexing — same as lists
person = ('Alice', 30, 'Cape Town')
print(person[0])     # Alice
print(person[-1])    # Cape Town
print(person[1:])    # (30, 'Cape Town')  — slicing works too
```

```python
# Tuples are IMMUTABLE — you cannot change them after creation
# person[0] = 'Bob'   # TypeError: 'tuple' object does not support item assignment

# This is a feature, not a bug — it prevents accidental changes
print('Tuples are safe from modification')
```

## Tuple packing and unpacking

Unpacking assigns each value to a variable in one line.

```python
person = ('Alice', 30, 'Cape Town')

# Unpack into variables
name, age, city = person
print(name)   # Alice
print(age)    # 30
print(city)   # Cape Town

# Practical — swap two values without a temp variable
x, y = 10, 20
x, y = y, x
print(x, y)   # 20 10
```

## Sets — unordered, unique values only

```python
# Create with curly braces
fruits = {'apple', 'banana', 'cherry'}
print(fruits)           # order may vary — sets are unordered
print(type(fruits))     # <class 'set'>

# Duplicates are removed automatically
nums = {1, 2, 2, 3, 3, 3}
print(nums)    # {1, 2, 3}
```

```python
# Convert a list to a set to remove duplicates
votes = ['yes', 'no', 'yes', 'yes', 'no', 'maybe']
unique_votes = set(votes)
print(unique_votes)     # {'yes', 'no', 'maybe'}
print(len(unique_votes))  # 3

# Empty set — use set(), NOT {} (that creates an empty dict)
empty = set()
print(type(empty))   # <class 'set'>
```

## Set methods — add, remove, discard

```python
fruits = {'apple', 'banana'}

fruits.add('cherry')      # add one item
print(fruits)

fruits.remove('banana')   # remove — raises KeyError if not found
print(fruits)

fruits.discard('mango')   # discard — no error if not found
print(fruits)
```

## Set operations — union, intersection, difference

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)    # union        → {1, 2, 3, 4, 5, 6}  — all items from both
print(a & b)    # intersection → {3, 4}               — only items in both
print(a - b)    # difference   → {1, 2}               — in a but not in b
print(b - a)    #              → {5, 6}               — in b but not in a
```

```python
# Practical: who attended both sessions?
session_1 = {'Alice', 'Bob', 'Charlie', 'Diana'}
session_2 = {'Bob', 'Diana', 'Eve', 'Frank'}

print('Attended both:', session_1 & session_2)
print('Attended either:', session_1 | session_2)
print('Only session 1:', session_1 - session_2)
```

## Summary

**Use a tuple when:** the data is fixed and should not be modified — coordinates, RGB values, config constants.

**Use a set when:** you only care about unique membership — removing duplicates, comparing groups.

| | Tuple | Set |
|-|-------|-----|
| Create | `(1, 2, 3)` | `{1, 2, 3}` |
| Ordered | Yes | No |
| Mutable | No | Yes |
| Duplicates | Yes | No |
| Index | Yes | No |
| `in` check | Yes (slow) | Yes (fast) |
