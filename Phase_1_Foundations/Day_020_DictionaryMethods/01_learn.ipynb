## .get() — safe key lookup

`d['key']` raises KeyError if the key is missing. `.get()` returns a default instead.

```python
inventory = {'apple': 50, 'banana': 30, 'cherry': 20}

# Direct access — raises KeyError
# inventory['mango']          # KeyError!

# .get() — safe, returns None by default
print(inventory.get('apple'))   # 50
print(inventory.get('mango'))   # None  — no error

# .get() with a custom default
print(inventory.get('mango', 0))    # 0
print(inventory.get('mango', 'out of stock'))   # 'out of stock'
```

```python
# Practical: look up user input safely
item = input('Enter an item: ').lower()
stock = inventory.get(item, 0)
if stock > 0:
    print(f'{item}: {stock} in stock')
else:
    print(f'{item} is not available')
```

## .keys(), .values(), .items()

```python
scores = {'Alice': 88, 'Bob': 75, 'Charlie': 92}

print(list(scores.keys()))     # ['Alice', 'Bob', 'Charlie']
print(list(scores.values()))   # [88, 75, 92]
print(list(scores.items()))    # [('Alice', 88), ('Bob', 75), ('Charlie', 92)]

# Calculate average using .values()
avg = sum(scores.values()) / len(scores)
print(f'Average: {avg:.2f}')
```

## .update() — merge or bulk update

```python
settings = {'theme': 'light', 'language': 'english', 'font_size': 14}
user_overrides = {'theme': 'dark', 'font_size': 16}

settings.update(user_overrides)   # existing keys overwritten, new keys added
print(settings)
# {'theme': 'dark', 'language': 'english', 'font_size': 16}
```

## .pop() — remove and return a value

```python
scores = {'Alice': 88, 'Bob': 75, 'Charlie': 92}

removed = scores.pop('Bob')     # removes 'Bob' and returns his score
print(removed)                  # 75
print(scores)                   # {'Alice': 88, 'Charlie': 92}

# Safe pop — provide a default to avoid KeyError
result = scores.pop('Diana', None)
print(result)    # None  — 'Diana' was not in the dict
```

## .setdefault() — set a key only if it doesn't exist

```python
d = {'a': 1}
d.setdefault('b', 0)    # 'b' is missing — adds 'b': 0
d.setdefault('a', 99)   # 'a' already exists — does nothing
print(d)                # {'a': 1, 'b': 0}

# Useful shortcut for the word-frequency pattern:
frequency = {}
words = 'the cat sat on the mat the cat'.split()
for word in words:
    frequency.setdefault(word, 0)   # ensure key exists
    frequency[word] += 1
print(frequency)
```

## Nested dictionaries

Dictionary values can be other dictionaries.

```python
students = {
    'Alice': {'score': 88, 'grade': 'B'},
    'Bob':   {'score': 75, 'grade': 'C'},
}

# Access nested values with chained keys
print(students['Alice']['score'])    # 88
print(students['Bob']['grade'])      # C

# Iterate over nested dict
for name, info in students.items():
    print(f"{name}: score={info['score']}, grade={info['grade']}")
```

## Summary

| Method | What it does |
|--------|--------------|
| `.get(key)` | Returns value or `None` — no KeyError |
| `.get(key, default)` | Returns value or your chosen default |
| `.keys()` | View of all keys |
| `.values()` | View of all values |
| `.items()` | View of all (key, value) pairs |
| `.update(other)` | Merge `other` into dict — overwrites existing |
| `.pop(key)` | Remove and return value — KeyError if missing |
| `.pop(key, default)` | Remove and return value — no error if missing |
| `.setdefault(key, val)` | Set key to val only if key is not already present |
