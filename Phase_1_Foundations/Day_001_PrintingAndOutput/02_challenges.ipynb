## Challenge 5 ⭐⭐⭐

Print a simple receipt that looks like this:

```
========== RECEIPT ==========
Item           Price
Coffee         R25.00
Sandwich       R45.00
==============================
TOTAL          R70.00
```

It does not need to be perfectly aligned yet — just get the structure right using `print()` and `\n`.

````python
# Your code here
print("========== RECEIPT ==========")
print(f"{'Item':<12}{'Price':>8}")
print(f"{'Coffee':<12}{'R25.00':>8}")
print(f"{'Sandwich':<12}{'R45.00':>8}")
print("=" * 30)
print(f"{'TOTAL':<12}{'R70.00':>8}")
````
