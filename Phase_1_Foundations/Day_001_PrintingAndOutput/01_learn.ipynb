## Printing multiple values

You can pass multiple values to `print()` separated by commas. Python puts a space between them by default.

```python
print("My name is", "Alex")
print("I am", 20, "years old")
print("Python", "is", "great")
```

## The sep parameter

`sep` controls what goes **between** the values. The default is a space `" "`. You can change it to anything.

```python
print("a", "b", "c")              # default: space between each
print("a", "b", "c", sep="-")     # dash between each
print("a", "b", "c", sep="")      # nothing between each
print("2024", "01", "15", sep="/") # date format
```

## The end parameter

`end` controls what goes **after** the last value. The default is `"\n"` (a newline), which is why each `print()` starts on a new line. Change it to keep output on the same line.

```python
print("Loading", end="...")
print("Done")          # continues on the same line

print()                # blank line

print("Step 1", end=" | ")
print("Step 2", end=" | ")
print("Step 3")
```

## Escape sequences

Escape sequences are special characters inside strings. They start with a backslash `\`.

| Sequence | Meaning |
|----------|---------|
| `\n` | New line |
| `\t` | Tab |
| `\\` | Literal backslash |
| `\"` | Literal double quote |

```python
print("Line 1\nLine 2\nLine 3")   # \n = new line

print()

print("Name:\tAlex")               # \t = tab
print("Age:\t25")
```

## Quick peek: f-strings

f-strings let you put variables directly inside a string. You will learn these fully on Day 10 — for now, just see what they look like.

```python
name = "Alex"
age = 25
print(f"Hello, {name}! You are {age} years old.")
```

## Summary

- `print(value)` — displays a value
- `print(a, b, c)` — displays multiple values with a space between them
- `sep="-"` — change the separator between values
- `end=""` — change what comes after the last value (default: newline)
- `\n` inside a string = new line, `\t` = tab
