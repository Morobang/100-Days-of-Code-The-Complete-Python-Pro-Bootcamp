# Day 6: User Input

So far your programs have used fixed values you typed in the code. Today you make them **interactive** — the user types something and your program responds.

The `input()` function pauses your program, waits for the user to type something, and returns what they typed.

```python
name = input("What is your name? ")
print("Hello,", name)
```

## input() always returns a string

This is the most important rule about `input()`. Even if the user types `42`, Python gives you the **string** `"42"`, not the number `42`.

```python
age = input("How old are you? ")
print(age)
print(type(age))   # <class 'str'> — always a string!
```

## Converting input to numbers

To do math with user input, you need to convert it. Wrap `input()` inside `int()` or `float()`.

```python
age = int(input("How old are you? "))
print(type(age))        # <class 'int'>
print(age + 10)         # now we can do math
```

```python
price = float(input("Enter price: "))
quantity = int(input("How many? "))
total = price * quantity
print("Total:", total)
```

## Building interactive programs

Let's build a program that asks for two pieces of information and combines them.

```python
first_name = input("Enter your first name: ")
last_name = input("Enter your last name: ")
full_name = first_name + " " + last_name
print("Full name:", full_name)
print("Name has", len(full_name), "characters")
```

## A complete interactive calculation

```python
length = float(input("Enter room length (metres): "))
width = float(input("Enter room width (metres): "))
area = length * width
print(f"Room area: {area} square metres")
```

## Summary

- `input("prompt")` — pauses and waits for the user to type
- The return value is **always a string**
- Wrap with `int()` or `float()` to use as a number
- Leave a space before the closing quote in your prompt: `"Enter name: "`
