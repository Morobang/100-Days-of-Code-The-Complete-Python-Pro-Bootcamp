# Day 3: String Basics

A **string** is any sequence of characters — text, numbers, symbols, spaces. Strings are one of the most common data types you will work with.

You create a string by wrapping text in quotes.

````python
greeting = "Hello, World!"
name = 'Python'             # single quotes work too
sentence = "It's a great day"  # double quotes when string has apostrophe
print(greeting)
print(name)
print(sentence)
````

## len() — the length of a string

`len()` returns how many characters are in a string. Spaces and punctuation count.

````python
print(len("Python"))       # 6
print(len("Hello World"))  # 11 (space counts)
print(len(""))             # 0 (empty string)
print(len("  hi  "))      # 6 (spaces count)
````

## Indexing — accessing single characters

Each character in a string has a position called an **index**. Python starts counting at **0**, not 1.

```
P  y  t  h  o  n
0  1  2  3  4  5
-6 -5 -4 -3 -2 -1
```

Negative indexes count from the end.

````python
word = "Python"

print(word[0])   # P — first character
print(word[1])   # y
print(word[5])   # n — last character
print(word[-1])  # n — also last character
print(word[-2])  # o — second to last
````

````python
# IndexError — accessing an index that doesn't exist
# This would crash — left as a comment:
# print(word[10])   # IndexError: string index out of range

# Safe: the valid indexes for a 6-character string are 0-5 (or -6 to -1)
print(f"Length: {len(word)}, last valid index: {len(word) - 1}")
````

## String concatenation — joining strings

Use `+` to join strings together.

````python
first_name = "Ada"
last_name = "Lovelace"

full_name = first_name + " " + last_name
print(full_name)

greeting = "Hello, " + full_name + "!"
print(greeting)
````

## String repetition — repeating a string

Use `*` to repeat a string a number of times.

````python
print("ha" * 3)      # hahaha
print("-" * 30)      # ------------------------------ (useful for dividers)
print("* " * 5)      # * * * * * 
````

## Summary

- Strings: wrap text in `' '` or `" "`
- `len(string)` — number of characters
- `string[0]` — first character, `string[-1]` — last character
- `+` joins strings, `*` repeats them
