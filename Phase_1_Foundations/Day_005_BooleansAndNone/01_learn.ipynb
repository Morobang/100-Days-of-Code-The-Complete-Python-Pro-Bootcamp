# Day 5: Booleans & None

Not all data is text or numbers. Sometimes you just need a yes/no answer. Python's **boolean** type has exactly two values: `True` and `False`.

```python
is_raining = True
is_sunny = False

print(is_raining)
print(is_sunny)
print(type(is_raining))   # <class 'bool'>
```

**Important:** `True` and `False` are capitalized. `true` and `false` are NOT valid Python.

## bool() — converting other values to booleans

You can convert any value to a boolean with `bool()`. This reveals which values Python considers "truthy" and which are "falsy".

```python
print(bool(1))        # True  — non-zero numbers are truthy
print(bool(0))        # False — zero is falsy
print(bool(-5))       # True  — negative non-zero is truthy
print(bool("hello"))  # True  — non-empty string is truthy
print(bool(""))       # False — empty string is falsy
print(bool(None))     # False — None is falsy
```

## Truthy and falsy

**Falsy** values (convert to `False`):
- `0`, `0.0`
- `""` (empty string)
- `None`
- `False`

**Truthy** values (everything else — convert to `True`):
- Any non-zero number
- Any non-empty string
- `True`

This will become very useful when you learn conditionals on Day 13.

```python
# A surprising one:
print(bool("False"))   # True — the STRING "False" is non-empty, so it's truthy!
print(bool(False))     # False — the boolean False is falsy
```

## None — the "nothing" value

`None` is Python's way of representing the absence of a value. It is its own type: `NoneType`.

```python
result = None
print(result)           # None
print(type(result))     # <class 'NoneType'>
```

## Where you'll see None

When a function doesn't explicitly return a value, it returns `None`. The `print()` function itself is a good example.

```python
# print() returns None — it just has a side effect (displaying text)
x = print("hello")    # this prints "hello"...
print(x)              # ...and x is None
print(type(x))        # <class 'NoneType'>
```

## Summary

- `True` and `False` are the two boolean values — capitalized
- `bool()` converts any value — `0`, `""`, and `None` are falsy; everything else is truthy
- `None` means "no value" — it is not `0`, not `False`, not `""`
