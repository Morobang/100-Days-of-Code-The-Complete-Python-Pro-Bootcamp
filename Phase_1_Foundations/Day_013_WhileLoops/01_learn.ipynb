# Day 13: while Loops

A `while` loop runs a block of code **repeatedly** as long as a condition is True.

```python
while condition:
    # this block runs over and over until condition is False
```

The condition is checked **before** each iteration. If it starts as False, the loop body never runs.

## Basic example

````python
n = 1
while n <= 5:
    print(n)
    n += 1      # n = n + 1 — CRITICAL: without this, n never changes and the loop runs forever

print('Done')
````

## Counting up and counting down

````python
# Count up by 1
i = 0
while i < 5:
    print(i)
    i += 1    # 0, 1, 2, 3, 4
````

````python
# Count down
n = 10
while n >= 1:
    print(n)
    n -= 1
print('Blast off!')
````

````python
# Count by 2 (step)
n = 0
while n <= 10:
    print(n)    # 0, 2, 4, 6, 8, 10
    n += 2
````

## while with user input

A very common pattern: keep asking until the user gives a valid answer.

````python
# Keep asking until user enters a positive number
age = int(input('Enter your age: '))
while age <= 0:
    print('Age must be greater than 0. Try again.')
    age = int(input('Enter your age: '))
print(f'Your age: {age}')
````

````python
# Keep processing until user types a sentinel value
word = input('Enter a word (or "quit" to stop): ')
while word != 'quit':
    print(f'You entered: {word.upper()}')
    word = input('Enter a word (or "quit" to stop): ')
print('Goodbye!')
````

## break — exit the loop immediately

````python
n = 1
while True:       # this would run forever without break
    print(n)
    n += 1
    if n > 5:
        break     # jumps out of the loop entirely

print('Loop ended')
````

## continue — skip the rest of this iteration

````python
# Print only odd numbers from 1 to 10
n = 0
while n < 10:
    n += 1
    if n % 2 == 0:
        continue   # skip even numbers — jump back to 'while n < 10'
    print(n)       # 1, 3, 5, 7, 9
````

## Accumulator pattern

Start with a variable at 0, then add to it inside the loop. This is how you build up a running total.

````python
total = 0
count = 0

number = float(input('Enter a number (0 to stop): '))
while number != 0:
    total += number
    count += 1
    number = float(input('Enter a number (0 to stop): '))

print(f'Numbers entered: {count}')
print(f'Total: {total}')
````

## The infinite loop trap

If the condition never becomes False, the loop runs forever and your program freezes. This is the most common while loop mistake.

````python
# DO NOT run this — it loops forever
# n = 1
# while n > 0:
#     print(n)
#     n += 1    # n keeps growing, condition is always True

# If you accidentally run an infinite loop:
# Press the STOP button in Jupyter, or press Ctrl+C in the terminal
print('Safe to run — the loop above is commented out')
````

## Summary

```python
while condition:    # check condition before each iteration
    ...             # loop body
    n += 1          # something must change to avoid infinite loop
```

| Keyword | Effect |
|---------|--------|
| `break` | Exit the loop immediately |
| `continue` | Skip to the next iteration |
| `n += 1` | Shorthand for `n = n + 1` |
| `n -= 1` | Shorthand for `n = n - 1` |
