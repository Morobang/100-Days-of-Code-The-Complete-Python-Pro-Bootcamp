# Day 9: String Methods

Every string in Python comes with a built-in toolkit of **methods** — functions you call with a dot:

```python
'hello'.upper()   # calling the upper() method on the string 'hello'
```

**Important:** methods never change the original string. They always return a **new** string.

## Case methods

````python
print('hello'.upper())          # HELLO
print('HELLO'.lower())          # hello
print('hello world'.title())    # Hello World
print('Hello World'.swapcase()) # hELLO wORLD
````

````python
# Methods return a NEW string — the original is unchanged
greeting = 'hello'
loud = greeting.upper()
print(greeting)  # still 'hello'
print(loud)      # 'HELLO'
````

## Whitespace methods

These remove leading and/or trailing whitespace — very useful for cleaning up `input()`.

````python
messy = '   hello world   '
print(repr(messy.strip()))    # 'hello world'   — both ends
print(repr(messy.lstrip()))   # 'hello world   ' — left only
print(repr(messy.rstrip()))   # '   hello world' — right only

# repr() shows quotes and spaces clearly — useful for debugging
````

## Searching — count() and find()

````python
word = 'banana'
print(word.count('a'))    # 3 — how many times 'a' appears
print(word.count('an'))   # 2 — substrings count too

print(word.find('n'))     # 2 — index of first 'n'
print(word.find('z'))     # -1 — not found (no error, just -1)
````

## Searching — startswith() and endswith()

````python
filename = 'report_2024.pdf'
print(filename.startswith('report'))  # True
print(filename.endswith('.pdf'))      # True
print(filename.endswith('.txt'))      # False

# Practical use: checking user input format
email = input('Enter your email: ')
print('Looks like a Gmail:', email.endswith('@gmail.com'))
````

## Modifying — replace()

`replace(old, new)` swaps every occurrence of `old` with `new`.

````python
sentence = 'I love cats. Cats are great.'
print(sentence.replace('cats', 'dogs'))   # only lowercase 'cats' changes
print(sentence.replace('Cats', 'Dogs'))   # only 'Cats' changes

# Replace is case-sensitive — chain two replaces to catch both:
fixed = sentence.replace('cats', 'dogs').replace('Cats', 'Dogs')
print(fixed)
````

## Splitting and joining

`split()` breaks a string into a **list** of parts.  
`join()` glues a list back into a string.

````python
# split(separator) — break on a separator
csv = 'apple,banana,cherry'
parts = csv.split(',')
print(parts)        # ['apple', 'banana', 'cherry']
print(parts[0])     # apple
print(parts[2])     # cherry

# split() with no argument splits on whitespace
sentence = 'hello world python'
words = sentence.split()
print(words)        # ['hello', 'world', 'python']
````

````python
# join(list) — glue list items together with a separator
# Note: the separator goes BEFORE .join(), the list goes INSIDE
items = ['apple', 'banana', 'cherry']
print(', '.join(items))   # apple, banana, cherry
print(' | '.join(items))  # apple | banana | cherry
print(''.join(items))     # applebananacherry
````

## Checking the content

These methods return `True` or `False`.

````python
print('123'.isdigit())     # True  — every character is a digit
print('12.3'.isdigit())    # False — '.' is not a digit
print('abc'.isalpha())     # True  — every character is a letter
print('abc123'.isalnum())  # True  — letters and digits, no spaces or symbols
print('   '.isspace())     # True  — all whitespace
````

## Chaining methods

Because each method returns a new string, you can chain calls together.

````python
raw = '   HELLO WORLD   '
clean = raw.strip().lower()
print(clean)    # 'hello world'

# Practical: clean user input before using it
name = input('Enter your name: ')
name = name.strip().title()
print('Hello,', name)
````

## Summary

| Method | What it does |
|--------|--------------|
| `.upper()` / `.lower()` | Change case |
| `.title()` | Capitalise each word |
| `.strip()` | Remove leading/trailing whitespace |
| `.count(x)` | Count occurrences of `x` |
| `.find(x)` | Index of first `x`, or `-1` if not found |
| `.startswith(x)` / `.endswith(x)` | Check start/end |
| `.replace(old, new)` | Swap substrings |
| `.split(sep)` | Break into list |
| `.join(list)` | Glue list into string |
| `.isdigit()` / `.isalpha()` / `.isalnum()` | Check content type |
