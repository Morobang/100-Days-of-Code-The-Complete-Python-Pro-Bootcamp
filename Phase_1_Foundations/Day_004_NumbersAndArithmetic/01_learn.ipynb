# Day 4: Numbers & Arithmetic

Python has two main number types:
- **int** — whole numbers: `5`, `-3`, `1000`
- **float** — decimal numbers: `3.14`, `-0.5`, `100.0`

```python
whole = 42
decimal = 3.14

print(type(whole))    # <class 'int'>
print(type(decimal))  # <class 'float'>
```

## Arithmetic operators

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `+` | Addition | `5 + 3` | `8` |
| `-` | Subtraction | `5 - 3` | `2` |
| `*` | Multiplication | `5 * 3` | `15` |
| `/` | Division | `10 / 3` | `3.333...` |
| `//` | Integer division | `10 // 3` | `3` |
| `%` | Modulo (remainder) | `10 % 3` | `1` |
| `**` | Exponentiation | `2 ** 8` | `256` |

```python
print(5 + 3)    # 8
print(5 - 3)    # 2
print(5 * 3)    # 15
print(10 / 3)   # 3.3333...
print(10 // 3)  # 3
print(10 % 3)   # 1
print(2 ** 8)   # 256
```

## True division vs integer division

`/` always gives a float. `//` gives the floor (rounded down to the nearest whole number).

```python
print(10 / 2)    # 5.0  — still a float even though it's a whole number
print(10 // 2)   # 5    — integer
print(10 / 3)    # 3.333...
print(10 // 3)   # 3    — truncates (does not round up)
```

## Modulo — the remainder operator

`%` gives you the remainder after division. Very useful in practice.

```python
print(10 % 3)   # 1  (10 = 3*3 + 1)
print(15 % 5)   # 0  (divides evenly)
print(7 % 2)    # 1  (odd number — remainder 1 when divided by 2)
print(8 % 2)    # 0  (even number — remainder 0)

# Practical: how many items fit, and how many are left over?
total = 17
per_box = 5
boxes = total // per_box    # 3 full boxes
leftover = total % per_box  # 2 left over
print(boxes, leftover)
```

## Order of operations

Python follows BODMAS/PEMDAS. Use parentheses to control order.

```python
print(2 + 3 * 4)      # 14 (multiplication first)
print((2 + 3) * 4)    # 20 (parentheses first)
print(10 - 2 ** 3)    # 2  (exponent first: 2**3=8, then 10-8)
print(100 / 10 / 2)   # 5.0 (left to right)
```

## The math module

Python's `math` module gives you extra mathematical functions. You need to `import` it first.

```python
import math

print(math.sqrt(16))    # 4.0  — square root
print(math.pi)          # 3.141592653589793
print(math.floor(4.9))  # 4    — round down
print(math.ceil(4.1))   # 5    — round up
print(round(3.567, 2))  # 3.57 — built-in round(), not math
```

## Summary

- `int` for whole numbers, `float` for decimals
- `/` always returns a float; `//` returns an integer (floor)
- `%` gives the remainder
- `**` is exponentiation
- `import math` for `sqrt`, `pi`, `floor`, `ceil`
