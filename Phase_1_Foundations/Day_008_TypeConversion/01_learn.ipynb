# Day 8: Type Conversion

Every value in Python has a type. Sometimes you need to change a value from one type to another — for example, `input()` gives you a string, but you need an integer to do math.

This is called **type conversion** (or **casting**).

## int() — convert to integer

```python
print(int("42"))     # string to int
print(int(3.9))      # float to int — TRUNCATES (not rounds!)
print(int(3.1))      # also 3 — truncates toward zero
print(int(True))     # 1
print(int(False))    # 0
```

**Watch out:** `int(3.9)` gives `3`, not `4`. It chops off the decimal — it does not round up.

## float() — convert to float

```python
print(float("3.14"))   # string to float
print(float(42))       # int to float — adds decimal
print(float("  2.5 ")) # strips surrounding whitespace
print(float(True))     # 1.0
print(float(False))    # 0.0
```

## str() — convert to string

```python
print(str(100))       # "100"
print(str(3.14))      # "3.14"
print(str(True))      # "True"
print(str(None))      # "None"

# str() never fails — everything converts to a string
age = 25
message = "I am " + str(age) + " years old."  # must convert to join strings
print(message)
```

## bool() — convert to boolean

You already saw this on Day 5. It follows the truthy/falsy rules.

```python
print(bool(0))        # False
print(bool(1))        # True
print(bool(""))       # False
print(bool("hello"))  # True
print(bool(None))     # False
```

## When conversion fails — ValueError

If you try to convert something that doesn't make sense, Python raises a `ValueError`. You will learn to handle this gracefully on Day 54.

```python
# These would cause ValueError — shown as comments so the notebook doesn't crash:
# int("hello")    # ValueError: invalid literal for int() with base 10: 'hello'
# int("3.14")     # ValueError: int() can't parse a decimal string directly
# float("abc")    # ValueError

# The fix for decimal strings:
value = "3.14"
result = int(float(value))   # convert to float first, then to int
print(result)                # 3
```

## Practical example — working with user input

```python
# Full example — converting input, calculating, printing with str()
price = float(input("Enter price: R"))
quantity = int(input("How many? "))
total = price * quantity

print("Total: R" + str(round(total, 2)))
```

## Summary

| Function | Converts to | Common use |
|----------|-------------|------------|
| `int()` | Integer | User input numbers, removing decimals |
| `float()` | Float | User input decimals, prices |
| `str()` | String | Joining numbers with text |
| `bool()` | Boolean | Checking if a value is "truthy" |
