# Day 12: if / elif / else

Until now every line of your program ran every time. Today you learn to **make decisions** — running different code depending on whether a condition is True or False.

## The if statement

````python
temperature = 35

if temperature > 30:
    print('It is hot outside.')    # only runs when the condition is True

print('This always runs.')         # outside the if block — always runs
````

**Indentation is critical.** The code inside the `if` block must be indented (4 spaces or 1 tab). Python uses indentation to decide what belongs to which block.

## if / else

````python
age = int(input('Enter your age: '))

if age >= 18:
    print('You are an adult.')
else:
    print('You are not yet an adult.')
````

## Comparison operators

These are the building blocks of conditions.

````python
x = 10
print(x == 10)    # True  — equal to (note: == not =)
print(x != 10)    # False — not equal to
print(x > 5)      # True  — greater than
print(x < 5)      # False — less than
print(x >= 10)    # True  — greater than or equal to
print(x <= 9)     # False — less than or equal to
````

````python
# Common mistake: using = instead of ==
# if x = 10:   # SyntaxError! = is assignment, not comparison
# if x == 10:  # correct

name = input('Enter your name: ')
if name == 'Alice':
    print('Hello, Alice!')
else:
    print(f'Hello, {name}. You are not Alice.')
````

## if / elif / else — multiple branches

`elif` means "else if". Only the **first** matching branch runs.

````python
score = int(input('Enter your score: '))

if score >= 90:
    print('Grade: A')
elif score >= 80:
    print('Grade: B')
elif score >= 70:
    print('Grade: C')
elif score >= 60:
    print('Grade: D')
else:
    print('Grade: F')
````

````python
# Why order matters — only the FIRST matching branch runs
x = 95

if x >= 60:      # this matches first...
    print('D')   # ...so this runs and we never reach the A check
elif x >= 90:
    print('A')   # this is never reached for x=95

# Always check from most specific to most general (high to low)
````

## Logical operators: and, or, not

````python
age = 20
has_id = True

# and — both conditions must be True
if age >= 18 and has_id:
    print('Entry allowed')

# or — at least one condition must be True
score = 55
if score < 0 or score > 100:
    print('Invalid score')

# not — flips True to False and vice versa
is_raining = False
if not is_raining:
    print('Leave the umbrella at home')
````

## Storing the result of a condition

You can assign values inside if/else blocks. This is a common pattern.

````python
score = int(input('Enter score: '))

if score >= 60:
    result = 'Pass'
else:
    result = 'Fail'

# Now use result later:
print(f'Your result: {result}')
````

## Summary

```python
if condition:          # runs when True
    ...
elif other_condition:  # runs when previous was False and this is True
    ...
else:                  # runs when none of the above matched
    ...
```

| Operator | Meaning |
|----------|---------|
| `==` | equal to |
| `!=` | not equal to |
| `>` / `<` | greater / less than |
| `>=` / `<=` | greater / less than or equal |
| `and` | both must be True |
| `or` | at least one must be True |
| `not` | flip True/False |
