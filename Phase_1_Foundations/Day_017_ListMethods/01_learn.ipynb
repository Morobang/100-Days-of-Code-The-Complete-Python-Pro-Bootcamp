# Day 17: List Methods

Lists come with a full toolkit of methods. Today you learn the most useful ones — for sorting, searching, inserting, removing, and calculating.

## sort() — sort in-place

```python
nums = [3, 1, 4, 1, 5, 9, 2, 6]
nums.sort()              # modifies the list in-place
print(nums)              # [1, 1, 2, 3, 4, 5, 6, 9]

nums.sort(reverse=True)  # descending order
print(nums)              # [9, 6, 5, 4, 3, 2, 1, 1]
```

```python
# Sorting strings — alphabetical
names = ['Charlie', 'Alice', 'Bob']
names.sort()
print(names)    # ['Alice', 'Bob', 'Charlie']
```

## sorted() — returns a new sorted list

Use `sorted()` when you want to keep the original list unchanged.

```python
original = [3, 1, 4, 1, 5]
new = sorted(original)       # returns a NEW list
print(original)              # [3, 1, 4, 1, 5]  — unchanged
print(new)                   # [1, 1, 3, 4, 5]

# Common mistake:
# result = original.sort()   # sort() returns None, not the list!
# print(result)              # None
```

## reverse()

```python
nums = [1, 2, 3, 4, 5]
nums.reverse()    # in-place
print(nums)       # [5, 4, 3, 2, 1]
```

## count() and index()

```python
items = ['a', 'b', 'a', 'c', 'a']
print(items.count('a'))    # 3 — how many times 'a' appears
print(items.count('z'))    # 0 — not found (no error)

print(items.index('b'))    # 1 — index of first 'b'
print(items.index('a'))    # 0 — index of FIRST 'a'
```

```python
# index() raises ValueError if not found — check first
items = ['apple', 'banana', 'cherry']
search = 'mango'
if search in items:
    print(items.index(search))
else:
    print(f'{search} not found')
```

## pop() and insert()

```python
nums = [1, 2, 3, 4, 5]

# pop() — remove and return an item
last = nums.pop()       # removes last item and returns it
print(last)             # 5
print(nums)             # [1, 2, 3, 4]

first = nums.pop(0)     # remove item at index 0
print(first)            # 1
print(nums)             # [2, 3, 4]
```

```python
nums = [1, 3, 4, 5]

# insert(index, value) — insert at a specific position
nums.insert(1, 99)    # insert 99 at index 1
print(nums)           # [1, 99, 3, 4, 5]

nums.insert(0, 0)     # insert at the very start
print(nums)           # [0, 1, 99, 3, 4, 5]
```

## min(), max(), sum() — built-in functions

```python
scores = [85, 92, 78, 95, 61]
print(min(scores))     # 61
print(max(scores))     # 95
print(sum(scores))     # 411

average = sum(scores) / len(scores)
print(f'Average: {average:.2f}')    # 82.20
```

## copy() — why it matters

Assigning a list to another variable does **not** copy it — both variables point to the same list.

```python
a = [1, 2, 3]
b = a            # b points to the SAME list as a
b.append(4)
print(a)         # [1, 2, 3, 4]  — a changed too!

# Fix: use .copy()
a = [1, 2, 3]
b = a.copy()     # b is an independent copy
b.append(4)
print(a)         # [1, 2, 3]     — a unchanged
print(b)         # [1, 2, 3, 4]
```

## Summary

| Method / Function | What it does |
|-------------------|--------------|
| `.sort()` | Sort in-place (returns None) |
| `sorted(lst)` | Return new sorted list |
| `.reverse()` | Reverse in-place |
| `.count(x)` | Count occurrences of `x` |
| `.index(x)` | Index of first `x` (ValueError if missing) |
| `.pop()` | Remove and return last item |
| `.pop(i)` | Remove and return item at index `i` |
| `.insert(i, x)` | Insert `x` at index `i` |
| `.remove(x)` | Remove first occurrence of `x` |
| `min(lst)` | Smallest value |
| `max(lst)` | Largest value |
| `sum(lst)` | Total of all values |
| `.copy()` | Independent copy of the list |
