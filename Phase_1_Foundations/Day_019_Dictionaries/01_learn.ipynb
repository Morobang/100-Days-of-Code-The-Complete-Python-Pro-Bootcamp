# Day 19: Dictionaries

A **dictionary** stores data as **key-value pairs**.

```python
person = {'name': 'Alice', 'age': 30}
```

Instead of looking up values by position (`list[0]`), you look them up by **name** (`person['name']`).  
This makes code much more readable and self-documenting.

## Creating a dictionary

````python
# Key: value pairs separated by commas, wrapped in {}
person = {'name': 'Alice', 'age': 30, 'city': 'Cape Town'}
print(person)
print(type(person))    # <class 'dict'>

# Keys and values can be any type
mixed = {1: 'one', 'two': 2, True: 'yes'}

# Empty dict
empty = {}
````

## Accessing values by key

````python
person = {'name': 'Alice', 'age': 30, 'city': 'Cape Town'}

print(person['name'])    # Alice
print(person['age'])     # 30

# KeyError if key doesn't exist:
# print(person['email'])   # KeyError: 'email'
````

## Adding and updating values

````python
person = {'name': 'Alice', 'age': 30}

# Add a new key
person['email'] = 'alice@example.com'
print(person)

# Update an existing key
person['age'] = 31
print(person)

# If the key already exists, the value is overwritten — no duplicates
````

## Deleting keys

````python
person = {'name': 'Alice', 'age': 30, 'city': 'Cape Town'}

del person['city']
print(person)    # {'name': 'Alice', 'age': 30}
````

## Checking for a key with `in`

````python
person = {'name': 'Alice', 'age': 30}

print('name' in person)    # True
print('email' in person)   # False

# Safe lookup pattern
key = 'email'
if key in person:
    print(person[key])
else:
    print(f'{key} not found')
````

## Iterating over a dictionary

````python
scores = {'Alice': 88, 'Bob': 75, 'Charlie': 92}

# Default iteration — over keys
for name in scores:
    print(name)

print('---')

# .items() — key AND value together (most common)
for name, score in scores.items():
    print(f'{name}: {score}')
````

````python
# .keys() and .values()
scores = {'Alice': 88, 'Bob': 75, 'Charlie': 92}

print(list(scores.keys()))     # ['Alice', 'Bob', 'Charlie']
print(list(scores.values()))   # [88, 75, 92]

# Sum of all values
total = sum(scores.values())
print(f'Total score: {total}')
````

## Building a dictionary with a loop

````python
# Word frequency counter
sentence = input('Enter a sentence: ').lower()
words = sentence.split()

frequency = {}
for word in words:
    if word in frequency:
        frequency[word] += 1    # word already seen — increment count
    else:
        frequency[word] = 1     # first time seeing this word

for word, count in frequency.items():
    print(f"'{word}': {count}")
````

## Summary

| Operation | Syntax |
|-----------|--------|
| Create | `{'key': value, ...}` |
| Access | `d['key']` — KeyError if missing |
| Add / Update | `d['key'] = value` |
| Delete | `del d['key']` |
| Check key | `'key' in d` |
| Iterate keys | `for k in d:` |
| Iterate both | `for k, v in d.items():` |
