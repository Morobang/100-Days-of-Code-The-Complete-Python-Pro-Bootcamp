# Day 10: String Formatting

You have already used f-strings on Day 1. Today you go deeper — learning how to control exactly how values look: decimal places, alignment, padding, and percentages.

## f-string basics — recap

An f-string starts with `f` before the quote. Any expression inside `{}` is evaluated and inserted.

```python
name = 'Alice'
age = 30
print(f'Name: {name}, Age: {age}')    # Name: Alice, Age: 30
print(f'Next year: {age + 1}')         # Next year: 31
print(f'Shouting: {name.upper()}')     # Shouting: ALICE
print(f'Pi approx: {22 / 7}')         # Pi approx: 3.142857142857143
```

## Format specifiers — decimal places

After the value, add `:` then a format code. `:.2f` means: fixed-point, 2 decimal places.

```python
price = 12.5
print(f'{price:.2f}')    # 12.50
print(f'{price:.4f}')    # 12.5000
print(f'{price:.0f}')    # 13  — rounds to nearest whole number

# The variable itself is unchanged:
print(price)             # still 12.5
```

## Thousands separator

```python
salary = 1234567
print(f'{salary:,}')       # 1,234,567
print(f'{salary:,.2f}')    # 1,234,567.00  — comma AND 2 decimal places
```

## Percentage formatting

`:.2%` multiplies the value by 100 and adds a `%` sign. Pass a decimal (0.85), not 85.

```python
score = 0.856
print(f'{score:.2%}')    # 85.60%
print(f'{score:.0%}')    # 86%

# Common mistake — do NOT pass 85 when you mean 85%:
wrong = 85
print(f'{wrong:.2%}')    # 8500.00%  — that's not what you want!
```

## Alignment — left, right, center

Format: `{value:[fill][align][width]}`

- `<` left-align
- `>` right-align  
- `^` center

```python
name = 'Alice'
print(f'|{name:<15}|')    # |Alice          |  — left, width 15
print(f'|{name:>15}|')    # |          Alice|  — right, width 15
print(f'|{name:^15}|')    # |     Alice     |  — center, width 15
```

```python
# Alignment works for numbers too
price = 9.5
print(f'|{price:>10.2f}|')    # |      9.50|  — right-aligned, 2dp

# Fill character before the alignment symbol:
print(f'{42:0>5}')            # 00042 — zero-padded
print(f'{"hi":*^10}')         # ****hi****
```

## Building a simple table

Alignment is how you make columns line up.

```python
print(f"{'Item':<20} {'Qty':>5} {'Price':>8}")
print(f"{'Coffee':<20} {2:>5} {35.50:>8.2f}")
print(f"{'Orange Juice':<20} {1:>5} {22.00:>8.2f}")
print(f"{'Sandwich':<20} {3:>5} {49.99:>8.2f}")
```

## Triple-quoted strings

Triple quotes (`"""` or `'''`) let a string span multiple lines without `\n`.

```python
message = """
Dear Customer,

Thank you for your order.

Regards,
The Shop
"""
print(message)
```

```python
# f-strings work with triple quotes too
name = input('Enter your name: ')
total = float(input('Enter total: '))
receipt = f"""
=== RECEIPT ===
Customer: {name}
Total:    R{total:.2f}
===============
"""
print(receipt)
```

## Summary

| Format spec | Example | Output |
|-------------|---------|--------|
| `:.2f` | `f'{12.5:.2f}'` | `12.50` |
| `:,` | `f'{1234567:,}'` | `1,234,567` |
| `:.2%` | `f'{0.856:.2%}'` | `85.60%` |
| `:<15` | `f'{"hi":<15}'` | `hi             ` |
| `:>15` | `f'{"hi":>15}'` | `             hi` |
| `:^15` | `f'{"hi":^15}'` | `      hi       ` |
| `:0>5` | `f'{42:0>5}'` | `00042` |
