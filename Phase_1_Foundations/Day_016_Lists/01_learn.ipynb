# Day 16: Lists

A **list** is an ordered, mutable collection of values. You can store any mix of types, and change the list at any time.

```python
fruits = ['apple', 'banana', 'cherry']
```

Lists are one of Python's most-used data structures. You will use them constantly.

## Creating lists

````python
numbers = [1, 2, 3, 4, 5]
names = ['Alice', 'Bob', 'Charlie']
mixed = [1, 'hello', True, 3.14]   # lists can hold any types
empty = []                          # empty list — you can add to it later

print(numbers)
print(names)
print(type(numbers))    # <class 'list'>
````

## Indexing — accessing items

Same rules as strings: zero-based, negative indices count from the end.

````python
fruits = ['apple', 'banana', 'cherry', 'mango']
#          0        1         2         3
#         -4       -3        -2        -1

print(fruits[0])     # apple   — first item
print(fruits[2])     # cherry
print(fruits[-1])    # mango   — last item
print(fruits[-2])    # cherry  — second from end
````

## Slicing lists

Same syntax as string slicing — returns a new list.

````python
fruits = ['apple', 'banana', 'cherry', 'mango', 'grape']
print(fruits[1:3])    # ['banana', 'cherry']
print(fruits[:2])     # ['apple', 'banana']
print(fruits[3:])     # ['mango', 'grape']
print(fruits[-2:])    # ['mango', 'grape']  — last 2 items
````

## len() — number of items

````python
fruits = ['apple', 'banana', 'cherry']
print(len(fruits))    # 3
print(len([]))        # 0

# Last item without knowing the length:
print(fruits[len(fruits) - 1])    # cherry — same as fruits[-1]
````

## Modifying items

Unlike strings, lists are **mutable** — you can change individual items.

````python
fruits = ['apple', 'banana', 'cherry']
fruits[1] = 'blueberry'    # replace the item at index 1
print(fruits)              # ['apple', 'blueberry', 'cherry']
````

## append() and remove()

````python
fruits = ['apple', 'banana']

# append() — add to the end
fruits.append('cherry')
print(fruits)    # ['apple', 'banana', 'cherry']

fruits.append('mango')
print(fruits)    # ['apple', 'banana', 'cherry', 'mango']

# remove() — remove first occurrence of a value
fruits.remove('banana')
print(fruits)    # ['apple', 'cherry', 'mango']
````

## The `in` operator

Check if an item is in a list.

````python
fruits = ['apple', 'banana', 'cherry']
print('banana' in fruits)    # True
print('mango' in fruits)     # False
print('mango' not in fruits) # True

# Practical use:
search = input('Enter a fruit: ').lower()
if search in fruits:
    print('Found!')
else:
    print('Not in the list.')
````

## Iterating over a list with for

````python
scores = [85, 92, 78, 95, 61]

for score in scores:
    print(score)
````

````python
# Building a list with a loop — the append pattern
numbers = []
for i in range(5):
    n = float(input(f'Enter number {i + 1}: '))
    numbers.append(n)

print('You entered:', numbers)
print('Total items:', len(numbers))
````

## Summary

| Operation | Syntax | Notes |
|-----------|--------|-------|
| Create | `[1, 2, 3]` | Square brackets, comma-separated |
| Access | `lst[0]`, `lst[-1]` | Zero-based, negative from end |
| Slice | `lst[1:3]` | Same rules as strings |
| Length | `len(lst)` | Number of items |
| Modify | `lst[1] = 'new'` | Lists are mutable |
| Add | `lst.append(x)` | Adds to the end |
| Remove | `lst.remove(x)` | Removes first occurrence |
| Check | `x in lst` | Returns True or False |
| Loop | `for item in lst:` | Iterates over items |
