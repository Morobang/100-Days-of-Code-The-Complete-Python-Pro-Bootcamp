# Day 15: for Loops

A `for` loop repeats a block of code once for each item in a sequence.

```python
for item in sequence:
    # runs once per item
```

Unlike `while`, you don't manage a counter yourself — Python moves through the sequence automatically.

## Iterating over a string

A string is a sequence of characters. `for` loops over them one at a time.

````python
for char in 'hello':
    print(char)    # h, e, l, l, o — one per line
````

````python
# Practical: count characters in a word
word = input('Enter a word: ')
count = 0
for char in word:
    count += 1
print(f'{word} has {count} characters')   # same as len(word), but shows the loop
````

## range() — generating number sequences

`range(stop)` generates integers from 0 up to (not including) stop.

````python
for i in range(5):
    print(i)    # 0, 1, 2, 3, 4
````

````python
# range(start, stop) — start is inclusive, stop is exclusive
for i in range(1, 6):
    print(i)    # 1, 2, 3, 4, 5
````

````python
# range(start, stop, step)
for i in range(0, 11, 2):
    print(i)    # 0, 2, 4, 6, 8, 10

print('---')

# Count down with a negative step
for i in range(10, 0, -1):
    print(i)    # 10, 9, 8, ..., 1
````

## Practical: multiplication table

````python
n = int(input('Enter a number: '))
for i in range(1, 11):
    print(f'{n} x {i} = {n * i}')
````

## break and continue in for loops

They work the same way as in `while` loops.

````python
# break — exit the loop early
for i in range(10):
    if i == 5:
        break
    print(i)    # 0, 1, 2, 3, 4
````

````python
# continue — skip this iteration
for i in range(10):
    if i % 2 == 0:
        continue    # skip even numbers
    print(i)        # 1, 3, 5, 7, 9
````

## enumerate() — loop with an index

When you need both the position and the value, use `enumerate()`.

````python
for i, char in enumerate('Python'):
    print(f'Index {i}: {char}')

# Output:
# Index 0: P
# Index 1: y  ... etc.
````

````python
# Start counting from 1 instead of 0
sentence = input('Enter a sentence: ')
words = sentence.split()
for i, word in enumerate(words, start=1):
    print(f'{i}. {word}')
````

## for vs while — when to use which

| Situation | Use |
|-----------|-----|
| You know how many times to loop | `for` with `range()` |
| You're looping over items in a sequence | `for` |
| You loop until a condition changes | `while` |
| Input validation (loop until valid input) | `while` |
