# Day 11: String Slicing

On Day 3 you learned that `s[0]` gives you the first character and `s[-1]` gives you the last.  
Today you learn **slicing** — extracting a range of characters at once:

```python
s = 'Python'
s[0:3]   # 'Pyt'
```

## Indexing review

````python
s = 'Python'
#    P y t h o n
#    0 1 2 3 4 5   (positive indices)
#   -6-5-4-3-2-1   (negative indices)

print(s[0])     # P   — first character
print(s[5])     # n   — last character
print(s[-1])    # n   — also last character
print(s[-2])    # o   — second from end
````

## Basic slicing: s[start:stop]

Gives characters from index `start` up to — but **not including** — index `stop`.

````python
s = 'Python'
print(s[0:3])    # 'Pyt'  — indices 0, 1, 2
print(s[2:5])    # 'tho'  — indices 2, 3, 4
print(s[1:4])    # 'yth'  — indices 1, 2, 3

# stop is exclusive — s[0:3] gives 0, 1, 2 (not 3)
````

## Omitting start or stop

If you leave out `start`, Python starts from the beginning.  
If you leave out `stop`, Python goes to the end.

````python
s = 'Python'
print(s[:3])     # 'Pyt'    — from start up to (not including) 3
print(s[3:])     # 'hon'    — from 3 to the end
print(s[:])      # 'Python' — full copy

# Useful patterns:
print(s[:1])     # 'P'   — just the first character
print(s[1:])     # 'ython' — everything after the first character
````

## Negative indices in slices

````python
s = 'Python'
print(s[-3:])    # 'hon'    — last 3 characters
print(s[:-3])    # 'Pyt'    — everything except the last 3
print(s[-4:-1])  # 'tho'    — from 4th-from-end up to (not including) last
````

## Step: s[start:stop:step]

The optional third value controls how many characters to skip between each pick.

````python
s = 'Python'
print(s[::1])    # 'Python'  — every character (default)
print(s[::2])    # 'Pto'     — every 2nd character
print(s[1::2])   # 'yhn'     — start at 1, every 2nd
print(s[0:5:2])  # 'Pto'     — from 0 to 4, every 2nd
````

## Reversing a string

Step `-1` walks backwards through the string. With no start or stop, it reverses the whole thing.

````python
s = 'Python'
print(s[::-1])    # 'nohtyP'  — reversed

word = 'racecar'
print(word[::-1])  # 'racecar'  — palindrome looks the same reversed
print(word == word[::-1])  # True
````

## Practical examples

````python
# Extract a file extension
filename = 'report_2024.pdf'
dot = filename.find('.')
base = filename[:dot]    # 'report_2024'
ext = filename[dot:]     # '.pdf'
print(base)
print(ext)
````

````python
# Split a sentence into two halves
sentence = 'Hello, World!'
mid = len(sentence) // 2
print(sentence[:mid])     # 'Hello, '
print(sentence[mid:])     # 'World!'
````

## Summary

| Syntax | Meaning |
|--------|---------|
| `s[i]` | Single character at index `i` |
| `s[start:stop]` | Characters from `start` up to (not including) `stop` |
| `s[:stop]` | From the beginning up to `stop` |
| `s[start:]` | From `start` to the end |
| `s[::step]` | Every `step`-th character |
| `s[::-1]` | Reversed |
| `s[-n:]` | Last `n` characters |
| `s[:-n]` | All except the last `n` characters |
