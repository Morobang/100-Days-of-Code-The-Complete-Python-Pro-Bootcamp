# Day 2: Variables & Naming

A **variable** is a named container that holds data. Instead of using raw values everywhere, you give them names so your code is readable and reusable.

You create a variable using `=` (the assignment operator).

```python
name = "Alex"
age = 25
city = "Cape Town"

print(name)
print(age)
print(city)
```

## Naming rules

Python has strict rules about variable names. Break them and you get a `SyntaxError`.

**Allowed:**
- Letters (`a-z`, `A-Z`), digits (`0-9`), and underscores (`_`)
- Must start with a letter or underscore — NOT a digit

**Not allowed:**
- Spaces: `first name` — invalid
- Starting with a digit: `2fast` — invalid
- Special characters: `my-var`, `my@var` — invalid

```python
# Valid names
first_name = "Alex"
age2 = 25
_hidden = "secret"
totalPrice = 99.99   # works, but not Python style

# These would cause SyntaxError — left as comments so the notebook doesn't crash
# 2name = "Alex"     # can't start with digit
# first-name = "Alex" # hyphens not allowed
# first name = "Alex" # spaces not allowed

print(first_name, age2)
```

## The snake_case convention

Python style (PEP 8) says variable names should use **snake_case**: all lowercase, words separated by underscores.

```python
# Python style
first_name = "Alex"
total_price = 99.99
is_logged_in = True

# Works but not Python style
firstName = "Alex"   # camelCase (used in JavaScript)
FirstName = "Alex"   # PascalCase (used for class names in Python — Day 57)

print(first_name, total_price, is_logged_in)
```

## Reassigning variables

You can change what a variable holds at any time. The old value is replaced.

```python
score = 0
print("Score:", score)

score = 10
print("Score:", score)

score = score + 5   # take current value, add 5, store result back
print("Score:", score)
```

## The type() function

Every value in Python has a type. Use `type()` to check what type a variable is.

```python
name = "Alex"
age = 25
height = 1.75
is_student = True

print(type(name))       # <class 'str'>
print(type(age))        # <class 'int'>
print(type(height))     # <class 'float'>
print(type(is_student)) # <class 'bool'>
```

## Multiple assignment

Python lets you assign multiple variables on one line.

```python
x, y, z = 1, 2, 3
print(x, y, z)

a = b = c = 0   # all three start at 0
print(a, b, c)
```

## Summary

- Create variables with `=`
- Names: letters, digits, underscores — can't start with a digit
- Use `snake_case` by convention
- Variables can be reassigned at any time
- `type()` tells you what type a variable is
