## Creating sets

Three ways — each useful in different situations.

```python
# 1. Literal — curly braces
colours = {'red', 'green', 'blue'}

# 2. set() from a list — removes duplicates
numbers = set([1, 2, 2, 3, 3, 3, 4])
print(numbers)   # {1, 2, 3, 4}

# 3. set() from a string — unique characters
chars = set('mississippi')
print(chars)   # {'m', 'i', 's', 'p'} — order not guaranteed

# Empty set — must use set(), NOT {}
empty = set()
print(type(empty))    # <class 'set'>
print(type({}))       # <class 'dict'> — {} creates an empty dict!

# Sets are unordered — no indexing
# colours[0]   # TypeError: 'set' object is not subscriptable
```

## Modifying sets

| Method | What it does |
|--------|--------------|
| `.add(x)` | Add one element |
| `.update(iter)` | Add all elements from an iterable |
| `.remove(x)` | Remove x — `KeyError` if not present |
| `.discard(x)` | Remove x — **no error** if not present |
| `.pop()` | Remove and return an arbitrary element |
| `.clear()` | Empty the set |

```python
fruits = {'apple', 'banana', 'cherry'}

fruits.add('mango')
print(fruits)

fruits.add('apple')   # duplicate — silently ignored
print(fruits)         # still 4 elements

fruits.update(['kiwi', 'grape', 'banana'])   # banana already there
print(fruits)

fruits.discard('banana')   # safe remove
fruits.discard('papaya')   # doesn't exist — no error
print(fruits)

# fruits.remove('papaya')  # KeyError — use discard for safety
```

## The four set operations

Both operator syntax and method syntax work.

```python
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}

# Union — all elements from either set
print(a | b)              # {1, 2, 3, 4, 5, 6, 7, 8}
print(a.union(b))         # same

# Intersection — only elements in BOTH
print(a & b)              # {4, 5}
print(a.intersection(b))  # same

# Difference — in a but NOT in b
print(a - b)              # {1, 2, 3}
print(a.difference(b))    # same
print(b - a)              # {6, 7, 8}  — order matters!

# Symmetric difference — in one but not both
print(a ^ b)                       # {1, 2, 3, 6, 7, 8}
print(a.symmetric_difference(b))   # same
```

## Relationship tests

```python
evens = {2, 4, 6, 8, 10}
small_evens = {2, 4}
odds = {1, 3, 5, 7}

# issubset — is every element in this set also in the other?
print(small_evens.issubset(evens))   # True — {2,4} ⊆ {2,4,6,8,10}
print(evens.issubset(small_evens))   # False

# issuperset — is every element of the other in this set?
print(evens.issuperset(small_evens)) # True
print(small_evens.issuperset(evens)) # False

# isdisjoint — do the two sets share NO elements?
print(evens.isdisjoint(odds))        # True — no overlap
print(evens.isdisjoint(small_evens)) # False — they share 2 and 4

# Operator equivalents
print(small_evens <= evens)   # issubset
print(evens >= small_evens)   # issuperset
```

## `frozenset` — an immutable set

A `frozenset` behaves like a set but cannot be modified. Because it's immutable, it can be used as a dictionary key or stored inside another set.

```python
allowed = frozenset(['admin', 'editor', 'viewer'])
print(allowed)
print('admin' in allowed)   # True
print(type(allowed))        # <class 'frozenset'>

# frozenset as a dict key (regular set can't be a key)
permissions = {
    frozenset(['read']): 'viewer',
    frozenset(['read', 'write']): 'editor',
    frozenset(['read', 'write', 'delete']): 'admin'
}
user_perms = frozenset(['read', 'write'])
print(permissions.get(user_perms, 'unknown'))   # editor

# Trying to modify raises an error
# allowed.add('superuser')  # AttributeError: 'frozenset' object has no attribute 'add'
```

## Sets for fast membership testing

Checking `x in set` is O(1) regardless of size. `x in list` is O(n) — it scans every element.

```python
# When you only need to test membership and don't need order,
# convert to a set.

banned_words = {'spam', 'scam', 'fake', 'fraud', 'phishing'}

messages = [
    'buy this now',
    'this is a scam',
    'hello friend',
    'fake offer inside'
]

for msg in messages:
    words = set(msg.split())
    flagged = words & banned_words   # intersection
    if flagged:
        print(f'FLAGGED: "{msg}" — matched: {flagged}')
    else:
        print(f'OK: "{msg}"')
```

| Symmetric diff | `a ^ b` | `a.symmetric_difference(b)` |
| Subset | `a <= b` | `a.issubset(b)` |
| Superset | `a >= b` | `a.issuperset(b)` |
| Disjoint | — | `a.isdisjoint(b)` |
