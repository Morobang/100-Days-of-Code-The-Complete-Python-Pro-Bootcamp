# Day 27: Dictionaries — Deeper Patterns

Days 19 and 20 gave you the fundamentals and methods. Today you apply them to four patterns that come up constantly in real programs: counting, grouping, building from parallel data, and inverting.

## Pattern 1: Counting occurrences

A dictionary is the natural tool for counting how many times each value appears in a sequence.

```python
votes = ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob', 'Alice', 'Bob']

# Method 1: if/else
counts = {}
for name in votes:
    if name in counts:
        counts[name] += 1
    else:
        counts[name] = 1
print(counts)

# Method 2: .get() with default — cleaner
counts2 = {}
for name in votes:
    counts2[name] = counts2.get(name, 0) + 1
print(counts2)

# Method 3: .setdefault() — also clean
counts3 = {}
for name in votes:
    counts3.setdefault(name, 0)
    counts3[name] += 1
print(counts3)

# Find the winner
winner = max(counts, key=counts.get)
print(f'Winner: {winner} with {counts[winner]} votes')
```

## Pattern 2: Grouping items

Sometimes you want to group items into categories — the values become **lists**.

```python
words = ['apple', 'ant', 'banana', 'bear', 'avocado', 'blueberry', 'cherry']

# Group words by their first letter
groups = {}
for word in words:
    letter = word[0]
    if letter not in groups:
        groups[letter] = []     # start a new list for this letter
    groups[letter].append(word)

for letter, group in sorted(groups.items()):
    print(f'{letter}: {group}')

# Same result using .setdefault()
groups2 = {}
for word in words:
    groups2.setdefault(word[0], []).append(word)
print(groups2)
```

## Pattern 3: Building a dict from parallel lists

When you have two lists of the same length — one for keys, one for values — use a loop to pair them up.

```python
countries = ['France', 'Japan', 'Brazil', 'Australia']
capitals  = ['Paris', 'Tokyo', 'Brasilia', 'Canberra']

# Build a dict by iterating over indices
country_to_capital = {}
for i in range(len(countries)):
    country_to_capital[countries[i]] = capitals[i]

print(country_to_capital)
print(country_to_capital['Japan'])  # Tokyo

# Lookup in both directions
query = input('Country: ')
print(country_to_capital.get(query, 'Not found'))
```

## Pattern 4: Inverting a dictionary

Swap every key and value so lookups work in the other direction.

```python
country_to_capital = {'France': 'Paris', 'Japan': 'Tokyo', 'Brazil': 'Brasilia'}

# Build the inverted dict
capital_to_country = {}
for country, capital in country_to_capital.items():
    capital_to_country[capital] = country

print(capital_to_country)
print(capital_to_country['Tokyo'])  # Japan

# Note: this only works cleanly if all values are unique.
# If two keys share a value, one will overwrite the other when inverted.
```

## Sorting a dictionary's output

Dicts don't have a sort method, but you can sort `.items()` before iterating.

```python
scores = {'Alice': 88, 'Bob': 75, 'Charlie': 92, 'Diana': 81}

# Sort by key (name)
print('By name:')
for name, score in sorted(scores.items()):
    print(f'  {name}: {score}')

# Sort by value (score) — rebuild as (score, name) then sort
by_score = []
for name, score in scores.items():
    by_score.append((score, name))
by_score.sort(reverse=True)

print('\nBy score (highest first):')
for score, name in by_score:
    print(f'  {name}: {score}')
```

## Summary

| Pattern | Core idea |
|---------|----------|
| Counting | `d[key] = d.get(key, 0) + 1` |
| Grouping | `d.setdefault(key, []).append(item)` |
| From parallel lists | `d[keys[i]] = values[i]` in a loop |
| Inverting | `new[v] = k` for `k, v` in `d.items()` |
| Sort by key | `sorted(d.items())` |
| Sort by value | build `(value, key)` list, sort, iterate |
