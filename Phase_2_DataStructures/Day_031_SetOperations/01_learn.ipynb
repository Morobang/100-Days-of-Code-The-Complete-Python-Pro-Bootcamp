# Day 31: Set Operations — In-Depth

Day 30 introduced the four operations and how to read them. Today you learn the **in-place** versions that modify a set directly, how to chain operations across three or more sets, and real-world patterns where sets solve problems that would be clunky with loops.

## In-place operators

`a | b` creates a **new** set. `a |= b` modifies `a` directly — like `+=` for numbers.

```python
registered = {'Alice', 'Bob', 'Charlie'}
new_signups = {'Diana', 'Eve', 'Bob'}   # Bob already registered

# |=  equivalent to .update()
registered |= new_signups
print('After union update:', registered)
# {'Alice', 'Bob', 'Charlie', 'Diana', 'Eve'}

# &=  keep only elements in BOTH
active = {'Alice', 'Bob', 'Diana', 'Frank'}
verified = {'Bob', 'Diana', 'Grace'}
active &= verified
print('After intersection update:', active)
# {'Bob', 'Diana'}

# -=  remove elements that are in the right-hand set
inventory = {'apple', 'banana', 'cherry', 'mango'}
sold_out = {'banana', 'mango', 'papaya'}
inventory -= sold_out
print('After difference update:', inventory)
# {'apple', 'cherry'}

# ^=  keep only elements in exactly one set
team = {'Alice', 'Bob', 'Charlie'}
transfers = {'Bob', 'Diana'}   # Bob leaves, Diana joins
team ^= transfers
print('After symmetric diff update:', team)
# {'Alice', 'Charlie', 'Diana'}
```

## Method equivalents

Each in-place operator has an equivalent method that does the same thing.

```python
s = {1, 2, 3, 4}

s.update({4, 5, 6})              # |=
print('update:', s)

s.intersection_update({2, 4, 6, 8})  # &=
print('intersection_update:', s)

s.difference_update({4})        # -=
print('difference_update:', s)

s.symmetric_difference_update({2, 7})  # ^=
print('symmetric_difference_update:', s)
```

## Operations across three or more sets

Chain operators or use a loop to find elements common to any number of sets.

```python
monday    = {'Alice', 'Bob', 'Charlie', 'Diana'}
tuesday   = {'Bob', 'Charlie', 'Eve', 'Frank'}
wednesday = {'Charlie', 'Diana', 'Eve', 'Grace'}

# Students present ALL three days
all_three = monday & tuesday & wednesday
print('All three days:', all_three)   # {'Charlie'}

# Students present on at least one day
any_day = monday | tuesday | wednesday
print('Any day:', sorted(any_day))

# Students present on Monday but not Tuesday or Wednesday
monday_only = monday - tuesday - wednesday
print('Monday only:', monday_only)   # {'Alice'}

# Using a loop for any number of sets
all_sets = [monday, tuesday, wednesday]
common = all_sets[0].copy()
for s in all_sets[1:]:
    common &= s
print('Common (loop):', common)
```

## Pattern: filtering with a blocklist

```python
stopwords = {'the', 'a', 'an', 'in', 'on', 'at', 'to', 'for', 'is', 'are'}

sentence = 'the cat is sitting on a mat in the sun'
words = set(sentence.split())

# Remove stopwords — set difference in one line
keywords = words - stopwords
print('Keywords:', sorted(keywords))

# Count how many were removed
print(f'Removed {len(words) - len(keywords)} stopwords')
```

## Pattern: deduplication while processing

```python
# Track which URLs have been visited — fast O(1) lookup
visited = set()
to_crawl = ['page1', 'page2', 'page1', 'page3', 'page2', 'page4']

for url in to_crawl:
    if url not in visited:
        print(f'Processing: {url}')
        visited.add(url)
    else:
        print(f'Skipping (already seen): {url}')

print(f'\nProcessed {len(visited)} unique pages')
```

## Summary

| In-place operator | Method | Effect |
|-------------------|--------|--------|
| `a \|= b` | `a.update(b)` | Add all of b into a |
| `a &= b` | `a.intersection_update(b)` | Keep only elements in both |
| `a -= b` | `a.difference_update(b)` | Remove elements of b from a |
| `a ^= b` | `a.symmetric_difference_update(b)` | Keep elements in exactly one |
