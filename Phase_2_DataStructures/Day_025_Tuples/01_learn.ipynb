## Quick review

A tuple is an **immutable**, **ordered** sequence. Once created, it cannot be changed.

```python
# Creating tuples
point = (3, 7)
rgb = (255, 128, 0)
single = (42,)     # trailing comma required for a one-element tuple
empty = ()

# Indexing and slicing work the same as lists
print(rgb[0])      # 255
print(rgb[-1])     # 0
print(rgb[1:])     # (128, 0)

# Unpacking
x, y = point
red, green, blue = rgb
print(f'x={x}, y={y}')
print(f'red={red}, green={green}, blue={blue}')
```

## `.count()` and `.index()`

Tuples have exactly two methods — much fewer than lists, because they can't be modified.

```python
grades = ('A', 'B', 'A', 'C', 'A', 'B', 'D')

# .count(x) — how many times x appears
print(grades.count('A'))    # 3
print(grades.count('B'))    # 2
print(grades.count('F'))    # 0

# .index(x) — position of first occurrence
print(grades.index('A'))    # 0
print(grades.index('B'))    # 1
print(grades.index('D'))    # 6

# .index() with start position — search from index 1 onwards
print(grades.index('A', 1)) # 2  — finds the second 'A'
```

## Converting between list and tuple

```python
# tuple → list (to modify)
t = (1, 2, 3, 4, 5)
as_list = list(t)
as_list.append(6)
as_list.sort(reverse=True)
print(as_list)   # [6, 5, 4, 3, 2, 1]

# list → tuple (to freeze/protect)
as_tuple = tuple(as_list)
print(as_tuple)  # (6, 5, 4, 3, 2, 1)

# String → tuple of characters
print(tuple('hello'))  # ('h', 'e', 'l', 'l', 'o')

# len, min, max, sum all work on tuples
numbers = (8, 3, 9, 1, 5)
print(len(numbers), min(numbers), max(numbers), sum(numbers))
```

## Sorting a list of tuples

When you have a list of tuples, Python sorts them **lexicographically** — first by the first element, then by the second if first elements are equal.

```python
scores = [('Alice', 88), ('Bob', 75), ('Charlie', 92), ('Diana', 75)]

# Default sort — by first element (name), alphabetical
print('By name:', sorted(scores))

# Descending by first element
print('Desc:', sorted(scores, reverse=True))

# To sort by score, rebuild as (score, name) tuples using a loop
flipped = []
for name, score in scores:
    flipped.append((score, name))
flipped.sort()   # sorts by score (first element), then name
print('\nBy score (low to high):', flipped)

flipped.sort(reverse=True)
print('By score (high to low):', flipped)
```

## Tuples as dictionary keys

Lists cannot be used as dictionary keys because they are mutable (changeable).  
Tuples **can** be dict keys because they are immutable.

```python
# Grid coordinates as keys — very common pattern
grid_labels = {
    (0, 0): 'top-left',
    (0, 1): 'top-right',
    (1, 0): 'bottom-left',
    (1, 1): 'bottom-right'
}

print(grid_labels[(0, 0)])   # top-left
print(grid_labels[(1, 1)])   # bottom-right

# Check if a coordinate exists
pos = (0, 1)
if pos in grid_labels:
    print(f'{pos} is labelled: {grid_labels[pos]}')

# Trying a list as a key would raise a TypeError:
# bad = {[0, 0]: 'error'}  # TypeError: unhashable type: 'list'
```

## When to use a tuple vs a list

| Use a **tuple** when... | Use a **list** when... |
|-------------------------|------------------------|
| The data won't change | The data may change |
| You need it as a dict key | Order matters but content evolves |
| Representing a fixed record | Accumulating items in a loop |
| Returning multiple values (later) | Removing or sorting items |

## Summary

| Operation | Syntax |
|-----------|--------|
| Count occurrences | `t.count(x)` |
| Find position | `t.index(x)` |
| To list | `list(t)` |
| To tuple | `tuple(lst)` |
| As dict key | `{(x, y): value}` |
| Sort list of tuples | `sorted(lst)` — lexicographic |
