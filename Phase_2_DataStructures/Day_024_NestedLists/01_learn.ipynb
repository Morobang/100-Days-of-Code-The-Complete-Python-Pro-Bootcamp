# Day 24: Nested Lists

A list can contain any value — including another list. When a list contains lists as its elements, it's called a **nested list** (or 2D list). This is how you represent grids, tables, and matrices in Python.

## Creating a 2D list

```python
# A 3x3 grid (3 rows, 3 columns)
grid = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(grid)           # entire 2D list
print(grid[0])        # first row: [1, 2, 3]
print(grid[1])        # second row: [4, 5, 6]
print(len(grid))      # 3 — number of rows
print(len(grid[0]))   # 3 — number of columns in row 0
```

## Accessing elements: `[row][col]`

Think of it as: first pick the row, then pick the column within that row.

```python
grid = [
    [1, 2, 3],   # row 0
    [4, 5, 6],   # row 1
    [7, 8, 9]    # row 2
]
#       ^col0 ^col1 ^col2

print(grid[0][0])   # 1  — top-left
print(grid[0][2])   # 3  — top-right
print(grid[2][0])   # 7  — bottom-left
print(grid[2][2])   # 9  — bottom-right
print(grid[1][1])   # 5  — centre

# Negative indices work too
print(grid[-1][-1]) # 9  — same as grid[2][2]
```

## Modifying nested elements

```python
seats = [
    ['O', 'O', 'O'],   # row 0 — O = open
    ['O', 'O', 'O'],   # row 1
    ['O', 'O', 'O']    # row 2
]

# Book seat at row 1, column 2
seats[1][2] = 'X'   # X = booked
seats[0][0] = 'X'

for row in seats:
    print(row)
```

## Iterating with nested `for` loops

The outer loop moves through rows; the inner loop moves through each item in the row.

```python
grid = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Print every value on its own line
for row in grid:
    for value in row:
        print(value, end=' ')
    print()   # newline after each row

print()

# Use enumerate to get row and column numbers
for row_idx, row in enumerate(grid):
    for col_idx, value in enumerate(row):
        print(f'[{row_idx}][{col_idx}] = {value}')
```

## Building a 2D list with a loop

Use nested loops to build rows and append them.

```python
# Build a 4x4 zero grid
rows = 4
cols = 4
grid = []

for r in range(rows):
    row = []
    for c in range(cols):
        row.append(0)
    grid.append(row)

for row in grid:
    print(row)
```

## Extracting a column

Rows are easy to access: `grid[0]`. Columns need a loop.

```python
grid = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Extract column 0: [1, 4, 7]
col0 = []
for row in grid:
    col0.append(row[0])
print('Column 0:', col0)

# Extract column 2: [3, 6, 9]
col2 = []
for row in grid:
    col2.append(row[2])
print('Column 2:', col2)
```

## Practical example: seating chart

```python
cinema = [
    ['O', 'O', 'X', 'O', 'O'],
    ['X', 'O', 'O', 'O', 'X'],
    ['O', 'X', 'O', 'X', 'O']
]

# Print with row labels
print('   1  2  3  4  5')
for i, row in enumerate(cinema):
    print(f'Row {i+1}:', ' '.join(row))

# Count available seats
available = 0
for row in cinema:
    for seat in row:
        if seat == 'O':
            available += 1
print(f'\nAvailable seats: {available}')
```

## Summary

| Operation | Syntax |
|-----------|--------|
| Create 2D list | `[[1,2],[3,4]]` |
| Access element | `grid[row][col]` |
| Access row | `grid[row]` |
| Modify element | `grid[row][col] = value` |
| Number of rows | `len(grid)` |
| Number of cols | `len(grid[0])` |
| Iterate all | Nested `for` loops |
