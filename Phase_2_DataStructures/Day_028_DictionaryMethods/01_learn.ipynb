# Day 28: Dictionary Methods — Deeper

Days 20 and 27 covered the essential methods. Today you add four more tools: `dict()` constructor forms, `dict.fromkeys()`, merging with `**`, and `.pop()` with a default.

## `dict()` constructor

Three ways to create the same dictionary.

```python
# 1. Literal (most common)
d1 = {'name': 'Alice', 'age': 30}

# 2. Keyword arguments — keys must be valid identifiers
d2 = dict(name='Alice', age=30)

# 3. List of (key, value) pairs
d3 = dict([('name', 'Alice'), ('age', 30)])

print(d1 == d2 == d3)   # True — all three are identical
print(d2)
```

## `dict.fromkeys()` — initialise with a default value

Creates a dict from a sequence of keys, all set to the same default value.

```python
students = ['Alice', 'Bob', 'Charlie', 'Diana']

# Create a grade book with everyone starting at 0
grades = dict.fromkeys(students, 0)
print(grades)  # {'Alice': 0, 'Bob': 0, 'Charlie': 0, 'Diana': 0}

# Default is None if you omit the second argument
flags = dict.fromkeys(['debug', 'verbose', 'quiet'])
print(flags)   # {'debug': None, 'verbose': None, 'quiet': None}

# Fill in values after creation
grades['Alice'] = 88
grades['Bob'] = 75
print(grades)
```

## `.update()` with keyword arguments

`.update()` accepts either a dict or keyword arguments — you can mix both.

```python
config = {'theme': 'light', 'language': 'english', 'font_size': 14}

# Update with a dict
config.update({'theme': 'dark'})
print(config)

# Update with keyword arguments — cleaner for a few keys
config.update(font_size=16, language='french')
print(config)

# Add a new key
config.update(show_line_numbers=True)
print(config)
```

## `.pop()` with a default — safe removal

`.pop(key)` raises `KeyError` if the key doesn't exist. `.pop(key, default)` returns the default instead.

```python
user = {'name': 'Alice', 'age': 30, 'temp_token': 'abc123'}

# Safe pop — key exists
token = user.pop('temp_token', None)
print('Token removed:', token)   # abc123
print(user)                      # temp_token gone

# Safe pop — key does NOT exist
missing = user.pop('session_id', 'no session')
print('Missing pop:', missing)   # no session — no error!

# Unsafe pop (no default) — raises KeyError
# user.pop('email')  # KeyError: 'email'
```

## Merging dicts with `**` unpacking

`{**d1, **d2}` creates a new dict combining both. Keys in `d2` overwrite matching keys in `d1`.

```python
defaults = {'theme': 'light', 'font_size': 14, 'language': 'english'}
user_prefs = {'theme': 'dark', 'font_size': 16}

# Merge: user_prefs wins on conflicts
merged = {**defaults, **user_prefs}
print(merged)
# {'theme': 'dark', 'font_size': 16, 'language': 'english'}

# Originals are unchanged
print(defaults)
print(user_prefs)

# You can also add extra keys inline
merged2 = {**defaults, **user_prefs, 'debug': True}
print(merged2)
```

## Finding the key with the highest or lowest value

```python
scores = {'Alice': 88, 'Bob': 75, 'Charlie': 92, 'Diana': 81}

top = max(scores, key=scores.get)
bottom = min(scores, key=scores.get)

print(f'Highest: {top} — {scores[top]}')
print(f'Lowest:  {bottom} — {scores[bottom]}')

# Useful with any dict of comparable values
word_counts = {'the': 12, 'cat': 3, 'sat': 5, 'on': 4}
most_common = max(word_counts, key=word_counts.get)
print(f'Most common word: "{most_common}" ({word_counts[most_common]} times)')
```

## Summary

| Tool | What it does |
|------|--------------|
| `dict(a=1, b=2)` | Create dict from keyword args |
| `dict(pairs)` | Create dict from list of tuples |
| `dict.fromkeys(keys, val)` | Create dict, all values set to `val` |
| `.update(key=val)` | Update/add using keyword args |
| `.pop(key, default)` | Remove key, return default if missing |
| `{**d1, **d2}` | Merge two dicts into a new one |
| `max(d, key=d.get)` | Key with the highest value |
