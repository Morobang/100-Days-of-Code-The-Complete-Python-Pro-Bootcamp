# Day 29: Nested Dictionaries

A dictionary value can be anything — including another dictionary. This lets you represent rich, structured data: student records, product catalogues, configuration objects. Day 20 showed the syntax briefly; today you build real fluency with it.

## Creating nested dicts

```python
# Each student maps to a dict of their details
students = {
    'Alice': {'age': 20, 'grade': 'A', 'score': 92},
    'Bob':   {'age': 21, 'grade': 'B', 'score': 78},
    'Charlie': {'age': 19, 'grade': 'A', 'score': 88}
}

print(students)
print(type(students['Alice']))   # <class 'dict'>
```

## Accessing nested values

```python
students = {
    'Alice': {'age': 20, 'grade': 'A', 'score': 92},
    'Bob':   {'age': 21, 'grade': 'B', 'score': 78},
}

# Chained [] access
print(students['Alice']['score'])   # 92
print(students['Bob']['grade'])     # B

# Safe access with .get() — avoid KeyError at either level
score = students.get('Diana', {}).get('score', 'unknown')
print(score)   # unknown — no error even though 'Diana' doesn't exist

# Store the inner dict first for readability
alice = students['Alice']
print(alice['age'])     # 20
print(alice['grade'])   # A
```

## Modifying nested values

```python
students = {
    'Alice': {'age': 20, 'grade': 'A', 'score': 92},
    'Bob':   {'age': 21, 'grade': 'B', 'score': 78},
}

# Update an existing nested value
students['Bob']['score'] = 85
students['Bob']['grade'] = 'A'
print(students['Bob'])

# Add a new key to an inner dict
students['Alice']['email'] = 'alice@uni.ac'
print(students['Alice'])

# Add an entirely new outer entry
students['Charlie'] = {'age': 22, 'grade': 'C', 'score': 65}
print(students)
```

## Iterating over a nested dict

```python
students = {
    'Alice':   {'age': 20, 'grade': 'A', 'score': 92},
    'Bob':     {'age': 21, 'grade': 'B', 'score': 78},
    'Charlie': {'age': 22, 'grade': 'C', 'score': 65}
}

# Outer loop: name and their info dict
for name, info in students.items():
    print(f'{name}:')
    for field, value in info.items():
        print(f'  {field}: {value}')

print()

# Formatted summary in one line per student
for name, info in students.items():
    print(f"{name:10} | age {info['age']} | grade {info['grade']} | score {info['score']}")
```

## Building a nested dict from user input

```python
contacts = {}

while True:
    name = input("Contact name (or 'done'): ").strip()
    if name == 'done':
        break
    phone = input('  Phone: ').strip()
    email = input('  Email: ').strip()
    contacts[name] = {'phone': phone, 'email': email}

# Look up a contact
query = input('\nLook up: ').strip()
if query in contacts:
    info = contacts[query]
    print(f"{query}: {info['phone']} | {info['email']}")
else:
    print('Not found.')
```

## Aggregating values across nested dicts

```python
students = {
    'Alice':   {'score': 92},
    'Bob':     {'score': 78},
    'Charlie': {'score': 65},
    'Diana':   {'score': 88}
}

# Average score across all students
total = 0
for info in students.values():
    total += info['score']
average = total / len(students)
print(f'Average score: {average:.1f}')

# Highest scorer
top_name = None
top_score = -1
for name, info in students.items():
    if info['score'] > top_score:
        top_score = info['score']
        top_name = name
print(f'Top scorer: {top_name} ({top_score})')
```

## Summary

| Operation | Syntax |
|-----------|--------|
| Access nested value | `d['outer']['inner']` |
| Safe nested access | `d.get('outer', {}).get('inner', default)` |
| Modify nested value | `d['outer']['inner'] = new_value` |
| Add to inner dict | `d['outer']['new_key'] = value` |
| Add outer entry | `d['new_outer'] = {inner dict}` |
| Iterate outer | `for k, v in d.items():` |
| Iterate inner | `for field, val in v.items():` |
