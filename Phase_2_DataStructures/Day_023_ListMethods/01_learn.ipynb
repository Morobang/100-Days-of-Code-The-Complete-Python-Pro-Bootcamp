## `.append()` vs `.extend()`

This is a common source of bugs. Know the difference.

```python
cart = ['milk', 'eggs']

# .append() adds ONE item — even if that item is a list
cart.append('bread')
print(cart)   # ['milk', 'eggs', 'bread']

cart.append(['butter', 'jam'])   # adds a list as a single item
print(cart)   # ['milk', 'eggs', 'bread', ['butter', 'jam']]  — nested!

# .extend() unpacks an iterable and adds each item individually
cart2 = ['milk', 'eggs']
cart2.extend(['butter', 'jam'])
print(cart2)  # ['milk', 'eggs', 'butter', 'jam']  — flat list

# extend also works with any iterable
cart2.extend('abc')   # adds 'a', 'b', 'c' as separate items
print(cart2)
```

## Three ways to remove items

| Method | Removes by | Returns |
|--------|------------|---------|
| `.remove(value)` | First matching value | Nothing (`None`) |
| `.pop(index)` | Index (default: last) | The removed item |
| `del a[i]` | Index or slice | Nothing |

```python
items = ['apple', 'banana', 'cherry', 'banana', 'date']

# .remove() — removes first occurrence of a value
items.remove('banana')   # removes the first 'banana'
print(items)  # ['apple', 'cherry', 'banana', 'date']

# .pop() — removes by index and returns the value
removed = items.pop(1)   # removes 'cherry', returns it
print('Removed:', removed)
print(items)  # ['apple', 'banana', 'date']

last = items.pop()       # no index = removes last item
print('Last removed:', last)

# del — removes by index or slice, no return value
nums = [10, 20, 30, 40, 50]
del nums[2]       # removes 30
print(nums)       # [10, 20, 40, 50]

del nums[1:3]     # removes items at index 1 and 2
print(nums)       # [10, 50]
```

## `.sort()` with `key=` and `reverse=`

```python
words = ['banana', 'apple', 'fig', 'cherry', 'kiwi', 'date']

# Default alphabetical
words.sort()
print(words)  # ['apple', 'banana', 'cherry', 'date', 'fig', 'kiwi']

# Reverse alphabetical
words.sort(reverse=True)
print(words)

# Sort by length — key= takes a function name (no parentheses)
words.sort(key=len)
print(words)  # ['fig', 'kiwi', 'date', 'apple', 'banana', 'cherry']

# Shortest to longest, then reversed (longest first)
words.sort(key=len, reverse=True)
print(words)

# Case-insensitive sort
names = ['charlie', 'Alice', 'bob', 'Diana']
names.sort(key=str.lower)
print(names)  # ['Alice', 'bob', 'charlie', 'Diana']
```

## `sorted()` vs `.sort()`

`.sort()` modifies the list in-place. `sorted()` returns a **new** sorted list and leaves the original unchanged.

```python
scores = [88, 75, 92, 61, 79]

# sorted() — returns a new list
sorted_scores = sorted(scores)
print('Original:', scores)         # unchanged
print('Sorted copy:', sorted_scores)

# sorted() with key=
words = ['banana', 'fig', 'cherry']
by_length = sorted(words, key=len)
print('By length:', by_length)
print('Original:', words)  # unchanged

# .sort() — modifies in place, returns None
scores.sort()
print('After .sort():', scores)  # scores itself changed
```

## Lists as stacks and queues

Two classic patterns that use only `append()` and `pop()`.

```python
# STACK — Last In, First Out (LIFO)
# Push with append(), pop with pop()
# Example: undo history

history = []
history.append('open file')    # push
history.append('edit text')    # push
history.append('bold word')    # push
print('History:', history)

last_action = history.pop()    # pop the most recent
print('Undone:', last_action)
print('History after undo:', history)
```

```python
# QUEUE — First In, First Out (FIFO)
# Enqueue with append(), dequeue with pop(0)
# Example: customer queue

queue = []
queue.append('Alice')    # joins queue
queue.append('Bob')
queue.append('Charlie')
print('Queue:', queue)

served = queue.pop(0)    # first person served
print('Serving:', served)
print('Queue:', queue)
```

## `.clear()`

Removes all items from the list (the list still exists — it's just empty).

```python
data = [1, 2, 3, 4, 5]
data.clear()
print(data)       # []
print(len(data))  # 0
```

## Summary

| Method | What it does |
|--------|--------------|
| `.append(x)` | Add one item to the end |
| `.extend(iter)` | Add all items from an iterable |
| `.remove(x)` | Remove first occurrence of x |
| `.pop(i)` | Remove and return item at index i (default: last) |
| `del a[i]` | Remove item at index i (no return) |
| `.sort(key=, reverse=)` | Sort in place |
| `sorted(a, key=, reverse=)` | Return new sorted list |
| `.clear()` | Empty the list |
