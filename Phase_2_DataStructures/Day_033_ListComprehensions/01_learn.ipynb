# Day 33: List Comprehensions

A **list comprehension** is a concise way to build a list from an iterable. Instead of a three-line `for` loop with `.append()`, you write one expression. It's one of the most-used Python features.

## Basic syntax

```python
[expression for item in iterable]
```

Read it as: *'for each item in iterable, evaluate expression and collect the result.'*

````python
# Loop version
squares_loop = []
for n in range(1, 6):
    squares_loop.append(n ** 2)
print(squares_loop)   # [1, 4, 9, 16, 25]

# Comprehension version — same result
squares = [n ** 2 for n in range(1, 6)]
print(squares)        # [1, 4, 9, 16, 25]

# Transform a list of strings
words = ['hello', 'world', 'python']
upper = [w.upper() for w in words]
print(upper)   # ['HELLO', 'WORLD', 'PYTHON']

# Build a list of lengths
lengths = [len(w) for w in words]
print(lengths)   # [5, 5, 6]
````

## With a filter condition

```python
[expression for item in iterable if condition]
```

Only items that pass the condition are included.

````python
numbers = [3, -1, 7, -4, 0, 8, -2, 5]

# Keep only positives
positives = [n for n in numbers if n > 0]
print(positives)   # [3, 7, 8, 5]

# Keep only even numbers
evens = [n for n in numbers if n % 2 == 0]
print(evens)   # [-4, 0, 8, -2]

# Filter and transform together
# Square only the positives
pos_squares = [n ** 2 for n in numbers if n > 0]
print(pos_squares)   # [9, 49, 64, 25]
````

## Working with strings

````python
words = ['apple', 'banana', 'fig', 'cherry', 'kiwi', 'date']

# Only words longer than 4 characters
long_words = [w for w in words if len(w) > 4]
print(long_words)   # ['apple', 'banana', 'cherry']

# Words starting with a vowel
vowels = 'aeiou'
vowel_words = [w for w in words if w[0] in vowels]
print(vowel_words)  # ['apple']

# Title-case all words
titled = [w.title() for w in words]
print(titled)
````

## Comprehensions over other iterables

Comprehensions work on any iterable — strings, tuples, ranges, even other lists.

````python
# From a string — characters that are digits
mixed = 'a1b2c3d4'
digits = [ch for ch in mixed if ch.isdigit()]
print(digits)   # ['1', '2', '3', '4']

# From a tuple
scores = (88, 75, 92, 61, 79)
passing = [s for s in scores if s >= 70]
print(passing)  # [88, 75, 92, 79]

# From a range
multiples_of_7 = [n for n in range(1, 50) if n % 7 == 0]
print(multiples_of_7)

# From .items() — extract keys where value meets a condition
stock = {'apple': 50, 'banana': 0, 'cherry': 20, 'mango': 0}
in_stock = [item for item, qty in stock.items() if qty > 0]
print(in_stock)   # ['apple', 'cherry']
````

## Nested comprehensions — flattening a 2D list

Two `for` clauses let you iterate over nested structures in one expression.

````python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# Loop version
flat_loop = []
for row in matrix:
    for val in row:
        flat_loop.append(val)

# Comprehension version
flat = [val for row in matrix for val in row]
print(flat)   # [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Build a 3x3 grid of zeros using a nested comprehension
grid = [[0 for _ in range(3)] for _ in range(3)]
for row in grid:
    print(row)
````

## When NOT to use comprehensions

Comprehensions are for **building lists**. Avoid them when you need side effects, complex logic, or the code becomes hard to read.

````python
# BAD — using a comprehension just for side effects (printing)
# [print(x) for x in range(5)]   # works but is confusing — use a for loop

# BAD — too complex to read at a glance
# result = [x**2 if x > 0 else -x for x in range(-5, 6) if x != 0]

# GOOD — simple transform or simple filter
names = ['alice', 'bob', 'charlie']
capitalised = [n.capitalize() for n in names]
print(capitalised)

# Rule of thumb: if it reads naturally as one sentence, use it.
# If you need to think hard to parse it, use a regular loop.
````

## Summary

| Pattern | Syntax |
|---------|--------|
| Transform | `[expr for x in it]` |
| Filter | `[x for x in it if cond]` |
| Transform + filter | `[expr for x in it if cond]` |
| Flatten 2D | `[v for row in matrix for v in row]` |
| Build 2D grid | `[[val for _ in range(cols)] for _ in range(rows)]` |
