# Day 34: Dictionary and Set Comprehensions

Day 33 introduced list comprehensions. The same idea extends to dictionaries and sets. The syntax is almost identical — you just use `{}` and provide a `key: value` pair for dicts.

## Dictionary comprehension — basic syntax

```python
{key_expr: val_expr for item in iterable}
```

````python
# Loop version
squares_loop = {}
for n in range(1, 6):
    squares_loop[n] = n ** 2
print(squares_loop)

# Comprehension version — same result
squares = {n: n ** 2 for n in range(1, 6)}
print(squares)   # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# Word → length mapping
words = ['apple', 'banana', 'fig', 'cherry']
word_lengths = {w: len(w) for w in words}
print(word_lengths)

# Word → reversed word
reversed_words = {w: w[::-1] for w in words}
print(reversed_words)
````

## With a filter condition

````python
scores = {'Alice': 88, 'Bob': 45, 'Charlie': 72, 'Diana': 91, 'Eve': 55}

# Keep only passing students
passing = {name: score for name, score in scores.items() if score >= 70}
print(passing)   # {'Alice': 88, 'Charlie': 72, 'Diana': 91}

# Keep only failing students
failing = {name: score for name, score in scores.items() if score < 70}
print(failing)

# Words longer than 4 characters → their length
words = ['apple', 'fig', 'banana', 'kiwi', 'cherry']
long = {w: len(w) for w in words if len(w) > 4}
print(long)
````

## Transforming an existing dict's values

````python
prices = {'apple': 1.20, 'banana': 0.50, 'cherry': 2.00}

# Apply a 10% discount
discounted = {item: round(price * 0.9, 2) for item, price in prices.items()}
print(discounted)

# Normalise scores to 0–1 range
scores = {'Alice': 80, 'Bob': 60, 'Charlie': 100}
max_score = max(scores.values())
normalised = {name: round(s / max_score, 2) for name, s in scores.items()}
print(normalised)

# Uppercase all keys
config = {'host': 'localhost', 'port': 5432}
upper_keys = {k.upper(): v for k, v in config.items()}
print(upper_keys)
````

## Inverting a dict in one line

````python
country_capital = {'France': 'Paris', 'Japan': 'Tokyo', 'Brazil': 'Brasilia'}

# Swap keys and values
capital_country = {v: k for k, v in country_capital.items()}
print(capital_country)
# {'Paris': 'France', 'Tokyo': 'Japan', 'Brasilia': 'Brazil'}

# Practical: build an index (value → key lookup)
symbol_to_name = {'H': 'Hydrogen', 'He': 'Helium', 'Li': 'Lithium'}
name_to_symbol = {v: k for k, v in symbol_to_name.items()}
print(name_to_symbol.get('Helium'))   # He
````

## Building from two lists

````python
keys   = ['name', 'age', 'city']
values = ['Alice', 30, 'Cape Town']

# Using range and index
combined = {keys[i]: values[i] for i in range(len(keys))}
print(combined)
# {'name': 'Alice', 'age': 30, 'city': 'Cape Town'}
````

## Set comprehensions

Same pattern as a list comprehension but with `{}` and no key — produces a set (unique values only).

````python
# Unique first letters from a list of words
words = ['apple', 'avocado', 'banana', 'blueberry', 'cherry']
first_letters = {w[0] for w in words}
print(first_letters)   # {'a', 'b', 'c'} — a set

# Unique lengths
unique_lengths = {len(w) for w in words}
print(unique_lengths)

# Even numbers in a range, as a set
evens = {n for n in range(20) if n % 2 == 0}
print(evens)

# Extract unique domain names from email addresses
emails = ['alice@gmail.com', 'bob@yahoo.com', 'charlie@gmail.com', 'diana@work.org']
domains = {e.split('@')[1] for e in emails}
print(domains)
````

## Choosing the right comprehension

| Result type | Use | Example |
|-------------|-----|---------|
| List (ordered, allows duplicates) | `[...]` | `[n**2 for n in nums]` |
| Dict (key→value mapping) | `{k: v ...}` | `{w: len(w) for w in words}` |
| Set (unique values, no order) | `{expr ...}` | `{w[0] for w in words}` |
